```python
# -*- coding: utf-8 -*-
"""
K-Fold training (by patient) for brain tumor classification (ResNet50)
- No-leak by patient_id (groups)
- StratifiedGroupKFold if available; fallback to GroupKFold
- Optional StyleGAN2-ADA augmentation:
    * Real-only run (no GAN)
    * GAN-balanced run: para cada fold se agrega GAN hasta igualar la clase mayor (por nº de imágenes REALES)
- Per-fold metrics + CV summary
- OOF (out-of-fold) predictions & metrics (one row per pool sample, predicted by a model that didn't see it)
- External TEST evaluation per fold + K-fold ensemble (avg probs across folds)
- Full INTTEST metrics per fold + INTTEST ensemble, and CV summary with mean±std for VAL/INTTEST/TEST
- Optional final fine-tuning epochs using only REAL images (no GAN) per fold and in full-train

Author: Jun + ChatGPT
"""

import os, re, io, json, time, random, csv, math, shutil, pickle
from pathlib import Path
from typing import List, Tuple, Dict, Optional
from dataclasses import dataclass, asdict

import numpy as np
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models

from sklearn.metrics import (
    accuracy_score, precision_recall_fscore_support,
    classification_report, confusion_matrix
)

# ====== Splitters ======
try:
    from sklearn.model_selection import StratifiedGroupKFold
    HAS_SGTK = True
except Exception:
    HAS_SGTK = False
    from sklearn.model_selection import GroupKFold

# ====== Utils ======

def set_seed(seed: int = 42):
    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True
    try:
        torch.set_float32_matmul_precision("high")
    except Exception:
        pass

def timestamp() -> str:
    return time.strftime("%Y%m%d-%H%M%S")

def extract_patient_id(p: Path) -> str:
    stem = p.stem
    token = stem.split("_")[0]
    return token.strip().lower()

IMG_EXTS = {".png", ".jpg", ".jpeg", ".bmp", ".tif", ".tiff"}

# ====== Config ======

@dataclass
class Config:
    data_root: str
    out_dir: str
    test_root: str = ""
    gan_pkl: str = ""
    images_per_class: int = 0        # sólo se usa si balance_gan_to_max=False
    truncation: float = 1.0
    internal_test_ratio: float = 0.0   # optional: holdout before CV
    image_size: int = 256
    batch_size: int = 32
    epochs: int = 15
    lr: float = 1e-3
    weight_decay: float = 1e-4
    num_workers: int = 4
    seed: int = 42
    kfolds: int = 5
    amp: bool = True
    pretrained: bool = True
    train_full_after_cv: bool = False   # optional: retrain on full pool at end

    # Fine-tuning final sólo con reales
    fine_tune_real_epochs: int = 0      # 0 = desactivado

    # NUEVO: balancear clases con GAN hasta la clase mayor de REALES
    balance_gan_to_max: bool = False

# ====== Dataset ======

class MRIDataset(Dataset):
    def __init__(self, items: List[Tuple[Path, int, str]], image_size: int = 256, aug: bool = False):
        self.items = items
        self.image_size = image_size
        self.aug = aug
        mean = [0.5, 0.5, 0.5]; std = [0.5, 0.5, 0.5]
        if aug:
            self.tf = transforms.Compose([
                transforms.RandomResizedCrop(image_size, scale=(0.9, 1.0), ratio=(0.95, 1.05)),
                transforms.RandomHorizontalFlip(),
                transforms.RandomVerticalFlip(),
                transforms.RandomRotation(10),
                transforms.ToTensor(),
            ])
        else:
            self.tf = transforms.Compose([
                transforms.Resize((image_size, image_size)),
                transforms.ToTensor(),
            ])
        self.norm = transforms.Normalize(mean=mean, std=std)

    def __len__(self):
        return len(self.items)

    def __getitem__(self, idx):
        path, y, pid = self.items[idx]
        img = Image.open(path).convert("L")
        img = self.tf(img)           # [1,H,W]
        img = img.repeat(3, 1, 1)    # [3,H,W]
        img = self.norm(img)
        return img, y, str(path), pid

# ====== Scan & Helpers ======

def scan_folder(root: Path) -> Tuple[List[Tuple[Path,int,str]], List[str]]:
    assert root.is_dir(), f"Not a directory: {root}"
    class_names = sorted([d.name for d in root.iterdir() if d.is_dir()])
    items: List[Tuple[Path,int,str]] = []
    for ci, cname in enumerate(class_names):
        cdir = root / cname
        for p in cdir.rglob("*"):
            if p.suffix.lower() in IMG_EXTS:
                pid = extract_patient_id(p)
                items.append((p, ci, pid))
    return items, class_names

def ensure_no_overlap(ids_a: List[str], ids_b: List[str]) -> Tuple[int, List[str]]:
    a = set(ids_a); b = set(ids_b)
    inter = sorted(a.intersection(b))
    return len(inter), inter

def summarize_id_counts(items: List[Tuple[Path,int,str]], class_names: List[str], split_name: str, out_dir: Path):
    from collections import defaultdict
    ids_by_class = defaultdict(set); imgs_by_class = defaultdict(int); all_ids = set()
    for p, ci, pid in items:
        ids_by_class[ci].add(pid)
        imgs_by_class[ci] += 1
        all_ids.add(pid)

    print(f"[ID-STATS] {split_name}: unique patient IDs total = {len(all_ids)} | images = {len(items)}")
    for ci, cname in enumerate(class_names):
        print(f"  - {cname}: ids={len(ids_by_class[ci])} images={imgs_by_class[ci]}")

    out_dir.mkdir(parents=True, exist_ok=True)
    with open(out_dir / f"id_counts_{split_name}.csv", 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(["class", "n_unique_ids", "n_images"])
        for ci, cname in enumerate(class_names):
            w.writerow([cname, len(ids_by_class[ci]), imgs_by_class[ci]])
        w.writerow(["__TOTAL__", len(all_ids), len(items)])

# ====== Model ======

def build_model(num_classes: int, pretrained: bool = True) -> nn.Module:
    m = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1 if pretrained else None)
    in_features = m.fc.in_features
    m.fc = nn.Linear(in_features, num_classes)
    return m

# ====== Eval (writes artifacts) ======

@torch.no_grad()
def evaluate(model: nn.Module, loader: DataLoader, device: torch.device, class_names: List[str], split_name: str, out_dir: Path):
    model.eval()
    all_probs, all_preds, all_labels = [], [], []
    all_paths, all_pids = [], []

    for imgs, ys, paths, pids in loader:
        imgs = imgs.to(device, non_blocking=True)
        ys = ys.to(device, non_blocking=True)
        logits = model(imgs)
        probs = F.softmax(logits, dim=1)
        preds = probs.argmax(dim=1)

        all_probs.append(probs.cpu().numpy())
        all_preds.append(preds.cpu().numpy())
        all_labels.append(ys.cpu().numpy())
        all_paths.extend(list(paths))
        all_pids.extend(list(pids))

    probs = np.concatenate(all_probs, axis=0) if all_probs else np.zeros((0, len(class_names)))
    preds = np.concatenate(all_preds, axis=0) if all_preds else np.zeros((0,), dtype=np.int64)
    labels = np.concatenate(all_labels, axis=0) if all_labels else np.zeros((0,), dtype=np.int64)

    acc = float(accuracy_score(labels, preds)) if len(labels) else 0.0
    prec_macro, rec_macro, f1_macro, _ = precision_recall_fscore_support(labels, preds, average='macro', zero_division=0)
    prec_weighted, rec_weighted, f1_weighted, _ = precision_recall_fscore_support(labels, preds, average='weighted', zero_division=0)
    per_class = classification_report(labels, preds, target_names=class_names, output_dict=True, zero_division=0)
    cm = confusion_matrix(labels, preds, labels=list(range(len(class_names))))

    metrics = {
        "overall": {
            "accuracy": acc,
            "precision_macro": prec_macro,
            "recall_macro": rec_macro,
            "f1_macro": f1_macro,
            "precision_weighted": prec_weighted,
            "recall_weighted": rec_weighted,
            "f1_weighted": f1_weighted,
            "num_samples": len(labels),
        },
        "per_class": per_class,
    }

    out_dir.mkdir(parents=True, exist_ok=True)
    with open(out_dir / f"metrics_{split_name}.json", 'w', encoding='utf-8') as f:
        json.dump(metrics, f, indent=2)

    with open(out_dir / f"metrics_overall_{split_name}.csv", 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(["split","accuracy","precision_macro","recall_macro","f1_macro","precision_weighted","recall_weighted","f1_weighted","num_samples"])
        o = metrics['overall']
        w.writerow([split_name, o['accuracy'], o['precision_macro'], o['recall_macro'], o['f1_macro'],
                    o['precision_weighted'], o['recall_weighted'], o['f1_weighted'], o['num_samples']])

    with open(out_dir / f"per_class_{split_name}.csv", 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(["class", "precision", "recall", "f1", "support"])
        for cname in class_names:
            row = per_class.get(cname, {"precision":0, "recall":0, "f1-score":0, "support":0})
            w.writerow([cname, row.get("precision",0), row.get("recall",0),
                        row.get("f1-score",0), row.get("support",0)])

    with open(out_dir / f"confusion_{split_name}.csv", 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(["true\\pred"] + class_names)
        for i, cname in enumerate(class_names):
            w.writerow([cname] + list(cm[i]))

    with open(out_dir / f"predictions_{split_name}.csv", 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(["path", "patient_id", "true", "pred"] + [f"p_{c}" for c in class_names])
        for i in range(len(labels)):
            w.writerow([all_paths[i], all_pids[i], class_names[labels[i]],
                        class_names[preds[i]]] + list(map(float, probs[i].tolist())))

    return metrics, cm

# Helper: fast inference to collect arrays (for OOF & ensemble)
@torch.no_grad()
def infer_loader(model: nn.Module, loader: DataLoader, device: torch.device):
    model.eval()
    paths, pids = [], []
    all_probs, all_preds, all_labels = [], [], []
    for imgs, ys, batch_paths, batch_pids in loader:
        imgs = imgs.to(device, non_blocking=True)
        ys = ys.to(device, non_blocking=True)
        logits = model(imgs)
        probs = F.softmax(logits, dim=1)
        preds = probs.argmax(dim=1)

        all_probs.append(probs.cpu().numpy())
        all_preds.append(preds.cpu().numpy())
        all_labels.append(ys.cpu().numpy())
        paths.extend(list(batch_paths))
        pids.extend(list(batch_pids))

    probs = np.concatenate(all_probs, axis=0) if all_probs else np.zeros((0, 0))
    preds = np.concatenate(all_preds, axis=0) if all_preds else np.zeros((0,), dtype=np.int64)
    labels = np.concatenate(all_labels, axis=0) if all_labels else np.zeros((0,), dtype=np.int64)
    return {"probs": probs, "preds": preds, "labels": labels, "paths": paths, "pids": pids}

# ====== Optional GAN loader (TRAIN only) ======

@torch.no_grad()
def _load_stylegan2_ada_G(gan_pkl: str, device: torch.device):
    try:
        import dnnlib
        from training import network as _dummy
        from legacy import load_network_pkl
    except Exception:
        try:
            from torch_utils import misc as _m  # noqa
            from torch_utils import training_stats as _ts  # noqa
            from legacy import load_network_pkl
        except Exception:
            return None
    try:
        with open(gan_pkl, 'rb') as f:
            data = load_network_pkl(f)
        G_ema = data.get('G_ema', None) or data.get('G', None)
        if G_ema is None:
            return None
        G_ema = G_ema.to(device)
        G_ema.eval()
        return G_ema
    except Exception:
        return None

@torch.no_grad()
def maybe_generate_gan_images(gan_pkl: str,
                              class_names: List[str],
                              n_per_class: int,
                              truncation: float,
                              out_dir: Path,
                              device: torch.device,
                              seed: int = 42) -> Optional[Path]:
    if (not gan_pkl) or n_per_class <= 0:
        return None
    G = _load_stylegan2_ada_G(gan_pkl, device)
    if G is None:
        print("[GAN] Could not load StyleGAN2-ADA G_ema. Skipping GAN augmentation.")
        return None
    if getattr(G, 'c_dim', 0) <= 0:
        print("[GAN] Generator is not conditional (c_dim=0). Skipping.")
        return None

    rng = np.random.RandomState(seed)
    synth_root = out_dir / "gan_synth"
    for cname in class_names:
        (synth_root / cname).mkdir(parents=True, exist_ok=True)

    batch = 16
    for ci, cname in enumerate(class_names):
        saved = 0
        while saved < n_per_class:
            bs = min(batch, n_per_class - saved)
            z = torch.from_numpy(rng.randn(bs, G.z_dim)).to(device)
            c = torch.zeros(bs, G.c_dim, device=device); c[:, ci] = 1.0
            try:
                img = G(z, c, truncation_psi=float(truncation), noise_mode='const')
            except TypeError:
                img = G(z, c, truncation_psi=float(truncation))
            img = (img * 127.5 + 128).clamp(0,255).to(torch.uint8).cpu()
            for k in range(bs):
                x = img[k]
                if x.shape[0] == 1:
                    arr = x[0].numpy()
                    pil = Image.fromarray(arr).convert('L')
                else:
                    arr = x.permute(1,2,0).numpy()
                    pil = Image.fromarray(arr).convert('L')
                fname = synth_root / cname / f"gan_{cname}_{saved+k:05d}.png"
                pil.save(fname)
            saved += bs
        print(f"[GAN] Saved {saved} images for class '{cname}'")
    return synth_root

def collate_fn(batch):
    imgs, ys, paths, pids = zip(*batch)
    return torch.stack(imgs, dim=0), torch.tensor(ys, dtype=torch.long), list(paths), list(pids)

# ====== Helper: construir subset balanceado real + GAN ======

def build_balanced_train_subset(real_train_subset: List[Tuple[Path,int,str]],
                                synth_items: List[Tuple[Path,int,str]],
                                num_classes: int,
                                seed: int) -> List[Tuple[Path,int,str]]:
    """
    Toma:
      - real_train_subset: sólo imágenes reales de ese fold
      - synth_items: TODAS las sintéticas (escaneadas desde gan_synth/)
    Devuelve:
      - lista con reales + suficientes GAN por clase para igualar la clase mayor (en REALES).
    """
    from collections import defaultdict

    # Reales por clase
    real_by_class: Dict[int, List[Tuple[Path,int,str]]] = defaultdict(list)
    for item in real_train_subset:
        _, ci, _ = item
        real_by_class[ci].append(item)

    # GAN por clase
    synth_by_class: Dict[int, List[Tuple[Path,int,str]]] = defaultdict(list)
    for item in synth_items:
        _, ci, _ = item
        synth_by_class[ci].append(item)

    # Target = tamaño de la clase mayor (sólo contando REALES)
    counts_real = [len(real_by_class[c]) for c in range(num_classes)]
    target = max(counts_real) if counts_real else 0
    print(f"[BAL] Real class counts (fold): {counts_real} -> target={target}")

    rng = random.Random(seed)
    balanced: List[Tuple[Path,int,str]] = list(real_train_subset)

    for ci in range(num_classes):
        need = target - len(real_by_class[ci])
        if need <= 0:
            continue
        pool = synth_by_class[ci]
        if not pool:
            print(f"[BAL] WARNING: no GAN images for class {ci}, cannot fully balance.")
            continue
        # muestramos con reemplazo si need > len(pool)
        for _ in range(need):
            balanced.append(rng.choice(pool))

    return balanced

# ====== K-Fold Train (CV + OOF + TEST + INTTEST ensemble) ======

def train_kfold(cfg: Config):
    if cfg.fine_tune_real_epochs < 0 or cfg.fine_tune_real_epochs > cfg.epochs:
        raise ValueError("fine_tune_real_epochs must be between 0 and cfg.epochs")

    base_run = Path(cfg.out_dir) / f"kfold_{cfg.kfolds}f_run_{timestamp()}"
    base_run.mkdir(parents=True, exist_ok=True)
    with open(base_run / 'config.json', 'w', encoding='utf-8') as f:
        json.dump(asdict(cfg), f, indent=2)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"[Device] Using {device}")

    # Scan pool
    all_items, class_names = scan_folder(Path(cfg.data_root))
    print(f"[Scan] Pool images: {len(all_items)} | Classes: {class_names}")

    # Optional internal-test holdout (by patient)
    inttest_subset: List[Tuple[Path,int,str]] = []
    if cfg.internal_test_ratio and cfg.internal_test_ratio > 0:
        from sklearn.model_selection import GroupShuffleSplit
        X_all = np.arange(len(all_items))
        y_all = np.array([lbl for _, lbl, _ in all_items])
        g_all = np.array([pid for _, _, pid in all_items])
        gss = GroupShuffleSplit(n_splits=1, test_size=cfg.internal_test_ratio, random_state=cfg.seed)
        trval_idx, inttest_idx = next(gss.split(X_all, y_all, g_all))
        base_pool = [all_items[i] for i in trval_idx]
        inttest_subset = [all_items[i] for i in inttest_idx]
        print(f"[Split] Internal test: {len(inttest_subset)} images ({cfg.internal_test_ratio*100:.1f}%)")
    else:
        base_pool = all_items

    # External test (leak-sanitized vs pool & internal test)
    ext_subset: List[Tuple[Path,int,str]] = []
    if cfg.test_root and len(cfg.test_root.strip()) > 0 and Path(cfg.test_root).exists():
        ext_items, _ = scan_folder(Path(cfg.test_root))
        ids_pool = set(pid for _,_,pid in base_pool)
        if len(inttest_subset) > 0:
            ids_pool |= set(pid for _,_,pid in inttest_subset)
        for it in ext_items:
            if it[2] not in ids_pool:
                ext_subset.append(it)
        removed = len(ext_items) - len(ext_subset)
        if removed > 0:
            print(f"[LEAK-FIX] Removed {removed} overlapping ext-test images by patient ID.")
        print(f"[Scan] External test images (sanitized): {len(ext_subset)}")

    # Build folds grouped by patient (SOLO con REALES)
    X = np.arange(len(base_pool))
    y = np.array([lbl for _, lbl, _ in base_pool])
    groups = np.array([pid for _, _, pid in base_pool])

    if cfg.kfolds < 2:
        raise ValueError("kfolds must be >= 2")

    if HAS_SGTK:
        splitter = StratifiedGroupKFold(n_splits=cfg.kfolds, shuffle=True, random_state=cfg.seed)
        folds_iter = list(splitter.split(X, y, groups))
        print("[Split] Using StratifiedGroupKFold")
    else:
        splitter = GroupKFold(n_splits=cfg.kfolds)
        folds_iter = list(splitter.split(X, groups=groups))
        print("[Split] Using GroupKFold (no stratification)")

    # ===== Pre-generate GAN synth una sola vez =====
    synth_items: List[Tuple[Path,int,str]] = []
    synth_root: Optional[Path] = None
    if cfg.gan_pkl and cfg.gan_pkl.strip():
        # si queremos balancear, ignoramos images_per_class y usamos el máximo global de REALES
        if cfg.balance_gan_to_max:
            from collections import defaultdict
            real_counts_global = defaultdict(int)
            for _, ci, _ in base_pool:
                real_counts_global[ci] += 1
            # Vector ordenado por índice de clase
            counts_vec = [real_counts_global[ci] for ci in range(len(class_names))]
            n_per_class = max(counts_vec) if counts_vec else 0
            print(f"[GAN] balance_gan_to_max=True -> global real counts per class = {counts_vec}, using n_per_class={n_per_class}")
        else:
            n_per_class = cfg.images_per_class

        if n_per_class > 0:
            synth_root = maybe_generate_gan_images(
                gan_pkl=cfg.gan_pkl,
                class_names=class_names,
                n_per_class=n_per_class,
                truncation=cfg.truncation,
                out_dir=base_run,
                device=device,
                seed=cfg.seed,
            )
            if synth_root is not None:
                synth_items, _ = scan_folder(synth_root)
                print(f"[GAN] Synth images generated (global): {len(synth_items)}")
        else:
            print("[GAN] n_per_class <= 0 -> no synthetic images will be generated.")
    else:
        print("[GAN] gan_pkl empty -> NO GAN will be used.")

    # Prepare fixed loaders for ensembles (TEST e INTTEST)
    test_loader_fixed: Optional[DataLoader] = None
    test_ref_paths = test_ref_pids = test_ref_labels = None
    if len(ext_subset) > 0:
        test_ds_fixed = MRIDataset(ext_subset, image_size=cfg.image_size, aug=False)
        test_loader_fixed = DataLoader(test_ds_fixed, batch_size=cfg.batch_size, shuffle=False,
                                       num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)
        # Se rellenan luego con infer_loader

    inttest_loader_fixed: Optional[DataLoader] = None
    int_ref_paths = int_ref_pids = int_ref_labels = None
    if len(inttest_subset) > 0:
        int_ds_fixed = MRIDataset(inttest_subset, image_size=cfg.image_size, aug=False)
        inttest_loader_fixed = DataLoader(int_ds_fixed, batch_size=cfg.batch_size, shuffle=False,
                                          num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)

    # ===== OOF buffers =====
    n_pool = len(base_pool)
    n_cls = len(class_names)
    oof_probs = np.full((n_pool, n_cls), np.nan, dtype=np.float32)
    oof_pred  = np.full((n_pool,), -1, dtype=np.int64)
    oof_true  = np.full((n_pool,), -1, dtype=np.int64)
    oof_path  = [""] * n_pool
    oof_pid   = [""] * n_pool

    # For ensembles
    test_probs_folds: List[np.ndarray] = []
    inttest_probs_folds: List[np.ndarray] = []

    # Per-fold training
    cv_rows = []
    for fold_id, (tr_idx, va_idx) in enumerate(folds_iter, start=1):
        print(f"\n===== FOLD {fold_id}/{cfg.kfolds} =====")
        set_seed(cfg.seed + fold_id)

        fold_dir = base_run / f"fold_{fold_id:02d}"
        fold_dir.mkdir(parents=True, exist_ok=True)

        # REAL train subset and VAL subset
        real_train_subset = [base_pool[i] for i in tr_idx]
        val_subset        = [base_pool[i] for i in va_idx]

        # Leak checks (sólo reales)
        ids_train = [pid for _,_,pid in real_train_subset]
        ids_val   = [pid for _,_,pid in val_subset]
        inter_n, inter_ids = ensure_no_overlap(ids_train, ids_val)
        if inter_n > 0:
            raise RuntimeError(f"[LEAK ERROR] Train↔Val share {inter_n} patient IDs (fold {fold_id}): {inter_ids[:10]}")

        if len(inttest_subset) > 0:
            ids_int = [pid for _,_,pid in inttest_subset]
            n1,_ = ensure_no_overlap(ids_train, ids_int)
            n2,_ = ensure_no_overlap(ids_val, ids_int)
            if n1>0 or n2>0:
                raise RuntimeError("[LEAK ERROR] Fold and INTTEST share patients")

        # ===== construir train_subset =====
        if cfg.balance_gan_to_max and len(synth_items) > 0:
            train_subset = build_balanced_train_subset(
                real_train_subset=real_train_subset,
                synth_items=synth_items,
                num_classes=len(class_names),
                seed=cfg.seed + fold_id,
            )
            print(f"[BAL] Fold {fold_id}: real={len(real_train_subset)} | train_balanced(real+GAN)={len(train_subset)}")
        else:
            # sin balanceo: opcionalmente append de todo synth_items
            train_subset = list(real_train_subset)
            if len(synth_items) > 0:
                train_subset = train_subset + synth_items
                print(f"[TRAIN] Fold {fold_id}: real={len(real_train_subset)} + synth={len(synth_items)} -> {len(train_subset)}")
            else:
                print(f"[TRAIN] Fold {fold_id}: only real images, n={len(real_train_subset)}")

        # Stats CSV
        summarize_id_counts(train_subset, class_names, f'fold{fold_id}_train', fold_dir)
        summarize_id_counts(val_subset,   class_names, f'fold{fold_id}_val',   fold_dir)
        if len(inttest_subset) > 0:
            summarize_id_counts(inttest_subset, class_names, f'fold{fold_id}_inttest', fold_dir)
        if len(ext_subset) > 0:
            summarize_id_counts(ext_subset,     class_names, f'fold{fold_id}_test',    fold_dir)

        # Loaders
        train_ds       = MRIDataset(train_subset,       image_size=cfg.image_size, aug=True)
        real_train_ds  = MRIDataset(real_train_subset,  image_size=cfg.image_size, aug=True)
        val_ds         = MRIDataset(val_subset,         image_size=cfg.image_size, aug=False)

        train_loader      = DataLoader(train_ds,      batch_size=cfg.batch_size, shuffle=True,
                                       num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)
        real_train_loader = DataLoader(real_train_ds, batch_size=cfg.batch_size, shuffle=True,
                                       num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)
        val_loader        = DataLoader(val_ds,        batch_size=cfg.batch_size, shuffle=False,
                                       num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)

        inttest_loader = None
        if len(inttest_subset) > 0:
            int_ds = MRIDataset(inttest_subset, image_size=cfg.image_size, aug=False)
            inttest_loader = DataLoader(int_ds, batch_size=cfg.batch_size, shuffle=False,
                                        num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)

        # Para TEST por fold usamos loader fijo
        test_loader = test_loader_fixed

        # Model/optim
        model = build_model(num_classes=len(class_names), pretrained=cfg.pretrained).to(device)
        optimizer = torch.optim.AdamW(model.parameters(), lr=cfg.lr, weight_decay=cfg.weight_decay)
        scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=cfg.epochs)
        criterion = nn.CrossEntropyLoss()

        amp_enabled = (cfg.amp and device.type == 'cuda')
        scaler = torch.amp.GradScaler('cuda', enabled=amp_enabled)

        best_f1 = -1.0

        # Epoch split: main (train_subset) + fine-tuning (real only)
        epochs_main = cfg.epochs - cfg.fine_tune_real_epochs
        epochs_ft   = cfg.fine_tune_real_epochs

        # ----- Phase 1: train con train_subset (real + GAN o sólo real) -----
        for epoch in range(1, epochs_main + 1):
            model.train()
            running_loss = 0.0; n_seen = 0
            for imgs, ys, _, _ in train_loader:
                imgs = imgs.to(device, non_blocking=True)
                ys = ys.to(device, non_blocking=True)

                optimizer.zero_grad(set_to_none=True)
                with torch.amp.autocast('cuda', enabled=amp_enabled):
                    logits = model(imgs)
                    loss = criterion(logits, ys)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()

                running_loss += float(loss.item()) * imgs.size(0)
                n_seen += imgs.size(0)

            scheduler.step()
            train_loss = running_loss / max(1, n_seen)

            m_val, _ = evaluate(model, val_loader, device, class_names,
                                split_name='val', out_dir=fold_dir)
            f1_macro = m_val['overall']['f1_macro']
            acc_val  = m_val['overall']['accuracy']
            print(f"[Fold {fold_id} | Epoch {epoch:03d}] "
                  f"train_loss={train_loss:.4f} | val_acc={acc_val:.4f} | val_f1_macro={f1_macro:.4f}")

            torch.save({'model': model.state_dict(), 'epoch': epoch, 'class_names': class_names},
                       fold_dir / 'last.ckpt')

            if f1_macro > best_f1:
                best_f1 = f1_macro
                torch.save({'model': model.state_dict(), 'epoch': epoch, 'class_names': class_names},
                           fold_dir / 'best.ckpt')
                print(f"[Fold {fold_id}] New best (F1_macro={best_f1:.4f}) saved.")

        # ----- Phase 2: fine-tuning sólo con REALES -----
        for ft_idx in range(1, epochs_ft + 1):
            epoch = epochs_main + ft_idx
            model.train()
            running_loss = 0.0; n_seen = 0
            for imgs, ys, _, _ in real_train_loader:
                imgs = imgs.to(device, non_blocking=True)
                ys = ys.to(device, non_blocking=True)

                optimizer.zero_grad(set_to_none=True)
                with torch.amp.autocast('cuda', enabled=amp_enabled):
                    logits = model(imgs)
                    loss = criterion(logits, ys)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()

                running_loss += float(loss.item()) * imgs.size(0)
                n_seen += imgs.size(0)

            scheduler.step()
            train_loss = running_loss / max(1, n_seen)

            m_val, _ = evaluate(model, val_loader, device, class_names,
                                split_name='val', out_dir=fold_dir)
            f1_macro = m_val['overall']['f1_macro']
            acc_val  = m_val['overall']['accuracy']
            print(f"[Fold {fold_id} | FT Epoch {epoch:03d}] "
                  f"train_loss={train_loss:.4f} | val_acc={acc_val:.4f} | val_f1_macro={f1_macro:.4f}")

            torch.save({'model': model.state_dict(), 'epoch': epoch, 'class_names': class_names},
                       fold_dir / 'last.ckpt')

            if f1_macro > best_f1:
                best_f1 = f1_macro
                torch.save({'model': model.state_dict(), 'epoch': epoch, 'class_names': class_names},
                           fold_dir / 'best.ckpt')
                print(f"[Fold {fold_id}] New best (F1_macro={best_f1:.4f}) saved (fine-tune).")

        # Re-eval BEST and collect OOF for this fold
        best = torch.load(fold_dir / 'best.ckpt', map_location=device)
        model.load_state_dict(best['model'])

        print(f"[Fold {fold_id}] Re-evaluating BEST on VAL (and collecting OOF)...")
        m_val_best, _ = evaluate(model, val_loader, device, class_names,
                                 split_name='val_best', out_dir=fold_dir)

        # Collect arrays for OOF
        val_inf = infer_loader(model, val_loader, device)  # same order as val_subset (shuffle=False)
        for j, base_i in enumerate(va_idx):
            oof_probs[base_i] = val_inf["probs"][j]
            oof_pred[base_i]  = int(val_inf["preds"][j])
            oof_true[base_i]  = int(val_inf["labels"][j])
            oof_path[base_i]  = val_inf["paths"][j]
            oof_pid[base_i]   = val_inf["pids"][j]

        # Row base with full VAL metrics
        row = {
            "fold": fold_id,
            "val_acc": m_val_best["overall"]["accuracy"],
            "val_precision_macro": m_val_best["overall"]["precision_macro"],
            "val_recall_macro": m_val_best["overall"]["recall_macro"],
            "val_f1_macro": m_val_best["overall"]["f1_macro"],
            "val_precision_weighted": m_val_best["overall"]["precision_weighted"],
            "val_recall_weighted": m_val_best["overall"]["recall_weighted"],
            "val_f1_weighted": m_val_best["overall"]["f1_weighted"],
            "val_n": m_val_best["overall"]["num_samples"],
        }

        # INTERNAL TEST per fold + probs for ensemble
        if inttest_loader is not None:
            print(f"[Fold {fold_id}] Evaluating on INTERNAL TEST...")
            m_int, _ = evaluate(model, inttest_loader, device, class_names,
                                split_name='inttest', out_dir=fold_dir)
            row.update({
                "int_acc": m_int["overall"]["accuracy"],
                "int_precision_macro": m_int["overall"]["precision_macro"],
                "int_recall_macro": m_int["overall"]["recall_macro"],
                "int_f1_macro": m_int["overall"]["f1_macro"],
                "int_precision_weighted": m_int["overall"]["precision_weighted"],
                "int_recall_weighted": m_int["overall"]["recall_weighted"],
                "int_f1_weighted": m_int["overall"]["f1_weighted"],
                "int_n": m_int["overall"]["num_samples"],
            })
            t_int = infer_loader(model, inttest_loader, device)
            inttest_probs_folds.append(t_int["probs"])
            if int_ref_paths is None:
                int_ref_paths  = t_int["paths"]
                int_ref_pids   = t_int["pids"]
                int_ref_labels = t_int["labels"]

        # EXTERNAL TEST per fold + probs for ensemble
        if test_loader is not None:
            print(f"[Fold {fold_id}] Evaluating on EXTERNAL TEST...")
            m_ext, _ = evaluate(model, test_loader, device, class_names,
                                split_name='test', out_dir=fold_dir)
            row.update({
                "test_acc": m_ext["overall"]["accuracy"],
                "test_precision_macro": m_ext["overall"]["precision_macro"],
                "test_recall_macro": m_ext["overall"]["recall_macro"],
                "test_f1_macro": m_ext["overall"]["f1_macro"],
                "test_precision_weighted": m_ext["overall"]["precision_weighted"],
                "test_recall_weighted": m_ext["overall"]["recall_weighted"],
                "test_f1_weighted": m_ext["overall"]["f1_weighted"],
                "test_n": m_ext["overall"]["num_samples"],
            })
            t_inf = infer_loader(model, test_loader, device)
            test_probs_folds.append(t_inf["probs"])
            if test_ref_paths is None:
                test_ref_paths  = t_inf["paths"]
                test_ref_pids   = t_inf["pids"]
                test_ref_labels = t_inf["labels"]

        cv_rows.append(row)

    # === CV summary (mean ± std) ===
    summary_csv = base_run / "cv_summary.csv"
    keys = sorted(set().union(*[r.keys() for r in cv_rows]))
    with open(summary_csv, 'w', newline='', encoding='utf-8') as f:
        w = csv.DictWriter(f, fieldnames=keys)
        w.writeheader()
        for r in cv_rows:
            w.writerow(r)

    def _agg_for(prefix: str) -> Dict[str, Tuple[Optional[float], Optional[float]]]:
        metrics = ["acc","precision_macro","recall_macro","f1_macro",
                   "precision_weighted","recall_weighted","f1_weighted"]
        out: Dict[str, Tuple[Optional[float], Optional[float]]] = {}
        for m in metrics:
            col = f"{prefix}_{m}"
            vals = [float(r[col]) for r in cv_rows if col in r]
            if len(vals) == 0:
                out[m] = (None, None)
            else:
                out[m] = (float(np.mean(vals)), float(np.std(vals)))
        return out

    cv_stats_full = {
        "kfolds": cfg.kfolds,
        "val":     _agg_for("val"),
        "inttest": _agg_for("int") if any(("int_acc" in r) for r in cv_rows) else {},
        "test":    _agg_for("test") if any(("test_acc" in r) for r in cv_rows) else {},
    }
    with open(base_run / "cv_summary_full.json", 'w', encoding='utf-8') as f:
        json.dump(cv_stats_full, f, indent=2)

    with open(base_run / "cv_summary_full.csv", 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(["split","metric","mean","std","k"])
        for split in ["val","inttest","test"]:
            if split not in cv_stats_full or not cv_stats_full[split]:
                continue
            for metric, (mu, sd) in cv_stats_full[split].items():
                if mu is None:
                    continue
                w.writerow([split, metric, mu, sd, cfg.kfolds])

    print("[CV] Summary (mean±std) written to cv_summary_full.{json,csv}")

    # === OOF metrics & artifacts ===
    valid_mask = (~np.isnan(oof_probs).any(axis=1)) & (oof_true != -1)
    oof_probs_v = oof_probs[valid_mask]
    oof_pred_v  = oof_pred[valid_mask]
    oof_true_v  = oof_true[valid_mask]
    oof_path_v  = [oof_path[i] for i, ok in enumerate(valid_mask) if ok]
    oof_pid_v   = [oof_pid[i]  for i, ok in enumerate(valid_mask) if ok]

    if len(oof_true_v) > 0:
        acc = float(accuracy_score(oof_true_v, oof_pred_v))
        prec_m, rec_m, f1_m, _ = precision_recall_fscore_support(
            oof_true_v, oof_pred_v, average='macro', zero_division=0)
        prec_w, rec_w, f1_w, _ = precision_recall_fscore_support(
            oof_true_v, oof_pred_v, average='weighted', zero_division=0)
        per_class = classification_report(
            oof_true_v, oof_pred_v, target_names=class_names,
            output_dict=True, zero_division=0)
        cm = confusion_matrix(oof_true_v, oof_pred_v, labels=list(range(len(class_names))))

        oof_metrics = {
            "overall": {
                "accuracy": acc,
                "precision_macro": prec_m,
                "recall_macro": rec_m,
                "f1_macro": f1_m,
                "precision_weighted": prec_w,
                "recall_weighted": rec_w,
                "f1_weighted": f1_w,
                "num_samples": int(len(oof_true_v)),
            },
            "per_class": per_class,
        }
        with open(base_run / "oof_metrics.json", 'w', encoding='utf-8') as f:
            json.dump(oof_metrics, f, indent=2)

        with open(base_run / "metrics_overall_oof.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["split","accuracy","precision_macro","recall_macro","f1_macro",
                        "precision_weighted","recall_weighted","f1_weighted","num_samples"])
            o = oof_metrics['overall']
            w.writerow(["oof", o['accuracy'], o['precision_macro'], o['recall_macro'],
                        o['f1_macro'], o['precision_weighted'], o['recall_weighted'],
                        o['f1_weighted'], o['num_samples']])

        with open(base_run / "per_class_oof.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["class","precision","recall","f1","support"])
            for cname in class_names:
                row = per_class.get(cname, {"precision":0, "recall":0,
                                            "f1-score":0, "support":0})
                w.writerow([cname, row.get("precision",0), row.get("recall",0),
                            row.get("f1-score",0), row.get("support",0)])

        with open(base_run / "confusion_oof.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["true\\pred"] + class_names)
            for i, cname in enumerate(class_names):
                w.writerow([cname] + list(cm[i]))

        with open(base_run / "oof_predictions.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["path","patient_id","true","pred"] +
                       [f"p_{c}" for c in class_names])
            for i in range(len(oof_true_v)):
                w.writerow([oof_path_v[i], oof_pid_v[i],
                            class_names[oof_true_v[i]],
                            class_names[oof_pred_v[i]]] +
                           list(map(float, oof_probs_v[i].tolist())))

    # === TEST ensemble (avg across folds) ===
    if test_loader_fixed is not None and len(test_probs_folds) > 0:
        probs_stack = np.stack(test_probs_folds, axis=0)     # [K, N_test, C]
        probs_mean  = probs_stack.mean(axis=0)               # [N_test, C]
        preds_ens   = probs_mean.argmax(axis=1)

        acc = float(accuracy_score(test_ref_labels, preds_ens))
        prec_m, rec_m, f1_m, _ = precision_recall_fscore_support(
            test_ref_labels, preds_ens, average='macro', zero_division=0)
        prec_w, rec_w, f1_w, _ = precision_recall_fscore_support(
            test_ref_labels, preds_ens, average='weighted', zero_division=0)
        per_class = classification_report(
            test_ref_labels, preds_ens, target_names=class_names,
            output_dict=True, zero_division=0)
        cm = confusion_matrix(test_ref_labels, preds_ens, labels=list(range(len(class_names))))

        test_ens_metrics = {
            "overall": {
                "accuracy": acc,
                "precision_macro": prec_m,
                "recall_macro": rec_m,
                "f1_macro": f1_m,
                "precision_weighted": prec_w,
                "recall_weighted": rec_w,
                "f1_weighted": f1_w,
                "num_samples": int(len(test_ref_labels)),
            },
            "per_class": per_class,
        }
        with open(base_run / "metrics_test_ensemble.json", 'w', encoding='utf-8') as f:
            json.dump(test_ens_metrics, f, indent=2)

        with open(base_run / "metrics_overall_test_ensemble.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["split","accuracy","precision_macro","recall_macro","f1_macro",
                        "precision_weighted","recall_weighted","f1_weighted","num_samples"])
            o = test_ens_metrics['overall']
            w.writerow(["test_ensemble", o['accuracy'], o['precision_macro'],
                        o['recall_macro'], o['f1_macro'], o['precision_weighted'],
                        o['recall_weighted'], o['f1_weighted'], o['num_samples']])

        with open(base_run / "per_class_test_ensemble.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["class","precision","recall","f1","support"])
            for cname in class_names:
                row = per_class.get(cname, {"precision":0, "recall":0,
                                            "f1-score":0, "support":0})
                w.writerow([cname, row.get("precision",0), row.get("recall",0),
                            row.get("f1-score",0), row.get("support",0)])

        with open(base_run / "confusion_test_ensemble.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["true\\pred"] + class_names)
            for i, cname in enumerate(class_names):
                w.writerow([cname] + list(cm[i]))

        with open(base_run / "test_ensemble_predictions.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["path","patient_id","true","pred"] +
                       [f"p_{c}" for c in class_names])
            for i in range(len(test_ref_labels)):
                w.writerow([test_ref_paths[i], test_ref_pids[i],
                            class_names[int(test_ref_labels[i])],
                            class_names[int(preds_ens[i])]] +
                           list(map(float, probs_mean[i].tolist())))

    # === INTTEST ensemble (avg across folds) ===
    if inttest_loader_fixed is not None and len(inttest_probs_folds) > 0:
        int_probs_stack = np.stack(inttest_probs_folds, axis=0)  # [K, N_int, C]
        int_probs_mean  = int_probs_stack.mean(axis=0)           # [N_int, C]
        int_preds_ens   = int_probs_mean.argmax(axis=1)

        acc = float(accuracy_score(int_ref_labels, int_preds_ens))
        prec_m, rec_m, f1_m, _ = precision_recall_fscore_support(
            int_ref_labels, int_preds_ens, average='macro', zero_division=0)
        prec_w, rec_w, f1_w, _ = precision_recall_fscore_support(
            int_ref_labels, int_preds_ens, average='weighted', zero_division=0)
        per_class = classification_report(
            int_ref_labels, int_preds_ens, target_names=class_names,
            output_dict=True, zero_division=0)
        cm = confusion_matrix(int_ref_labels, int_preds_ens, labels=list(range(len(class_names))))

        int_ens_metrics = {
            "overall": {
                "accuracy": acc,
                "precision_macro": prec_m,
                "recall_macro": rec_m,
                "f1_macro": f1_m,
                "precision_weighted": prec_w,
                "recall_weighted": rec_w,
                "f1_weighted": f1_w,
                "num_samples": int(len(int_ref_labels)),
            },
            "per_class": per_class,
        }
        with open(base_run / "metrics_inttest_ensemble.json", 'w', encoding='utf-8') as f:
            json.dump(int_ens_metrics, f, indent=2)

        with open(base_run / "metrics_overall_inttest_ensemble.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["split","accuracy","precision_macro","recall_macro","f1_macro",
                        "precision_weighted","recall_weighted","f1_weighted","num_samples"])
            o = int_ens_metrics['overall']
            w.writerow(["inttest_ensemble", o['accuracy'], o['precision_macro'],
                        o['recall_macro'], o['f1_macro'], o['precision_weighted'],
                        o['recall_weighted'], o['f1_weighted'], o['num_samples']])

        with open(base_run / "per_class_inttest_ensemble.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["class","precision","recall","f1","support"])
            for cname in class_names:
                row = per_class.get(cname, {"precision":0, "recall":0,
                                            "f1-score":0, "support":0})
                w.writerow([cname, row.get("precision",0), row.get("recall",0),
                            row.get("f1-score",0), row.get("support",0)])

        with open(base_run / "confusion_inttest_ensemble.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["true\\pred"] + class_names)
            for i, cname in enumerate(class_names):
                w.writerow([cname] + list(cm[i]))

        with open(base_run / "inttest_ensemble_predictions.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["path","patient_id","true","pred"] +
                       [f"p_{c}" for c in class_names])
            for i in range(len(int_ref_labels)):
                w.writerow([int_ref_paths[i], int_ref_pids[i],
                            class_names[int(int_ref_labels[i])],
                            class_names[int(int_preds_ens[i])]] +
                           list(map(float, int_probs_mean[i].tolist())))

    # (Optional) train on FULL pool and evaluate test once
    if cfg.train_full_after_cv:
        print("\n[Full Train] Training with full pool (no internal test)...")
        run_full = base_run / "full_train"
        run_full.mkdir(parents=True, exist_ok=True)

        real_train_subset = list(base_pool)

        if cfg.balance_gan_to_max and len(synth_items) > 0:
            full_train_subset = build_balanced_train_subset(
                real_train_subset=real_train_subset,
                synth_items=synth_items,
                num_classes=len(class_names),
                seed=cfg.seed + 999,
            )
            print(f"[Full Train] real={len(real_train_subset)} | train_balanced(real+GAN)={len(full_train_subset)}")
        else:
            full_train_subset = real_train_subset + (synth_items if len(synth_items) > 0 else [])
            print(f"[Full Train] real={len(real_train_subset)} + synth={len(synth_items)} -> {len(full_train_subset)}")

        train_ds      = MRIDataset(full_train_subset, image_size=cfg.image_size, aug=False)
        real_train_ds = MRIDataset(real_train_subset, image_size=cfg.image_size, aug=False)

        train_loader      = DataLoader(train_ds,      batch_size=cfg.batch_size, shuffle=True,
                                       num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)
        real_train_loader = DataLoader(real_train_ds, batch_size=cfg.batch_size, shuffle=True,
                                       num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)

        model = build_model(num_classes=len(class_names), pretrained=cfg.pretrained).to(device)
        optimizer = torch.optim.AdamW(model.parameters(), lr=cfg.lr, weight_decay=cfg.weight_decay)
        scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=cfg.epochs)
        criterion = nn.CrossEntropyLoss()
        amp_enabled = (cfg.amp and device.type == 'cuda')
        scaler = torch.amp.GradScaler('cuda', enabled=amp_enabled)

        epochs_main = cfg.epochs - cfg.fine_tune_real_epochs
        epochs_ft   = cfg.fine_tune_real_epochs

        # Full-train Phase 1: full_train_subset (real + synth balanceado o no)
        for epoch in range(1, epochs_main + 1):
            model.train()
            for imgs, ys, _, _ in train_loader:
                imgs = imgs.to(device, non_blocking=True)
                ys = ys.to(device, non_blocking=True)
                optimizer.zero_grad(set_to_none=True)
                with torch.amp.autocast('cuda', enabled=amp_enabled):
                    loss = criterion(model(imgs), ys)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()
            scheduler.step()

        # Full-train Phase 2: sólo reales
        for ft_idx in range(1, epochs_ft + 1):
            epoch = epochs_main + ft_idx
            model.train()
            for imgs, ys, _, _ in real_train_loader:
                imgs = imgs.to(device, non_blocking=True)
                ys = ys.to(device, non_blocking=True)
                optimizer.zero_grad(set_to_none=True)
                with torch.amp.autocast('cuda', enabled=amp_enabled):
                    loss = criterion(model(imgs), ys)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()
            scheduler.step()

        torch.save({'model': model.state_dict(), 'class_names': class_names},
                   run_full / 'final_full.ckpt')

        if test_loader_fixed is not None:
            print("[Full Train] Evaluating on TEST ...")
            evaluate(model, test_loader_fixed, device, class_names,
                     split_name='test_full', out_dir=run_full)

    print(f"\n[Artifacts] Saved at: {base_run}")
    return base_run
```

```python
# === Batch projector with N per class + ETA + prototype computation ===
#
# It does:
#   1) For each class in DATA_ROOT/<class>/*.png
#        -> projects up to N_PER_CLASS images with projector.py
#        -> saves projected_w.npz under PROJ_ROOT/<class>/<image_stem>/
#   2) For each class, loads ALL projected_w.npz from PROJ_ROOT
#        -> computes mean W+ prototype
#        -> saves PROTO_ROOT/<class>_prototype_mean.npz
#
# Assumes:
#   - projector.py is in the current repo (StyleGAN2-ADA)
#   - projector.py saves np.savez(..., w=...)
#
# This uses GPU because projector.py uses CUDA.
aa
from pathlib import Path
from typing import List, Dict
import subprocess
import numpy as np
import time
import random

# --------------------
# CONFIG
# --------------------

# Folder with images per class (train or test, doesn't matter)
DATA_ROOT = Path("/content/localdata/test_unzipped/testreducido")

# Where all per-image projections will be stored
PROJ_ROOT = Path("/content/drive/MyDrive/TESIS/projections_batch_all")

# Where class prototypes (means) will be stored
PROTO_ROOT = Path("/content/drive/MyDrive/TESIS/prototypes")

# Your StyleGAN2-ADA network
NETWORK_PKL = Path(
    "/content/drive/MyDrive/TESIS/StyleGan2/Modelsinge/"
    "00000-finalnuevoinge-cond-mirror-24gb-gpu-kimg1000-batch32-ada-target0.6-bg/"
    "network-snapshot-000600.pkl"
)

# projector.py script (from sg2-ada repo)
PROJECTOR_SCRIPT = Path("projector.py")

# Classes (subfolders of DATA_ROOT). If None, we infer them automatically.
CLASSES: List[str] = ["glioma", "meningioma", "pituitary"]
# CLASSES = None  # alternative: infer automatically

# Max number of images to project per class:
#   - N_PER_CLASS = None  -> use ALL images per class
#   - N_PER_CLASS = 50    -> project at most 50 per class
N_PER_CLASS: int | None = 10

# Projector hyperparameters
NUM_STEPS = 300
BASE_SEED = 122

# Shuffle images before taking first N_PER_CLASS?
SHUFFLE = True

# Valid image extensions
IMG_EXTS = {".png", ".jpg", ".jpeg", ".bmp", ".tif", ".tiff"}


# --------------------
# Utility functions
# --------------------

def load_last_w(npz_path: Path) -> np.ndarray:
    """
    Load the last W+ code from an npz file.

    Supports:
      - w.shape == [T, num_ws, w_dim]  (projector steps)
      - w.shape == [num_ws, w_dim]     (already a single W+)
      - w.shape == [1, num_ws, w_dim]  (StyleGAN projector variant)
    """
    if not npz_path.is_file():
        raise FileNotFoundError(f"NPZ not found: {npz_path}")
    data = np.load(str(npz_path))
    if "w" not in data:
        raise RuntimeError(f"'w' key not found in {npz_path}")
    w = data["w"]
    if w.ndim == 3:
        # [T, num_ws, w_dim] or [1, num_ws, w_dim]
        return w[-1]  # last step or the only one
    elif w.ndim == 2:
        return w
    else:
        raise RuntimeError(f"Unexpected 'w' shape in {npz_path}: {w.shape}")


def run_projector(
    projector_script: Path,
    network_pkl: Path,
    image_path: Path,
    outdir: Path,
    num_steps: int,
    seed: int,
) -> Path:
    """
    Run projector.py on a single image.

    Equivalent to:
        python projector.py --network=... --target=... --outdir=... \
                            --num-steps=... --seed=...
    """
    outdir.mkdir(parents=True, exist_ok=True)

    cmd = [
        "python",
        str(projector_script),
        f"--network={network_pkl}",
        f"--target={image_path}",
        f"--outdir={outdir}",
        f"--num-steps={num_steps}",
        f"--seed={seed}",
    ]

    print(f"[PROJECTOR] {image_path.name} -> {outdir}")
    print("  CMD:", " ".join(cmd))

    result = subprocess.run(cmd, check=False, capture_output=True, text=True)
    print("[PROJECTOR] STDOUT:")
    print(result.stdout)
    print("[PROJECTOR] STDERR:")
    print(result.stderr)

    if result.returncode != 0:
        raise RuntimeError(
            f"projector.py failed with exit code {result.returncode} "
            f"for image {image_path}"
        )

    proj_npz = outdir / "projected_w.npz"
    if not proj_npz.is_file():
        raise FileNotFoundError(
            f"Expected projected_w.npz not found in {outdir}. "
            f"Check projector.py (it must save np.savez(..., w=...))."
        )

    return proj_npz


def run_projection_for_all_images(
    data_root: Path,
    proj_root: Path,
    network_pkl: Path,
    projector_script: Path,
    classes: List[str] | None = None,
    num_steps: int = 200,
    base_seed: int = 0,
    n_per_class: int | None = None,
    shuffle: bool = True,
) -> None:
    """
    Project up to n_per_class images per class and show a global ETA.

    Input structure:
      data_root/<class>/*.png

    Output structure:
      proj_root/<class>/<image_stem>/projected_w.npz
    """
    data_root = Path(data_root)
    proj_root = Path(proj_root)
    proj_root.mkdir(parents=True, exist_ok=True)

    # 1) Infer classes if not given
    if classes is None:
        classes = [d.name for d in data_root.iterdir() if d.is_dir()]
    print(f"[INFO] Using classes: {classes}")

    # 2) Collect images per class
    class_to_imgs: Dict[str, List[Path]] = {}
    total_tasks = 0

    for cls in classes:
        cls_dir = data_root / cls
        if not cls_dir.is_dir():
            print(f"[WARN] Class dir not found: {cls_dir}, skipping.")
            continue

        img_paths = sorted(
            p for p in cls_dir.rglob("*") if p.suffix.lower() in IMG_EXTS
        )
        if not img_paths:
            print(f"[WARN] No images in {cls_dir}, skipping.")
            continue

        if shuffle:
            random.shuffle(img_paths)

        if n_per_class is not None and len(img_paths) > n_per_class:
            img_paths = img_paths[:n_per_class]

        class_to_imgs[cls] = img_paths
        total_tasks += len(img_paths)
        print(f"[CLASS] {cls}: {len(img_paths)} images will be projected.")

    if total_tasks == 0:
        print("[INFO] No images to project. Exiting.")
        return

    print(f"\n[INFO] Total images to project: {total_tasks}")

    # 3) Main loop with ETA
    processed = 0
    global_idx = 0
    start_time = time.time()

    def _fmt_time(t: float) -> str:
        m = int(t // 60)
        s = int(t % 60)
        return f"{m:02d}:{s:02d}"

    for cls, img_paths in class_to_imgs.items():
        print(f"\n[RUN] Class {cls}: {len(img_paths)} images")
        for img_path in img_paths:
            stem = img_path.stem
            outdir = proj_root / cls / stem
            proj_npz = outdir / "projected_w.npz"

            processed += 1
            global_idx += 1

            if proj_npz.is_file():
                print(f"[SKIP] {proj_npz} already exists.")
            else:
                seed = base_seed + global_idx
                try:
                    run_projector(
                        projector_script=projector_script,
                        network_pkl=network_pkl,
                        image_path=img_path,
                        outdir=outdir,
                        num_steps=num_steps,
                        seed=seed,
                    )
                except Exception as e:
                    print(f"[ERROR] Failed on {img_path}: {e}")
                    # Continue with next image
                    continue

            # Update ETA
            elapsed = time.time() - start_time
            avg_per_img = elapsed / processed
            remaining_imgs = total_tasks - processed
            eta_sec = remaining_imgs * avg_per_img

            print(
                f"[ETA] {processed}/{total_tasks} "
                f"({processed/total_tasks*100:.1f}%) | "
                f"elapsed={_fmt_time(elapsed)} | remaining~{_fmt_time(eta_sec)}"
            )

    print("\n[DONE] All projections attempted.")


def compute_class_prototypes(
    proj_root: Path,
    proto_root: Path,
    classes: List[str] | None = None,
) -> None:
    """
    For each class, load ALL projected_w.npz from:
        proj_root/<class>/*/projected_w.npz
    compute the mean W+ and save to:
        proto_root/<class>_prototype_mean.npz  (key 'w')
    """
    proj_root = Path(proj_root)
    proto_root = Path(proto_root)
    proto_root.mkdir(parents=True, exist_ok=True)

    # If classes not provided, infer them from proj_root subfolders
    if classes is None:
        classes = [d.name for d in proj_root.iterdir() if d.is_dir()]

    print(f"[PROTO] Computing prototypes for classes: {classes}")

    for cls in classes:
        class_proj_dir = proj_root / cls
        if not class_proj_dir.is_dir():
            print(f"[PROTO] No projections for class '{cls}' at {class_proj_dir}, skipping.")
            continue

        npz_paths = sorted(class_proj_dir.rglob("projected_w.npz"))
        if not npz_paths:
            print(f"[PROTO] No projected_w.npz found under {class_proj_dir}, skipping.")
            continue

        ws = []
        for p in npz_paths:
            w = load_last_w(p)
            ws.append(w)
            print(f"[PROTO] Loaded {p}, w.shape={w.shape}")

        ws_stack = np.stack(ws, axis=0)  # [N, num_ws, w_dim]
        proto_w = ws_stack.mean(axis=0)  # [num_ws, w_dim]

        out_npz = proto_root / f"{cls}_prototype_mean.npz"
        np.savez(out_npz, w=proto_w)
        print(f"[PROTO] Saved {cls} prototype -> {out_npz}, w.shape={proto_w.shape}")

    print("\n[PROTO] All prototypes computed.")


# --------------------
# Main / Snippet usage
# --------------------

if __name__ == "__main__":
    # 1) Project up to N_PER_CLASS images per class with ETA
    run_projection_for_all_images(
        data_root=DATA_ROOT,
        proj_root=PROJ_ROOT,
        network_pkl=NETWORK_PKL,
        projector_script=PROJECTOR_SCRIPT,
        classes=CLASSES,
        num_steps=NUM_STEPS,
        base_seed=BASE_SEED,
        n_per_class=N_PER_CLASS,  # set to None to use ALL images
        shuffle=SHUFFLE,
    )

    # 2) Compute prototypes and save under PROTO_ROOT
    compute_class_prototypes(
        proj_root=PROJ_ROOT,
        proto_root=PROTO_ROOT,
        classes=CLASSES,
    )
```

```python
from pathlib import Path
from typing import Dict, List
from collections import defaultdict
import subprocess
import shutil
import numpy as np
import time

IMG_EXTS = {".png", ".jpg", ".jpeg", ".bmp", ".tif", ".tiff"}


def load_last_w(npz_path: Path) -> np.ndarray:
    if not npz_path.is_file():
        raise FileNotFoundError(f"NPZ not found: {npz_path}")
    data = np.load(str(npz_path))
    if "w" not in data:
        raise RuntimeError(f"'w' key not found in {npz_path}")
    w = data["w"]
    if w.ndim == 3:
        return w[-1]
    elif w.ndim == 2:
        return w
    else:
        raise RuntimeError(f"Unexpected 'w' shape in {npz_path}: {w.shape}")


def run_projector(
    projector_script: Path,
    network_pkl: Path,
    image_path: Path,
    outdir: Path,
    num_steps: int,
    seed: int,
    device: str = "cuda",
) -> Path:
    outdir.mkdir(parents=True, exist_ok=True)

    cmd = [
        "python",
        str(projector_script),
        f"--network={network_pkl}",
        f"--target={image_path}",
        f"--outdir={outdir}",
        f"--num-steps={num_steps}",
        f"--seed={seed}",
        f"--device={device}",
    ]

    print("[PROJECTOR] Running:")
    print(" ", " ".join(cmd))

    result = subprocess.run(cmd, check=False, capture_output=True, text=True)
    print("[PROJECTOR] STDOUT:")
    print(result.stdout)
    print("[PROJECTOR] STDERR:")
    print(result.stderr)

    if result.returncode != 0:
        raise RuntimeError(
            f"projector.py failed with exit code {result.returncode} "
            f"for image {image_path}"
        )

    proj_npz = outdir / "projected_w.npz"
    if not proj_npz.is_file():
        raise FileNotFoundError(
            f"Expected projected_w.npz not found in {outdir}. "
            f"Check projector.py (it must save np.savez(..., w=...))."
        )
    return proj_npz


def extract_patient_id(image_path: Path) -> str:
    stem = image_path.stem
    parts = stem.split("_")
    if len(parts) >= 1:
        return parts[0]
    return stem


def select_frames_by_percentile(frames_dir: Path, k: int) -> List[Path]:
    if not frames_dir.is_dir():
        raise FileNotFoundError(f"Frames directory not found: {frames_dir}")

    frames = sorted(frames_dir.glob("frame*.png"))
    if not frames:
        raise FileNotFoundError(f"No frame*.png found in {frames_dir}")

    n = len(frames)
    if k <= 0:
        return []
    if k >= n:
        return frames

    if k == 1:
        ps = [0.5]
    else:
        step = 1.0 / (k + 1)
        ps = [step * (i + 1) for i in range(k)]

    idxs: List[int] = []
    for p in ps:
        idx = int(round(p * (n - 1)))
        idxs.append(idx)

    idxs = sorted(set(idxs))
    return [frames[i] for i in idxs]


def _format_eta(seconds: float) -> str:
    seconds = int(seconds)
    h = seconds // 3600
    m = (seconds % 3600) // 60
    s = seconds % 60
    if h > 0:
        return f"{h:d}h {m:02d}m {s:02d}s"
    elif m > 0:
        return f"{m:d}m {s:02d}s"
    else:
        return f"{s:d}s"


def run_xai_counterfactual_from_folder(
    test_root: Path,
    network_pkl: Path,
    outdir_root: Path,
    projector_script: Path,
    generate_script: Path,
    proto_root: Path,
    classes: List[str],
    class_to_index: Dict[str, int],
    num_steps: int = 200,
    seed: int = 0,
    frames: int = 32,
    interp_mode: str = "linear",
    easing: str = "linear",
    truncation_psi: float = 1.0,
    noise_mode: str = "const",
    fps: int = 24,
    frames_to_keep_per_class: int = 1,
    max_images_per_patient: int = 0,
    device: str = "cuda",

) -> None:
    test_root = Path(test_root)
    network_pkl = Path(network_pkl)
    outdir_root = Path(outdir_root)
    projector_script = Path(projector_script)
    generate_script = Path(generate_script)
    proto_root = Path(proto_root)

    proj_root = outdir_root / "projection"
    pairs_root = outdir_root / "pairs"
    interps_root = outdir_root / "interpolations"
    classif_root = outdir_root / "classification"

    for d in [proj_root, pairs_root, interps_root, classif_root]:
        d.mkdir(parents=True, exist_ok=True)

    prototypes: Dict[str, np.ndarray] = {}
    for cls in classes:
        proto_path = proto_root / f"{cls}_prototype_mean.npz"
        if not proto_path.is_file():
            raise FileNotFoundError(
                f"Prototype for class '{cls}' not found at {proto_path}. "
                f"Run your prototype-building step first."
            )
        w_proto = load_last_w(proto_path)
        prototypes[cls] = w_proto
        print(f"[PROTO] Loaded {cls} prototype with shape {w_proto.shape} from {proto_path}")

    per_patient_counts: Dict[str, int] = defaultdict(int)
    cases: List[tuple] = []

    for true_cls in classes:
        cls_dir = test_root / true_cls
        if not cls_dir.is_dir():
            print(f"[WARN] Test class dir not found: {cls_dir}, skipping.")
            continue

        img_paths = sorted(
            p for p in cls_dir.rglob("*") if p.suffix.lower() in IMG_EXTS
        )
        if not img_paths:
            print(f"[WARN] No images found under {cls_dir}, skipping.")
            continue

        for img_path in img_paths:
            patient_id = extract_patient_id(img_path)
            current_count = per_patient_counts[patient_id]

            if max_images_per_patient > 0 and current_count >= max_images_per_patient:
                continue

            per_patient_counts[patient_id] += 1
            img_idx = per_patient_counts[patient_id]
            cases.append((true_cls, img_path, patient_id, img_idx))

    total_cases = len(cases)
    if total_cases == 0:
        print("[INFO] No cases to process. Exiting.")
        return

    print(f"[INFO] Total cases to process (patient-image pairs): {total_cases}")

    start_time = time.time()

    for case_i, (true_cls, img_path, patient_id, img_idx) in enumerate(cases, start=1):
        elapsed = time.time() - start_time
        if case_i == 1:
            elapsed_str = _format_eta(elapsed)
            eta_str = "estimating..."
        else:
            avg_per_case = elapsed / (case_i - 1)
            remaining = total_cases - (case_i - 1)
            eta_sec = avg_per_case * remaining
            elapsed_str = _format_eta(elapsed)
            eta_str = _format_eta(eta_sec)

        print(
            f"\n[PROGRESS] Case {case_i}/{total_cases} | "
            f"elapsed={elapsed_str} | ETA={eta_str}"
        )

        print(f"[CASE] Patient {patient_id} (img #{img_idx} for this patient, true class: {true_cls})")
        print(f"       Using image: {img_path}")

        proj_dir = proj_root / true_cls / f"{patient_id}_i{img_idx:02d}"
        proj_npz = run_projector(
            projector_script=projector_script,
            network_pkl=network_pkl,
            image_path=img_path,
            outdir=proj_dir,
            num_steps=num_steps,
            seed=seed,
        )
        w_real = load_last_w(proj_npz)
        print(f"[INFO] Loaded w_real from {proj_npz} with shape {w_real.shape}")

        case_classif_dir = classif_root / patient_id
        case_classif_dir.mkdir(parents=True, exist_ok=True)
        real_dest = case_classif_dir / f"real_{true_cls}_i{img_idx:02d}.png"
        shutil.copy2(img_path, real_dest)
        print(f"[CLASSIF] Saved real image -> {real_dest}")

        for target_cls in classes:
            w_proto = prototypes[target_cls]
            if w_real.shape != w_proto.shape:
                raise RuntimeError(
                    f"Shape mismatch for patient {patient_id}, target '{target_cls}': "
                    f"w_real {w_real.shape} vs w_proto {w_proto.shape}"
                )

            w_pair = np.stack([w_real, w_proto], axis=0)

            pair_dir = pairs_root / patient_id
            pair_dir.mkdir(parents=True, exist_ok=True)
            pair_npz = pair_dir / f"{patient_id}_i{img_idx:02d}_to_{target_cls}.npz"
            np.savez(pair_npz, w=w_pair)
            print(f"[PAIR] Saved W pair -> {pair_npz} (shape={w_pair.shape})")

            outdir_cls = interps_root / patient_id / f"i{img_idx:02d}" / target_cls
            outdir_cls.mkdir(parents=True, exist_ok=True)

            class_idx = class_to_index.get(target_cls, 0)

            cmd = [
                "python", str(generate_script),
                "--network", str(network_pkl),
                "--space=w",
                "--process=interpolation",
                "--projected-w", str(pair_npz),
                "--interpolation", interp_mode,
                "--easing", easing,
                "--frames", str(frames),
                "--trunc", str(truncation_psi),
                "--noise-mode", noise_mode,
                "--fps", str(fps),
                "--class", str(class_idx),
                "--outdir", str(outdir_cls),
            ]

            print(f"[GEN] Running interpolation to '{target_cls}' for patient {patient_id}, image #{img_idx}:")
            print(" ", " ".join(cmd))

            result = subprocess.run(cmd, check=False, capture_output=True, text=True)
            print("[GEN] STDOUT:")
            print(result.stdout)
            print("[GEN] STDERR:")
            print(result.stderr)

            if result.returncode != 0:
                raise RuntimeError(
                    f"generate.py failed for patient {patient_id}, img #{img_idx}, "
                    f"target '{target_cls}' with exit code {result.returncode}"
                )

            frames_dir = outdir_cls / "frames"
            frames_to_copy = select_frames_by_percentile(frames_dir, frames_to_keep_per_class)

            for j, frame_path in enumerate(frames_to_copy, start=1):
                cf_dest = case_classif_dir / f"cf_to_{target_cls}_i{img_idx:02d}_f{j:02d}.png"
                shutil.copy2(frame_path, cf_dest)
                print(f"[CLASSIF] Saved frame {frame_path.name} -> {cf_dest}")

    total_elapsed = time.time() - start_time
    print(f"\n[DONE] XAI counterfactual batch finished in {_format_eta(total_elapsed)}.")
    print(f"Classification images per patient are under: {classif_root}")
```

```python
# ensemble_cnn_lpips_patient_gated_norm.py
# -*- coding: utf-8 -*-

from pathlib import Path
from typing import List, Dict, Optional
import csv
import re

import numpy as np
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import transforms, models

from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix

import lpips  # pip install lpips


def lpips_distances_to_probs(
    D: np.ndarray,
    temp: float = 0.2,
) -> np.ndarray:
    """
    Convierte distancias LPIPS D (más chico = mejor) en probabilidades.

    Pasos:
      1) Trabaja solo con valores finitos.
      2) Normaliza D a [0,1] usando (D - min) / (max - min).
      3) Aplica softmax(-D_norm / temp) para que la clase con
         menor distancia tenga prob más alta.
    """
    D = np.asarray(D, dtype=np.float32)
    C = D.shape[0]

    # Mascara de valores finitos
    finite_mask = np.isfinite(D)
    if not finite_mask.any():
        # Si TODO es inf, devolvemos uniforme
        return np.ones(C, dtype=np.float32) / float(C)

    # Extrae rango de los valores finitos
    D_finite = D[finite_mask]
    d_min = float(D_finite.min())
    d_max = float(D_finite.max())

    # Normalizamos el rango a [0,1]
    D_norm = np.zeros_like(D, dtype=np.float32)
    if d_max > d_min:
        D_norm = (D - d_min) / (d_max - d_min)
    else:
        # Si todas las distancias son iguales, no hay info:
        # usamos distribución uniforme
        return np.ones(C, dtype=np.float32) / float(C)

    # Los infinitos los mandamos al peor caso (1.0)
    D_norm[~finite_mask] = 1.0

    # Softmax con temperatura (más chico = distribución más "picuda")
    logits = -D_norm / max(temp, 1e-6)
    logits = logits - logits.max()  # por estabilidad numérica
    e = np.exp(logits)
    p = e / e.sum()

    return p.astype(np.float32)


def ensemble_cnn_lpips_patient_gated(
    classification_root: Path,
    model: nn.Module,
    class_names: List[str],
    out_csv: Path,
    device: Optional[str] = None,
    image_size: int = 256,
    frame_weights: Optional[Dict[int, float]] = None,
    class_weights: Optional[Dict[str, float]] = None,
    alpha_cnn: float = 0.5,
    tau_cnn: float = 0.80,
    tau_lpips: float = 0.70,
    lpips_temp: float = 0.2,   # <<< temperatura para LPIPS normalizado
):
    """
    Patient-level ensemble de:
      (1) CNN (probabilidades agregadas sobre imágenes generadas)
      (2) LPIPS (inference by generation, real vs generadas)

    GATED ENSEMBLE:

      - Si max(p_cnn) >= tau_cnn  -> usamos solo CNN:
            p_ens = p_cnn
      - Else, si max(p_lpips) >= tau_lpips -> usamos solo LPIPS:
            p_ens = p_lpips
      - Else -> mezcla suave:
            p_ens = alpha_cnn * p_cnn + (1 - alpha_cnn) * p_lpips

    LPIPS -> prob se calcula con lpips_distances_to_probs(D, temp=lpips_temp).
    """

    classification_root = Path(classification_root)
    out_csv = Path(out_csv)

    # ---- Device ----
    if device is None:
        device = "cuda" if torch.cuda.is_available() else "cpu"
    print(f"[INFO] Using device (ensemble gated): {device}")

    # ---- CNN model ----
    model = model.to(device)
    model.eval()

    # ---- LPIPS network ----
    lpips_fn = lpips.LPIPS(net='alex').to(device)
    lpips_fn.eval()

    num_classes = len(class_names)
    class_to_idx = {c: i for i, c in enumerate(class_names)}
    idx_to_class = {i: c for c, i in class_to_idx.items()}

    # ---- Default weights ----
    if frame_weights is None:
        frame_weights = {1: 1/3, 2: 1/3, 3: 1/3}
    print(f"[INFO] frame_weights (ensemble) = {frame_weights}")

    if class_weights is None:
        class_weights = {c: 1.0 for c in class_names}
    print(f"[INFO] class_weights (ensemble) = {class_weights}")

    # ---- Transforms ----
    resize_tf = transforms.Resize((image_size, image_size))
    to_tensor = transforms.ToTensor()
    norm_cnn = transforms.Normalize(mean=[0.5, 0.5, 0.5],
                                    std=[0.5, 0.5, 0.5])

    def preprocess_cnn(pil_img: Image.Image) -> torch.Tensor:
        img = pil_img.convert("L")
        img = resize_tf(img)
        t = to_tensor(img)         # [1,H,W]
        t = t.repeat(3, 1, 1)      # [3,H,W]
        t = norm_cnn(t)
        return t

    def preprocess_lpips(pil_img: Image.Image) -> torch.Tensor:
        img = pil_img.convert("L")
        img = resize_tf(img)
        t = to_tensor(img)         # [1,H,W] in [0,1]
        t = t.repeat(3, 1, 1)      # [3,H,W]
        t = t * 2.0 - 1.0          # [-1,1]
        return t.unsqueeze(0)      # [1,3,H,W]

    # ---- Patient folders ----
    patient_dirs = [d for d in classification_root.iterdir() if d.is_dir()]
    patient_dirs = sorted(patient_dirs, key=lambda p: p.name)

    print(f"[INFO] Found {len(patient_dirs)} patient folders in {classification_root}")

    real_pattern = re.compile(r"^real_([^_]+)(?:_i(\d+))?$")
    cf_pattern = re.compile(r"^cf_to_([a-zA-Z0-9]+)(?:_i(\d+))?_f(\d+)$")

    all_rows = []

    y_true_all = []
    y_pred_cnn_all = []
    y_pred_lpips_all = []
    y_pred_ens_all = []

    for patient_dir in patient_dirs:
        patient_id = patient_dir.name

        # ===== Real images (for LPIPS) =====
        real_files = list(patient_dir.glob("real_*.png"))
        if not real_files:
            print(f"[WARN] No real_*.png for patient {patient_id}, skipping.")
            continue

        real_by_idx: Dict[int, Path] = {}
        true_cls = None

        for rf in real_files:
            m_real = real_pattern.match(rf.stem)
            if not m_real:
                continue
            cls = m_real.group(1)
            idx_str = m_real.group(2)

            if true_cls is None:
                true_cls = cls
            elif cls != true_cls:
                print(f"[WARN] Multiple classes in real_*.png for patient {patient_id}, skipping.")
                true_cls = None
                break

            idx_img = int(idx_str) if idx_str is not None else 0
            real_by_idx[idx_img] = rf

        if true_cls is None or true_cls not in class_to_idx:
            print(f"[WARN] Could not determine valid true class for patient {patient_id}, skipping.")
            continue

        true_idx = class_to_idx[true_cls]

        real_tensor_cache: Dict[int, torch.Tensor] = {}

        # ===== Generated images =====
        cf_files = sorted(patient_dir.glob("cf_to_*.png"))
        if not cf_files:
            print(f"[WARN] No cf_to_*.png for patient {patient_id}, skipping.")
            continue

        print(f"\n[PATIENT] {patient_id} | y_true = {true_cls} | #generated images = {len(cf_files)}")

        # Aggregadores CNN
        sum_probs_cnn = torch.zeros(num_classes, device=device)
        sum_w_cnn = 0.0

        # Aggregadores LPIPS
        sum_d_per_target = {c_idx: 0.0 for c_idx in range(num_classes)}
        sum_w_per_target = {c_idx: 0.0 for c_idx in range(num_classes)}
        count_imgs_per_target = {c_idx: 0 for c_idx in range(num_classes)}

        for img_path in cf_files:
            m_cf = cf_pattern.match(img_path.stem)
            if not m_cf:
                print(f"[WARN] Filename {img_path.name} does not match cf pattern, skipping.")
                continue

            target_cls = m_cf.group(1)
            idx_str = m_cf.group(2)
            frame_str = m_cf.group(3)

            if target_cls not in class_to_idx:
                print(f"[WARN] target_cls '{target_cls}' not in class_names, skipping {img_path.name}.")
                continue

            target_idx = class_to_idx[target_cls]
            img_idx = int(idx_str) if idx_str is not None else 0

            # Frame index
            try:
                frame_idx = int(frame_str)
            except Exception:
                frame_idx = None

            # Weights
            w_frame = frame_weights.get(frame_idx, 1.0)
            w_cls = class_weights.get(target_cls, 1.0)
            w_total = w_cls * w_frame

            # ---------- CNN ----------
            pil_cf = Image.open(img_path)
            x_cnn = preprocess_cnn(pil_cf).unsqueeze(0).to(device)
            with torch.no_grad():
                logits = model(x_cnn)
                probs = F.softmax(logits, dim=1)[0]  # [num_classes]
            sum_probs_cnn += w_total * probs
            sum_w_cnn += w_total

            # ---------- LPIPS ----------
            real_path = real_by_idx.get(img_idx, real_files[0])

            if img_idx not in real_tensor_cache:
                real_pil = Image.open(real_path)
                real_t = preprocess_lpips(real_pil).to(device)
                real_tensor_cache[img_idx] = real_t
            else:
                real_t = real_tensor_cache[img_idx]

            cf_t = preprocess_lpips(pil_cf).to(device)

            with torch.no_grad():
                d_lpips = lpips_fn(real_t, cf_t).item()

            sum_d_per_target[target_idx] += w_total * d_lpips
            sum_w_per_target[target_idx] += w_total
            count_imgs_per_target[target_idx] += 1

        # ===== CNN: prob paciente =====
        if sum_w_cnn == 0.0:
            print(f"[WARN] sum_w_cnn == 0 for patient {patient_id}, skipping.")
            continue

        p_cnn = (sum_probs_cnn / sum_w_cnn).detach().cpu().numpy()
        pred_idx_cnn = int(p_cnn.argmax())
        pred_cls_cnn = idx_to_class[pred_idx_cnn]
        max_p_cnn = float(p_cnn.max())

        # ===== LPIPS: distancias promedio -> prob normalizada =====
        D = np.zeros(num_classes, dtype=np.float32)
        for c_idx in range(num_classes):
            if sum_w_per_target[c_idx] > 0:
                D[c_idx] = float(sum_d_per_target[c_idx] / sum_w_per_target[c_idx])
            else:
                D[c_idx] = np.inf

        # D -> probs usando normalización min-max + temperatura
        p_lpips = lpips_distances_to_probs(D, temp=lpips_temp)
        pred_idx_lpips = int(p_lpips.argmax())
        pred_cls_lpips = idx_to_class[pred_idx_lpips]
        max_p_lpips = float(p_lpips.max())

        # ===== Ensemble (gated) =====
        if max_p_cnn >= tau_cnn:
            p_ens = p_cnn.copy()
            gate_mode = "cnn_only"
        elif max_p_lpips >= tau_lpips:
            p_ens = p_lpips.copy()
            gate_mode = "lpips_only"
        else:
            p_ens = alpha_cnn * p_cnn + (1.0 - alpha_cnn) * p_lpips
            gate_mode = "mixed"

        pred_idx_ens = int(p_ens.argmax())
        pred_cls_ens = idx_to_class[pred_idx_ens]

        print(f"  -> p_cnn       = {p_cnn}  (max={max_p_cnn:.3f})")
        print(f"  -> D (LPIPS)   = {D}")
        print(f"  -> p_lpips     = {p_lpips} (max={max_p_lpips:.3f})")
        print(f"  -> gate_mode   = {gate_mode}")
        print(f"  -> p_ensemble  = {p_ens}")
        print(f"  -> y_pred_cnn      = {pred_cls_cnn}")
        print(f"  -> y_pred_lpips    = {pred_cls_lpips}")
        print(f"  -> y_pred_ensemble = {pred_cls_ens}")

        y_true_all.append(true_idx)
        y_pred_cnn_all.append(pred_idx_cnn)
        y_pred_lpips_all.append(pred_idx_lpips)
        y_pred_ens_all.append(pred_idx_ens)

        row = {
            "patient_id": patient_id,
            "y_true": true_cls,
            "y_pred_cnn": pred_cls_cnn,
            "y_pred_lpips": pred_cls_lpips,
            "y_pred_ensemble": pred_cls_ens,
            "gate_mode": gate_mode,
            "max_p_cnn": max_p_cnn,
            "max_p_lpips": max_p_lpips,
        }
        for i, c in enumerate(class_names):
            row[f"p_cnn_{c}"] = float(p_cnn[i])
            row[f"D_{c}"] = float(D[i]) if np.isfinite(D[i]) else float("inf")
            row[f"p_lpips_{c}"] = float(p_lpips[i])
            row[f"p_ens_{c}"] = float(p_ens[i])
            row[f"n_imgs_{c}"] = int(count_imgs_per_target[i])
            row[f"sum_w_{c}"] = float(sum_w_per_target[i])

        all_rows.append(row)

    # ===== Save CSV =====
    out_csv.parent.mkdir(parents=True, exist_ok=True)
    fieldnames = (
        ["patient_id", "y_true", "y_pred_cnn", "y_pred_lpips", "y_pred_ensemble",
         "gate_mode", "max_p_cnn", "max_p_lpips"]
        + [f"p_cnn_{c}" for c in class_names]
        + [f"D_{c}" for c in class_names]
        + [f"p_lpips_{c}" for c in class_names]
        + [f"p_ens_{c}" for c in class_names]
        + [f"n_imgs_{c}" for c in class_names]
        + [f"sum_w_{c}" for c in class_names]
    )

    with out_csv.open("w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        for r in all_rows:
            writer.writerow(r)

    print(f"\n[INFO] Saved patient-level GATED ENSEMBLE results to: {out_csv}")

    # ===== Global metrics =====
    if len(y_true_all) == 0:
        print("[WARN] No patients evaluated. Check your classification_root.")
        return None

    results = {}

    # CNN
    acc_cnn = accuracy_score(y_true_all, y_pred_cnn_all)
    f1_cnn = f1_score(y_true_all, y_pred_cnn_all, average="macro")
    print("\n[METRICS] CNN (prob-based aggregation):")
    print(f"  Accuracy: {acc_cnn:.4f}")
    print(f"  F1-macro: {f1_cnn:.4f}")
    print("  Classification report:")
    print(classification_report(y_true_all, y_pred_cnn_all, target_names=class_names, digits=4))
    print("  Confusion matrix:")
    print(confusion_matrix(y_true_all, y_pred_cnn_all))
    results["cnn"] = {"accuracy": float(acc_cnn), "f1_macro": float(f1_cnn)}

    # LPIPS
    acc_lpips = accuracy_score(y_true_all, y_pred_lpips_all)
    f1_lpips = f1_score(y_true_all, y_pred_lpips_all, average="macro")
    print("\n[METRICS] LPIPS (inference by generation, normalized):")
    print(f"  Accuracy: {acc_lpips:.4f}")
    print(f"  F1-macro: {f1_lpips:.4f}")
    print("  Classification report:")
    print(classification_report(y_true_all, y_pred_lpips_all, target_names=class_names, digits=4))
    print("  Confusion matrix:")
    print(confusion_matrix(y_true_all, y_pred_lpips_all))
    results["lpips"] = {"accuracy": float(acc_lpips), "f1_macro": float(f1_lpips)}

    # Ensemble (gated)
    acc_ens = accuracy_score(y_true_all, y_pred_ens_all)
    f1_ens = f1_score(y_true_all, y_pred_ens_all, average="macro")
    print("\n[METRICS] GATED ENSEMBLE (alpha_cnn = %.2f, tau_cnn = %.2f, tau_lpips = %.2f, lpips_temp = %.2f):"
          % (alpha_cnn, tau_cnn, tau_lpips, lpips_temp))
    print(f"  Accuracy: {acc_ens:.4f}")
    print(f"  F1-macro: {f1_ens:.4f}")
    print("  Classification report:")
    print(classification_report(y_true_all, y_pred_ens_all, target_names=class_names, digits=4))
    print("  Confusion matrix:")
    print(confusion_matrix(y_true_all, y_pred_ens_all))
    results["ensemble_gated"] = {"accuracy": float(acc_ens), "f1_macro": float(f1_ens)}

    results["n_patients"] = len(y_true_all)
    return results
```

```python
# fusion_cnn_lpips_all.py
# -*- coding: utf-8 -*-
"""
Single script that:
- Loads a ResNet50 CNN checkpoint.
- Scans classification_root for patients with real_*.png and cf_to_*.png.
- Computes patient-level CNN and LPIPS probabilities.
- Builds a gated ensemble (CNN+LPIPS).
- Applies several decision-level fusion strategies:
    * Majority vote
    * Weighted majority
    * Bayes-style log-linear fusion
    * KL-based log-linear fusion
    * Joint posterior fusion
    * Dempster–Shafer (singleton case)

For each method, it reports:
- Global ACC and macro-F1
- Per-class precision/recall/F1
- Confusion matrix

All hyperparameters are grouped in a CONFIG section for easy editing.
"""

from pathlib import Path
from typing import List, Dict, Optional
import csv
import re

import numpy as np
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import models, transforms

from sklearn.metrics import (
    accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix,
)

import lpips  # pip install lpips


# ============================================================
# CONFIGURACIÓN GENERAL
# ============================================================

# Rutas que tú definiste
CLASSIFICATION_ROOT = Path("/content/localdata/class/classification")
CKPT_PATH = "/content/drive/MyDrive/TESIS/StyleGan2/CNNResnet50kfoldinge1/inge1.3fold/full_train/final_full.ckpt"
OUT_CSV = Path("/content/drive/MyDrive/TESIS/xai_batch_testset/results_fusion_all.csv")

# Clases (orden consistente en todo el script)
CLASS_NAMES = ["glioma", "meningioma", "pituitary"]


# ============================================================
# HIPERPARÁMETROS (EDITABLES FÁCILMENTE)
# ============================================================

# Pesos internos de agregación (mismo criterio que ya usabas)
FRAME_WEIGHTS = {
    1: 2.0,
    2: 0.3,
    3: 1.0,
}

CLASS_WEIGHTS = {
    "glioma": 400.0,
    "meningioma": 0.3,
    "pituitary": 1.0,
}

# Conversión de LPIPS → probabilidades (temperatura)
LPIPS_TEMP = 0.2

# Ensamble "gated" CNN+LPIPS
ALPHA_CNN_GATED = 0.5   # mezcla cuando ambos están por debajo de sus umbrales
TAU_CNN = 0.80          # umbral de confianza del CNN
TAU_LPIPS = 0.70        # umbral de confianza del LPIPS

# FUSIÓN A NIVEL DECISIÓN (CNN vs LPIPS)

# Weighted majority: pesos fijos para CNN y LPIPS (en desacuerdos)
WEIGHT_CNN_MAJ = 0.7
WEIGHT_LPIPS_MAJ = 0.3

# Bayes-style fusion (log-linear, α=0.5 es simétrico)
ALPHA_BAYES = 0.5

# KL-based fusion (otro α, típicamente más sesgado al CNN)
ALPHA_KL = 0.7

# Joint posterior (otro α, similar a KL pero conceptualmente distinto)
ALPHA_JOINT = 0.7


# ============================================================
# UTILIDADES DE MODELO / PREPROCESAMIENTO
# ============================================================

def build_model(num_classes: int, pretrained: bool = False) -> nn.Module:
    """
    Construye un ResNet50 y reemplaza la última capa fully connected.
    """
    m = models.resnet50(
        weights=models.ResNet50_Weights.IMAGENET1K_V1 if pretrained else None
    )
    in_features = m.fc.in_features
    m.fc = nn.Linear(in_features, num_classes)
    return m


def lpips_distances_to_probs(
    D: np.ndarray,
    temp: float = 0.2,
) -> np.ndarray:
    """
    Convierte distancias LPIPS por clase D[c] a probabilidades.
    Pasos:
      1) Min-max normalización en [0,1] (invirtiendo sentido: menor dist -> mejor).
      2) Softmax(-D_norm/temp).
    """
    D = np.asarray(D, dtype=np.float32)
    C = D.shape[0]

    finite_mask = np.isfinite(D)
    if not finite_mask.any():
        return np.ones(C, dtype=np.float32) / float(C)

    D_finite = D[finite_mask]
    d_min = float(D_finite.min())
    d_max = float(D_finite.max())

    D_norm = np.zeros_like(D, dtype=np.float32)
    if d_max > d_min:
        D_norm = (D - d_min) / (d_max - d_min)
    else:
        return np.ones(C, dtype=np.float32) / float(C)

    D_norm[~finite_mask] = 1.0

    logits = -D_norm / max(temp, 1e-6)
    logits = logits - logits.max()
    e = np.exp(logits)
    p = e / e.sum()

    return p.astype(np.float32)


# ============================================================
# FUSIÓN A NIVEL DECISIÓN
# ============================================================

def majority_vote(labels_list: List[np.ndarray]) -> np.ndarray:
    """
    Majority vote no ponderado.
    labels_list: lista de arrays [N] (uno por modelo).
    """
    labels_arr = np.stack(labels_list, axis=1)  # [N, M]
    N = labels_arr.shape[0]
    C = labels_arr.max() + 1
    y_pred = np.zeros(N, dtype=int)

    for i in range(N):
        votes = np.bincount(labels_arr[i], minlength=C)
        y_pred[i] = votes.argmax()
    return y_pred


def weighted_majority_vote(labels_list: List[np.ndarray], weights: List[float]) -> np.ndarray:
    """
    Majority vote ponderado entre modelos.
    """
    labels_arr = np.stack(labels_list, axis=1)  # [N, M]
    weights_arr = np.array(weights, dtype=float)

    N = labels_arr.shape[0]
    C = labels_arr.max() + 1
    y_pred = np.zeros(N, dtype=int)

    for i in range(N):
        scores = np.zeros(C, dtype=float)
        for label, w in zip(labels_arr[i], weights_arr):
            scores[int(label)] += w
        y_pred[i] = scores.argmax()

    return y_pred


def fuse_loglinear(
    probs_cnn: np.ndarray,
    probs_lpips: np.ndarray,
    alpha_cnn: float,
    eps: float = 1e-8,
) -> np.ndarray:
    """
    Fusión log-lineal / product-of-experts:
        p_fused ∝ (p_cnn)^α * (p_lpips)^(1-α)
    Esto representa:
      - Bayes-style fusion (α moderado)
      - KL-based fusion (mínimo de suma de KL)
      - Joint posterior (interpretación generativa)
    """
    assert 0.0 <= alpha_cnn <= 1.0
    pc = np.clip(probs_cnn, eps, 1.0)
    pl = np.clip(probs_lpips, eps, 1.0)

    log_pc = np.log(pc)
    log_pl = np.log(pl)

    log_pf = alpha_cnn * log_pc + (1.0 - alpha_cnn) * log_pl

    max_log = np.max(log_pf, axis=1, keepdims=True)
    exp_log = np.exp(log_pf - max_log)
    fused = exp_log / np.sum(exp_log, axis=1, keepdims=True)
    return fused


def fuse_dempster_shafer_singleton(
    probs_cnn: np.ndarray,
    probs_lpips: np.ndarray,
    eps: float = 1e-8,
) -> np.ndarray:
    """
    Fusión Dempster–Shafer en el caso sencillo donde cada fuente solo
    asigna masa a hipótesis singleton (sin masa explícita de ignorancia).
    Se reduce a un producto elemento a elemento y normalización.
    """
    pc = np.clip(probs_cnn, eps, 1.0)
    pl = np.clip(probs_lpips, eps, 1.0)

    m12 = pc * pl
    norm = np.sum(m12, axis=1, keepdims=True)
    fused = m12 / norm
    return fused


def print_metrics(name: str, y_true: np.ndarray, y_pred: np.ndarray) -> Dict[str, float]:
    """
    Imprime ACC y macro-F1.
    """
    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average="macro")
    print(f"{name:25s} ACC = {acc:.4f} | F1_macro = {f1:.4f}")
    return {"accuracy": float(acc), "f1_macro": float(f1)}


def print_class_metrics(name: str, y_true: np.ndarray, y_pred: np.ndarray, class_names: List[str]):
    """
    Imprime métricas por clase (precision/recall/F1) y matriz de confusión.
    """
    print(f"\n----- {name}: per-class metrics -----")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))
    cm = confusion_matrix(y_true, y_pred)
    print(f"Confusion matrix ({name}):")
    print(cm)
    print("--------------------------------------\n")


# ============================================================
# PIPELINE PRINCIPAL: CNN + LPIPS + FUSIONES
# ============================================================

def run_ensemble_and_fusions(
    classification_root: Path,
    ckpt_path: str,
    out_csv: Path,
    class_names: List[str],
    frame_weights: Optional[Dict[int, float]] = None,
    class_weights: Optional[Dict[str, float]] = None,
    alpha_cnn_gated: float = ALPHA_CNN_GATED,
    tau_cnn: float = TAU_CNN,
    tau_lpips: float = TAU_LPIPS,
    lpips_temp: float = LPIPS_TEMP,
):
    """
    - Carga CNN y LPIPS.
    - Para cada paciente:
        * Obtiene probabilidades CNN y LPIPS por clase.
        * Calcula ensamble gated.
    - Luego aplica fusiones:
        * Majority vote
        * Weighted majority
        * Bayes
        * KL-based
        * Joint posterior
        * Dempster–Shafer
    - Guarda CSV por paciente y muestra métricas globales y por clase.
    """

    classification_root = Path(classification_root)
    out_csv = Path(out_csv)

    # ---- Device ----
    device = "cuda" if torch.cuda.is_available() else "cpu"
    print(f"[INFO] Using device: {device}")

    # ---- CNN ----
    num_classes = len(class_names)
    print("[INFO] Loading CNN checkpoint...")
    model = build_model(num_classes=num_classes, pretrained=False)
    checkpoint = torch.load(ckpt_path, map_location=device)
    if "model" in checkpoint:
        model.load_state_dict(checkpoint["model"])
    else:
        model.load_state_dict(checkpoint)
    model = model.to(device)
    model.eval()
    print("[INFO] CNN loaded.")

    # ---- LPIPS ----
    print("[INFO] Loading LPIPS (alex)...")
    lpips_fn = lpips.LPIPS(net='alex').to(device)
    lpips_fn.eval()
    print("[INFO] LPIPS ready.")

    class_to_idx = {c: i for i, c in enumerate(class_names)}
    idx_to_class = {i: c for c, i in class_to_idx.items()}

    # ---- Pesos de agregación ----
    if frame_weights is None:
        frame_weights = FRAME_WEIGHTS
    if class_weights is None:
        class_weights = CLASS_WEIGHTS

    print(f"[INFO] frame_weights = {frame_weights}")
    print(f"[INFO] class_weights = {class_weights}")

    # ---- Transforms ----
    resize_tf = transforms.Resize((256, 256))
    to_tensor = transforms.ToTensor()
    norm_cnn = transforms.Normalize(mean=[0.5, 0.5, 0.5],
                                    std=[0.5, 0.5, 0.5])

    def preprocess_cnn(pil_img: Image.Image) -> torch.Tensor:
        img = pil_img.convert("L")
        img = resize_tf(img)
        t = to_tensor(img)         # [1,H,W]
        t = t.repeat(3, 1, 1)      # [3,H,W]
        t = norm_cnn(t)
        return t

    def preprocess_lpips(pil_img: Image.Image) -> torch.Tensor:
        img = pil_img.convert("L")
        img = resize_tf(img)
        t = to_tensor(img)         # [1,H,W] in [0,1]
        t = t.repeat(3, 1, 1)      # [3,H,W]
        t = t * 2.0 - 1.0          # [-1,1]
        return t.unsqueeze(0)      # [1,3,H,W]

    # ---- Pacientes ----
    patient_dirs = [d for d in classification_root.iterdir() if d.is_dir()]
    patient_dirs = sorted(patient_dirs, key=lambda p: p.name)

    print(f"[INFO] Found {len(patient_dirs)} patient folders in {classification_root}")

    real_pattern = re.compile(r"^real_([^_]+)(?:_i(\d+))?$")
    cf_pattern = re.compile(r"^cf_to_([a-zA-Z0-9]+)(?:_i(\d+))?_f(\d+)$")

    all_rows = []

    y_true_all = []
    y_pred_cnn_all = []
    y_pred_lpips_all = []
    y_pred_ens_all = []

    p_cnn_all = []
    p_lpips_all = []

    for patient_dir in patient_dirs:
        patient_id = patient_dir.name

        # ===== Real images =====
        real_files = list(patient_dir.glob("real_*.png"))
        if not real_files:
            print(f"[WARN] No real_*.png for patient {patient_id}, skipping.")
            continue

        real_by_idx: Dict[int, Path] = {}
        true_cls = None

        for rf in real_files:
            m_real = real_pattern.match(rf.stem)
            if not m_real:
                continue
            cls = m_real.group(1)
            idx_str = m_real.group(2)

            if true_cls is None:
                true_cls = cls
            elif cls != true_cls:
                print(f"[WARN] Multiple real classes for patient {patient_id}, skipping.")
                true_cls = None
                break

            idx_img = int(idx_str) if idx_str is not None else 0
            real_by_idx[idx_img] = rf

        if true_cls is None or true_cls not in class_to_idx:
            print(f"[WARN] Could not determine valid true class for patient {patient_id}, skipping.")
            continue

        true_idx = class_to_idx[true_cls]
        real_tensor_cache: Dict[int, torch.Tensor] = {}

        # ===== Generated images =====
        cf_files = sorted(patient_dir.glob("cf_to_*.png"))
        if not cf_files:
            print(f"[WARN] No cf_to_*.png for patient {patient_id}, skipping.")
            continue

        print(f"\n[PATIENT] {patient_id} | y_true = {true_cls} | #generated = {len(cf_files)}")

        sum_probs_cnn = torch.zeros(num_classes, device=device)
        sum_w_cnn = 0.0

        sum_d_per_target = {c_idx: 0.0 for c_idx in range(num_classes)}
        sum_w_per_target = {c_idx: 0.0 for c_idx in range(num_classes)}
        count_imgs_per_target = {c_idx: 0 for c_idx in range(num_classes)}

        for img_path in cf_files:
            m_cf = cf_pattern.match(img_path.stem)
            if not m_cf:
                print(f"[WARN] Filename {img_path.name} does not match cf pattern, skipping.")
                continue

            target_cls = m_cf.group(1)
            idx_str = m_cf.group(2)
            frame_str = m_cf.group(3)

            if target_cls not in class_to_idx:
                print(f"[WARN] target_cls '{target_cls}' not in class_names, skipping {img_path.name}.")
                continue

            target_idx = class_to_idx[target_cls]
            img_idx = int(idx_str) if idx_str is not None else 0

            try:
                frame_idx = int(frame_str)
            except Exception:
                frame_idx = None

            w_frame = frame_weights.get(frame_idx, 1.0)
            w_cls = class_weights.get(target_cls, 1.0)
            w_total = w_cls * w_frame

            # ----- CNN -----
            pil_cf = Image.open(img_path)
            x_cnn = preprocess_cnn(pil_cf).unsqueeze(0).to(device)
            with torch.no_grad():
                logits = model(x_cnn)
                probs = F.softmax(logits, dim=1)[0]
            sum_probs_cnn += w_total * probs
            sum_w_cnn += w_total

            # ----- LPIPS -----
            real_path = real_by_idx.get(img_idx, real_files[0])
            if img_idx not in real_tensor_cache:
                real_pil = Image.open(real_path)
                real_t = preprocess_lpips(real_pil).to(device)
                real_tensor_cache[img_idx] = real_t
            else:
                real_t = real_tensor_cache[img_idx]

            cf_t = preprocess_lpips(pil_cf).to(device)
            with torch.no_grad():
                d_lpips = lpips_fn(real_t, cf_t).item()

            sum_d_per_target[target_idx] += w_total * d_lpips
            sum_w_per_target[target_idx] += w_total
            count_imgs_per_target[target_idx] += 1

        if sum_w_cnn == 0.0:
            print(f"[WARN] sum_w_cnn == 0 for patient {patient_id}, skipping.")
            continue

        # CNN patient-level
        p_cnn = (sum_probs_cnn / sum_w_cnn).detach().cpu().numpy()
        pred_idx_cnn = int(p_cnn.argmax())
        pred_cls_cnn = idx_to_class[pred_idx_cnn]
        max_p_cnn = float(p_cnn.max())

        # LPIPS -> D -> p_lpips
        D = np.zeros(num_classes, dtype=np.float32)
        for c_idx in range(num_classes):
            if sum_w_per_target[c_idx] > 0:
                D[c_idx] = float(sum_d_per_target[c_idx] / sum_w_per_target[c_idx])
            else:
                D[c_idx] = np.inf

        p_lpips = lpips_distances_to_probs(D, temp=lpips_temp)
        pred_idx_lpips = int(p_lpips.argmax())
        pred_cls_lpips = idx_to_class[pred_idx_lpips]
        max_p_lpips = float(p_lpips.max())

        # Ensamble gated
        if max_p_cnn >= tau_cnn:
            p_ens = p_cnn.copy()
            gate_mode = "cnn_only"
        elif max_p_lpips >= tau_lpips:
            p_ens = p_lpips.copy()
            gate_mode = "lpips_only"
        else:
            p_ens = alpha_cnn_gated * p_cnn + (1.0 - alpha_cnn_gated) * p_lpips
            gate_mode = "mixed"

        pred_idx_ens = int(p_ens.argmax())
        pred_cls_ens = idx_to_class[pred_idx_ens]

        print(f"  -> p_cnn       = {p_cnn} (max={max_p_cnn:.3f})")
        print(f"  -> D (LPIPS)   = {D}")
        print(f"  -> p_lpips     = {p_lpips} (max={max_p_lpips:.3f})")
        print(f"  -> gate_mode   = {gate_mode}")
        print(f"  -> p_ensemble  = {p_ens}")
        print(f"  -> y_pred_cnn      = {pred_cls_cnn}")
        print(f"  -> y_pred_lpips    = {pred_cls_lpips}")
        print(f"  -> y_pred_ensemble = {pred_cls_ens}")

        # Acumular global
        y_true_all.append(true_idx)
        y_pred_cnn_all.append(pred_idx_cnn)
        y_pred_lpips_all.append(pred_idx_lpips)
        y_pred_ens_all.append(pred_idx_ens)
        p_cnn_all.append(p_cnn)
        p_lpips_all.append(p_lpips)

        # Fila CSV
        row = {
            "patient_id": patient_id,
            "y_true": true_cls,
            "y_pred_cnn": pred_cls_cnn,
            "y_pred_lpips": pred_cls_lpips,
            "y_pred_ensemble_gated": pred_cls_ens,
            "gate_mode": gate_mode,
            "max_p_cnn": max_p_cnn,
            "max_p_lpips": max_p_lpips,
        }
        for i, c in enumerate(class_names):
            row[f"p_cnn_{c}"] = float(p_cnn[i])
            row[f"D_{c}"] = float(D[i]) if np.isfinite(D[i]) else float("inf")
            row[f"p_lpips_{c}"] = float(p_lpips[i])
            row[f"p_ens_gated_{c}"] = float(p_ens[i])
            row[f"n_imgs_{c}"] = int(count_imgs_per_target[i])
            row[f"sum_w_{c}"] = float(sum_w_per_target[i])

        all_rows.append(row)

    # Guardar CSV
    out_csv.parent.mkdir(parents=True, exist_ok=True)
    fieldnames = (
        ["patient_id", "y_true", "y_pred_cnn", "y_pred_lpips", "y_pred_ensemble_gated",
         "gate_mode", "max_p_cnn", "max_p_lpips"]
        + [f"p_cnn_{c}" for c in class_names]
        + [f"D_{c}" for c in class_names]
        + [f"p_lpips_{c}" for c in class_names]
        + [f"p_ens_gated_{c}" for c in class_names]
        + [f"n_imgs_{c}" for c in class_names]
        + [f"sum_w_{c}" for c in class_names]
    )

    with out_csv.open("w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        for r in all_rows:
            writer.writerow(r)

    print(f"\n[INFO] Saved patient-level results to: {out_csv}")

    if len(y_true_all) == 0:
        print("[WARN] No patients evaluated. Check your classification_root.")
        return

    # Convertir a arrays
    y_true_all = np.array(y_true_all, dtype=int)
    y_pred_cnn_all = np.array(y_pred_cnn_all, dtype=int)
    y_pred_lpips_all = np.array(y_pred_lpips_all, dtype=int)
    y_pred_ens_all = np.array(y_pred_ens_all, dtype=int)
    p_cnn_all = np.vstack(p_cnn_all)
    p_lpips_all = np.vstack(p_lpips_all)

    print("\n================ METRICS ==================")

    # --- CNN baseline ---
    metrics_cnn = print_metrics("CNN baseline", y_true_all, y_pred_cnn_all)
    print_class_metrics("CNN baseline", y_true_all, y_pred_cnn_all, class_names)

    # --- LPIPS baseline ---
    metrics_lpips = print_metrics("LPIPS baseline", y_true_all, y_pred_lpips_all)
    print_class_metrics("LPIPS baseline", y_true_all, y_pred_lpips_all, class_names)

    # --- Gated ensemble ---
    metrics_gated = print_metrics("Gated ensemble", y_true_all, y_pred_ens_all)
    print_class_metrics("Gated ensemble", y_true_all, y_pred_ens_all, class_names)

    # --- Majority vote ---
    y_mv = majority_vote([y_pred_cnn_all, y_pred_lpips_all])
    metrics_mv = print_metrics("Majority vote", y_true_all, y_mv)
    print_class_metrics("Majority vote", y_true_all, y_mv, class_names)

    # --- Weighted majority ---
    y_wmv = weighted_majority_vote(
        [y_pred_cnn_all, y_pred_lpips_all],
        weights=[WEIGHT_CNN_MAJ, WEIGHT_LPIPS_MAJ],
    )
    metrics_wmv = print_metrics("Weighted majority", y_true_all, y_wmv)
    print_class_metrics("Weighted majority", y_true_all, y_wmv, class_names)

    # --- Bayes fusion (log-linear con ALPHA_BAYES) ---
    probs_bayes = fuse_loglinear(p_cnn_all, p_lpips_all, alpha_cnn=ALPHA_BAYES)
    y_bayes = probs_bayes.argmax(axis=1)
    metrics_bayes = print_metrics("Bayes fusion", y_true_all, y_bayes)
    print_class_metrics("Bayes fusion", y_true_all, y_bayes, class_names)

    # --- KL-based fusion (log-linear con ALPHA_KL) ---
    probs_kl = fuse_loglinear(p_cnn_all, p_lpips_all, alpha_cnn=ALPHA_KL)
    y_kl = probs_kl.argmax(axis=1)
    metrics_kl = print_metrics("KL-based fusion", y_true_all, y_kl)
    print_class_metrics("KL-based fusion", y_true_all, y_kl, class_names)

    # --- Joint posterior (log-linear con ALPHA_JOINT) ---
    probs_joint = fuse_loglinear(p_cnn_all, p_lpips_all, alpha_cnn=ALPHA_JOINT)
    y_joint = probs_joint.argmax(axis=1)
    metrics_joint = print_metrics("Joint posterior", y_true_all, y_joint)
    print_class_metrics("Joint posterior", y_true_all, y_joint, class_names)

    # --- Dempster–Shafer ---
    probs_ds = fuse_dempster_shafer_singleton(p_cnn_all, p_lpips_all)
    y_ds = probs_ds.argmax(axis=1)
    metrics_ds = print_metrics("Dempster–Shafer", y_true_all, y_ds)
    print_class_metrics("Dempster–Shafer", y_true_all, y_ds, class_names)

    print("\n[INFO] Done. Metrics computed for:")
    print("  - CNN baseline")
    print("  - LPIPS baseline")
    print("  - Gated ensemble")
    print("  - Majority vote")
    print("  - Weighted majority")
    print("  - Bayes fusion")
    print("  - KL-based fusion")
    print("  - Joint posterior fusion")
    print("  - Dempster–Shafer fusion")


# ============================================================
# MAIN
# ============================================================

if __name__ == "__main__":
    run_ensemble_and_fusions(
        classification_root=CLASSIFICATION_ROOT,
        ckpt_path=CKPT_PATH,
        out_csv=OUT_CSV,
        class_names=CLASS_NAMES,
        frame_weights=FRAME_WEIGHTS,
        class_weights=CLASS_WEIGHTS,
        alpha_cnn_gated=ALPHA_CNN_GATED,
        tau_cnn=TAU_CNN,
        tau_lpips=TAU_LPIPS,
        lpips_temp=LPIPS_TEMP,
    )
```

```python
# meta_fusion_classifier.py
# -*- coding: utf-8 -*-
"""
Meta-classifier for decision-level fusion between CNN and LPIPS.

This script:
- Loads patient-level predictions from results_fusion_all.csv
  (produced by the fusion_cnn_lpips_all.py script).
- Builds feature vectors combining:
    * CNN probabilities
    * LPIPS probabilities
    * Max probabilities
    * Entropies
    * Absolute differences |p_cnn - p_lpips|
    * Disagreement indicator between CNN and LPIPS
- Trains a multinomial Logistic Regression meta-classifier
  using StratifiedKFold cross-validation at patient level.
- Reports:
    * Fold-wise accuracy and macro-F1
    * Mean accuracy and macro-F1
    * Per-class precision/recall/F1 (classification_report)
    * Confusion matrix for cross-validated predictions
"""

from pathlib import Path
from typing import List, Tuple, Dict

import numpy as np
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix,
)

# ============================================================
# CONFIG
# ============================================================

# Path to the CSV generated by your fusion script
INPUT_CSV = Path(
    "/content/drive/MyDrive/TESIS/xai_batch_testset/results_fusion_all.csv"
)

# Class names (same order as your main pipeline)
CLASS_NAMES = ["glioma", "meningioma", "pituitary"]


# ============================================================
# FEATURE ENGINEERING
# ============================================================

def compute_entropy(probs: np.ndarray, eps: float = 1e-12) -> np.ndarray:
    """
    Compute entropy H(p) = -sum_c p_c log p_c for each sample.
    probs: array [N, C]
    Returns: array [N]
    """
    p = np.clip(probs, eps, 1.0)
    return -np.sum(p * np.log(p), axis=1)


def build_features(
    df: pd.DataFrame,
    class_names: List[str],
) -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Build feature matrix X and label vector y from the patient-level CSV.

    Features per patient (row):
        - p_cnn_c for each class
        - p_lpips_c for each class
        - max_p_cnn
        - max_p_lpips
        - entropy_cnn
        - entropy_lpips
        - |p_cnn_c - p_lpips_c| for each class
        - disagreement_flag (1 if argmax differs, 0 otherwise)

    Returns:
        X: [N, D] float32
        y: [N] int labels
        patient_ids: [N] object array (for reference)
    """
    class_to_idx = {c: i for i, c in enumerate(class_names)}

    # Ground truth as integer labels
    y_col = df["y_true"]
    y = np.array([class_to_idx[str(v)] for v in y_col], dtype=int)

    # CNN and LPIPS probabilities
    p_cnn_list = []
    p_lpips_list = []
    for c in class_names:
        p_cnn_list.append(df[f"p_cnn_{c}"].astype(float).values)
        p_lpips_list.append(df[f"p_lpips_{c}"].astype(float).values)

    p_cnn = np.stack(p_cnn_list, axis=1)     # [N, C]
    p_lpips = np.stack(p_lpips_list, axis=1) # [N, C]

    # Max probabilities and argmax
    max_p_cnn = p_cnn.max(axis=1)            # [N]
    max_p_lpips = p_lpips.max(axis=1)        # [N]
    argmax_cnn = p_cnn.argmax(axis=1)        # [N]
    argmax_lpips = p_lpips.argmax(axis=1)    # [N]

    # Entropies
    H_cnn = compute_entropy(p_cnn)           # [N]
    H_lpips = compute_entropy(p_lpips)       # [N]

    # Absolute differences |p_cnn - p_lpips|
    abs_diff = np.abs(p_cnn - p_lpips)       # [N, C]

    # Disagreement flag
    disagree_flag = (argmax_cnn != argmax_lpips).astype(float)  # [N]

    # Stack all features
    # Order:
    #   [p_cnn (3), p_lpips (3), max_p_cnn, max_p_lpips,
    #    H_cnn, H_lpips, abs_diff (3), disagree_flag]
    features = [
        p_cnn,
        p_lpips,
        max_p_cnn[:, None],
        max_p_lpips[:, None],
        H_cnn[:, None],
        H_lpips[:, None],
        abs_diff,
        disagree_flag[:, None],
    ]
    X = np.concatenate(features, axis=1).astype(np.float32)  # [N, D]

    patient_ids = df["patient_id"].values

    return X, y, patient_ids


# ============================================================
# META-CLASSIFIER TRAINING (CROSS-VALIDATION)
# ============================================================

def run_meta_classifier(
    X: np.ndarray,
    y: np.ndarray,
    class_names: List[str],
    n_splits: int = 5,
    random_state: int = 42,
):
    """
    Train and evaluate a meta-classifier using StratifiedKFold.

    - Uses LogisticRegression (multinomial).
    - Reports fold-wise ACC and macro-F1.
    - Reports overall classification_report and confusion_matrix for
      cross-validated predictions.
    """
    skf = StratifiedKFold(
        n_splits=n_splits,
        shuffle=True,
        random_state=random_state,
    )

    all_y_true = []
    all_y_pred = []

    fold_metrics = []

    print(f"[INFO] Starting StratifiedKFold with {n_splits} splits...")
    fold_idx = 0

    for train_idx, test_idx in skf.split(X, y):
        fold_idx += 1
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        clf = LogisticRegression(
            multi_class="multinomial",
            solver="lbfgs",
            max_iter=1000,
        )
        clf.fit(X_train, y_train)

        y_pred = clf.predict(X_test)

        acc = accuracy_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred, average="macro")

        fold_metrics.append((acc, f1))

        all_y_true.append(y_test)
        all_y_pred.append(y_pred)

        print(f"[FOLD {fold_idx}] ACC = {acc:.4f} | F1_macro = {f1:.4f}")

    # Concatenate all folds
    all_y_true = np.concatenate(all_y_true, axis=0)
    all_y_pred = np.concatenate(all_y_pred, axis=0)

    # Global metrics
    mean_acc = np.mean([m[0] for m in fold_metrics])
    mean_f1 = np.mean([m[1] for m in fold_metrics])

    print("\n========== META-CLASSIFIER (LogReg) ==========")
    print(f"Mean ACC over folds      = {mean_acc:.4f}")
    print(f"Mean F1_macro over folds = {mean_f1:.4f}")

    print("\nPer-class metrics (cross-validated predictions):")
    print(classification_report(all_y_true, all_y_pred,
                                target_names=class_names,
                                digits=4))

    print("Confusion matrix (all folds):")
    print(confusion_matrix(all_y_true, all_y_pred))


# ============================================================
# MAIN
# ============================================================

def main():
    print(f"[INFO] Loading CSV from: {INPUT_CSV}")
    df = pd.read_csv(INPUT_CSV)

    print(f"[INFO] Number of patients: {len(df)}")

    # Build features and labels
    X, y, patient_ids = build_features(df, CLASS_NAMES)
    print(f"[INFO] Feature matrix shape: {X.shape}")
    print(f"[INFO] Labels shape: {y.shape}")

    # Train and evaluate meta-classifier
    run_meta_classifier(X, y, CLASS_NAMES, n_splits=5, random_state=42)


if __name__ == "__main__":
    main()
```
