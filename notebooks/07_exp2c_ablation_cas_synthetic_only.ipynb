```python
# ===========================
# 1) Generation: StyleGAN2-ADA
# ===========================
import re, json, math
from pathlib import Path
from typing import List, Tuple, Optional, Dict

import numpy as np
from PIL import Image

import torch
import sys, os # Added imports for sys and os

# Robust loader for StyleGAN2-ADA G_ema
def load_stylegan2ada_generator(gan_pkl: str, device: str = "cuda"):
    # Add current directory to sys.path to help resolve module imports for pickled objects
    if os.getcwd() not in sys.path:
        sys.path.insert(0, os.getcwd()) # Use insert(0, ...) to prioritize current dir

    # Explicitly import training.networks to help resolve classes during unpickling
    try:
        import training.networks # noqa
    except Exception as e:
        print(f"Warning: Could not pre-import training.networks: {e}")

    dev = torch.device(device if torch.cuda.is_available() else "cpu")
    G = None
    err_msgs = []
    try:
        # StyleGAN2-ADA PyTorch (NVLabs) typical import
        from legacy import load_network_pkl  # NVLabs repo
        with open(gan_pkl, 'rb') as f:
            data = load_network_pkl(f)
        G = data.get('G_ema', None) or data.get('G', None)
    except Exception as e:
        err_msgs.append(f"[legacy] {e}")

    if G is None:
        try:
            # Some forks expose torch_utils as path
            from torch_utils import persistence  # noqa
            from legacy import load_network_pkl  # noqa
            with open(gan_pkl, 'rb') as f:
                data = load_network_pkl(f)
            G = data.get('G_ema', None) or data.get('G', None)
        except Exception as e:
            err_msgs.append(f"[torch_utils] {e}")

    if G is None:
        raise RuntimeError("Could not load G_ema from .pkl. Hints: mount repo, correct .pkl path. "
                           + " | ".join(err_msgs))

    G = G.to(dev).eval()
    if getattr(G, 'c_dim', 0) <= 0:
        raise ValueError(f"This generator is not conditional (c_dim={getattr(G,'c_dim',0)}).")
    return G, dev

@torch.no_grad()
def generate_synth_folders(
    gan_pkl: str,
    out_root: str,
    class_names: List[str],
    n_list: List[int],
    truncations: List[float],
    seed: int = 42,
    batch: int = 32,
    to_grayscale: bool = True
) -> List[Path]:
    """
    Generates per-class folders for each (N, psi) combination:

      out_root/
        N{N}_psi{psi}/
          <class_0>/*.png
          <class_1>/*.png
          ...

    Returns the list of created run folders.
    """
    out_root = Path(out_root)
    out_root.mkdir(parents=True, exist_ok=True)

    G, device = load_stylegan2ada_generator(gan_pkl, device="cuda")
    z_dim = int(getattr(G, "z_dim", 512))
    num_classes = len(class_names)
    created: List[Path] = []

    rng = np.random.RandomState(seed)

    for N in n_list:
        if N <= 0:
            # skip creating empty runs
            continue
        for psi in truncations:
            run_dir = out_root / f"N{int(N)}_psi{float(psi):.2f}"
            for cname in class_names:
                (run_dir / cname).mkdir(parents=True, exist_ok=True)

            # Generate per class
            for ci, cname in enumerate(class_names):
                saved = 0
                while saved < N:
                    bs = min(batch, N - saved)
                    z = torch.from_numpy(rng.randn(bs, z_dim)).to(device)
                    c = torch.zeros(bs, G.c_dim, device=device)
                    # one-hot condition
                    if ci >= G.c_dim:
                        raise ValueError(f"Class index {ci} exceeds G.c_dim={G.c_dim}. "
                                         "Check label order / class_names.")
                    c[:, ci] = 1.0
                    # Forward (handle signature differences)
                    try:
                        imgs = G(z, c, truncation_psi=float(psi), noise_mode='const')
                    except TypeError:
                        imgs = G(z, c, truncation_psi=float(psi))

                    # Convert to uint8 [H,W] grayscale (dataset is mono)
                    imgs = (imgs * 127.5 + 128).clamp(0, 255).to(torch.uint8).cpu()  # [B, C, H, W]
                    for k in range(bs):
                        x = imgs[k]
                        if x.shape[0] == 1:
                            arr = x[0].numpy()
                            pil = Image.fromarray(arr, mode='L')
                        else:
                            # stylegan outputs RGB; convert to L if requested
                            arr = x.permute(1, 2, 0).numpy()
                            pil = Image.fromarray(arr, mode='RGB')
                            if to_grayscale:
                                pil = pil.convert('L')
                        fname = run_dir / cname / f"synth_{cname}_psi{float(psi):.2f}_{saved+k:05d}.png"
                        pil.save(fname)
                    saved += bs

            created.append(run_dir)
            print(f"[GEN] Created: {run_dir} | total ~{N*num_classes} images")

    return created


# ==========================================
# 2) Evaluation: CNN ckpt on generated sets
# ==========================================
import csv
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from sklearn.metrics import (
    accuracy_score, precision_recall_fscore_support
)

class LabeledFolderDataset(torch.utils.data.Dataset):
    def __init__(self, root: str, class_names: List[str], image_size: int = 256):
        self.root = Path(root)
        self.class_names = class_names
        idx = {c:i for i,c in enumerate(class_names)}
        self.items = []
        for cname in class_names:
            for p in (self.root / cname).rglob("*"):
                if p.suffix.lower() in {".png",".jpg",".jpeg",".bmp",".tif",".tiff"}:
                    self.items.append((p, idx[cname]))
        self.tf = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor()
        ])
        self.norm = transforms.Normalize(mean=[0.5,0.5,0.5], std=[0.5,0.5,0.5])

    def __len__(self): return len(self.items)

    def __getitem__(self, i):
        p, y = self.items[i]
        img = Image.open(p).convert("L")
        img = self.tf(img).repeat(3,1,1)
        img = self.norm(img)
        return img, y, str(p)

def build_resnet50(num_classes: int) -> nn.Module:
    m = models.resnet50(weights=None)
    in_f = m.fc.in_features
    m.fc = nn.Linear(in_f, num_classes)
    return m

@torch.no_grad()
def _eval_one_folder(model: nn.Module, folder: Path, class_names: List[str],
                     device: torch.device, image_size: int, batch_size: int) -> Dict[str, float]:
    ds = LabeledFolderDataset(str(folder), class_names, image_size=image_size)
    if len(ds) == 0:
        return {"n": 0, "acc": np.nan, "prec_macro": np.nan, "rec_macro": np.nan, "f1_macro": np.nan}
    dl = DataLoader(ds, batch_size=batch_size, shuffle=False, num_workers=4, pin_memory=True)
    ys_true, ys_pred = [], []

    for imgs, ys, _ in dl:
        imgs = imgs.to(device, non_blocking=True)
        logits = model(imgs)
        probs = F.softmax(logits, dim=1)
        pred = probs.argmax(dim=1).cpu().numpy()
        ys_true.append(ys.numpy()); ys_pred.append(pred)

    y_true = np.concatenate(ys_true); y_pred = np.concatenate(ys_pred)
    acc = float(accuracy_score(y_true, y_pred))
    pm, rm, fm, _ = precision_recall_fscore_support(y_true, y_pred, average='macro', zero_division=0)
    return {"n": int(len(y_true)), "acc": acc, "prec_macro": float(pm), "rec_macro": float(rm), "f1_macro": float(fm)}

def _parse_N_psi(folder_name: str) -> Tuple[Optional[int], Optional[float]]:
    m = re.search(r"N(\d+)", folder_name)
    n = int(m.group(1)) if m else None
    m2 = re.search(r"psi([0-9]+(?:\.[0-9]+)?)", folder_name)
    psi = float(m2.group(1)) if m2 else None
    return n, psi

def eval_cnn_on_synth_root(
    ckpt_path: str,
    synth_root: str,
    class_names: Optional[List[str]] = None,
    image_size: int = 256,
    batch_size: int = 64,
    device: str = "cuda",
    save_csv: bool = True
):
    """
    Finds all subfolders in synth_root that look like runs (contain subfolders for all classes),
    evaluates the CNN checkpoint on each, and returns a sorted list of dict rows + prints a summary table.
    """
    dev = torch.device(device if torch.cuda.is_available() else "cpu")
    ckpt = torch.load(ckpt_path, map_location=dev)

    # Resolve class_names order
    ckpt_classes = None
    if isinstance(ckpt, dict) and "class_names" in ckpt:
        ckpt_classes = list(ckpt["class_names"])
    if class_names is None:
        if ckpt_classes is None:
            raise ValueError("Provide class_names or store 'class_names' in the checkpoint.")
        class_names = ckpt_classes

    # Build / load model
    m = build_resnet50(num_classes=len(class_names)).to(dev).eval()
    state = ckpt["model"] if (isinstance(ckpt, dict) and "model" in ckpt) else ckpt
    m.load_state_dict(state, strict=True)

    synth_root = Path(synth_root)
    run_folders = []
    for d in sorted([p for p in synth_root.iterdir() if p.is_dir()]):
        # Must contain all class subfolders
        if all((d / c).is_dir() for c in class_names):
            run_folders.append(d)

    rows = []
    for d in run_folders:
        res = _eval_one_folder(m, d, class_names, dev, image_size, batch_size)
        N, psi = _parse_N_psi(d.name)
        rows.append({
            "folder": d.name,
            "N": N,
            "psi": psi,
            "n_images": res["n"],
            "accuracy": res["acc"],
            "precision_macro": res["prec_macro"],
            "recall_macro": res["rec_macro"],
            "f1_macro": res["f1_macro"]
        })
        print(f"[EVAL] {d.name}  ->  acc={res['acc']:.3f}  f1_macro={res['f1_macro']:.3f}  (n={res['n']})")

    # Sort by (psi, N)
    rows.sort(key=lambda r: (float('inf') if r["psi"] is None else r["psi"],
                             float('inf') if r["N"]   is None else r["N"]))

    # Optional save CSV
    if save_csv:
        out_csv = synth_root / "summary_synth_eval.csv"
        import csv
        with open(out_csv, "w", newline="", encoding="utf-8") as f:
            w = csv.DictWriter(f, fieldnames=list(rows[0].keys()) if rows else
                               ["folder","N","psi","n_images","accuracy","precision_macro","recall_macro","f1_macro"])
            w.writeheader()
            for r in rows:
                w.writerow(r)
        print(f"[OK] Summary saved: {out_csv}")

    return rows
```

Para el experimento 2C se quizo comparar si incluir imagenes sinteticas generadas por el stylegan-c a una red cnn mejoran las metricas a otra que no tiene estas imagenes. para esto se separaron los expreimentos en dos datasets. el dataset completo que tienen todas las imagenes y el data set reducido que solo tiene imagenes de buena calidad. La diferencia entre estos dos sets es de 761 pacientes, es decir, aprox un 25% menos de imagenes (refernciar tabla de reduceddataset).
Ambos data sets estan con control de leakage de pacientes.

Se hicieron diversos entrenamientos con distintos valores de truncation para ver como varian las metricas en funcion de esto. Los experimentos se corrieron en fold para que fueran mas robustos y no entrara en juego el azar de una buena corrida, ademas todos los entrenamientos se hicieron con el mismo seed para garantizar comparabilidad. En la tabla 5.12 3k fold se puede observar un aumento de accuracy de 0.93 a 0.945 desde la baseline hasta la augmentacion. Para hacer los resultados mas robustos aun se hizo el experimento con 5kfold y variando el truncation tabla 5.13 5kfold, donde tambien se puede observar una mejora de 0.941 a 0.948 al usar truncation trick 1.4. Con esto se puede notar que no existe un truncation trick fijo que sea mejor para todos los modelos, esto sucede porque cada vez que se generan imagenes, estas son diferentes y esto puede variar los resultados de las metricas.

La media de kfold solo nos indica que tan estable es el modelo reduciendo la varianza al correr varios modelos, pero estos modelos se pueden juntar y crear uno mejor. por esto se crea un ensemble con los folds. Al hacer esto se puede notar el baseline aumenta de 0.941 a 0.947 en las imagenes reales, esto es porque el modelo llama a cada una de los folds para obtener un resultado y se combinan. de esta manera en la tabla 5.14 se nota que el acc pasa de 0.947 a 0.956 usando el truncation trick de 1.4. Tambien se analiza la performance por clase en la tabla 5.16 ensemble 5k donde se puede observar que todas los f1 aumentan del real al augmentado y el acc aumenta de 0.947 a 0.956.
