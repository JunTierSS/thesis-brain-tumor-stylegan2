```python
# -*- coding: utf-8 -*-
"""
K-Fold training (by patient) for brain tumor classification (ResNet50)
- No-leak by patient_id (groups)
- StratifiedGroupKFold if available; fallback to GroupKFold
- Optional StyleGAN2-ADA augmentation:
    * Real-only run (no GAN)
    * GAN-balanced run: para cada fold se agrega GAN hasta igualar la clase mayor (por nº de imágenes REALES)
- Per-fold metrics + CV summary
- OOF (out-of-fold) predictions & metrics (one row per pool sample, predicted by a model that didn't see it)
- External TEST evaluation per fold + K-fold ensemble (avg probs across folds)
- Full INTTEST metrics per fold + INTTEST ensemble, and CV summary with mean±std for VAL/INTTEST/TEST
- Optional final fine-tuning epochs using only REAL images (no GAN) per fold and in full-train

Author: Jun + ChatGPT
"""

import os, re, io, json, time, random, csv, math, shutil, pickle
from pathlib import Path
from typing import List, Tuple, Dict, Optional
from dataclasses import dataclass, asdict

import numpy as np
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models

from sklearn.metrics import (
    accuracy_score, precision_recall_fscore_support,
    classification_report, confusion_matrix
)

# ====== Splitters ======
try:
    from sklearn.model_selection import StratifiedGroupKFold
    HAS_SGTK = True
except Exception:
    HAS_SGTK = False
    from sklearn.model_selection import GroupKFold

# ====== Utils ======

def set_seed(seed: int = 42):
    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True
    try:
        torch.set_float32_matmul_precision("high")
    except Exception:
        pass

def timestamp() -> str:
    return time.strftime("%Y%m%d-%H%M%S")

def extract_patient_id(p: Path) -> str:
    stem = p.stem
    token = stem.split("_")[0]
    return token.strip().lower()

IMG_EXTS = {".png", ".jpg", ".jpeg", ".bmp", ".tif", ".tiff"}

# ====== Config ======

@dataclass
class Config:
    data_root: str
    out_dir: str
    test_root: str = ""
    gan_pkl: str = ""
    images_per_class: int = 0        # sólo se usa si balance_gan_to_max=False
    truncation: float = 1.0
    internal_test_ratio: float = 0.0   # optional: holdout before CV
    image_size: int = 256
    batch_size: int = 32
    epochs: int = 15
    lr: float = 1e-3
    weight_decay: float = 1e-4
    num_workers: int = 4
    seed: int = 42
    kfolds: int = 5
    amp: bool = True
    pretrained: bool = True
    train_full_after_cv: bool = False   # optional: retrain on full pool at end

    # Fine-tuning final sólo con reales
    fine_tune_real_epochs: int = 0      # 0 = desactivado

    # NUEVO: balancear clases con GAN hasta la clase mayor de REALES
    balance_gan_to_max: bool = False

# ====== Dataset ======

class MRIDataset(Dataset):
    def __init__(self, items: List[Tuple[Path, int, str]], image_size: int = 256, aug: bool = False):
        self.items = items
        self.image_size = image_size
        self.aug = aug
        mean = [0.5, 0.5, 0.5]; std = [0.5, 0.5, 0.5]
        if aug:
            self.tf = transforms.Compose([
                transforms.RandomResizedCrop(image_size, scale=(0.9, 1.0), ratio=(0.95, 1.05)),
                transforms.RandomHorizontalFlip(),
                transforms.RandomVerticalFlip(),
                transforms.RandomRotation(10),
                transforms.ToTensor(),
            ])
        else:
            self.tf = transforms.Compose([
                transforms.Resize((image_size, image_size)),
                transforms.ToTensor(),
            ])
        self.norm = transforms.Normalize(mean=mean, std=std)

    def __len__(self):
        return len(self.items)

    def __getitem__(self, idx):
        path, y, pid = self.items[idx]
        img = Image.open(path).convert("L")
        img = self.tf(img)           # [1,H,W]
        img = img.repeat(3, 1, 1)    # [3,H,W]
        img = self.norm(img)
        return img, y, str(path), pid

# ====== Scan & Helpers ======

def scan_folder(root: Path) -> Tuple[List[Tuple[Path,int,str]], List[str]]:
    assert root.is_dir(), f"Not a directory: {root}"
    class_names = sorted([d.name for d in root.iterdir() if d.is_dir()])
    items: List[Tuple[Path,int,str]] = []
    for ci, cname in enumerate(class_names):
        cdir = root / cname
        for p in cdir.rglob("*"):
            if p.suffix.lower() in IMG_EXTS:
                pid = extract_patient_id(p)
                items.append((p, ci, pid))
    return items, class_names

def ensure_no_overlap(ids_a: List[str], ids_b: List[str]) -> Tuple[int, List[str]]:
    a = set(ids_a); b = set(ids_b)
    inter = sorted(a.intersection(b))
    return len(inter), inter

def summarize_id_counts(items: List[Tuple[Path,int,str]], class_names: List[str], split_name: str, out_dir: Path):
    from collections import defaultdict
    ids_by_class = defaultdict(set); imgs_by_class = defaultdict(int); all_ids = set()
    for p, ci, pid in items:
        ids_by_class[ci].add(pid)
        imgs_by_class[ci] += 1
        all_ids.add(pid)

    print(f"[ID-STATS] {split_name}: unique patient IDs total = {len(all_ids)} | images = {len(items)}")
    for ci, cname in enumerate(class_names):
        print(f"  - {cname}: ids={len(ids_by_class[ci])} images={imgs_by_class[ci]}")

    out_dir.mkdir(parents=True, exist_ok=True)
    with open(out_dir / f"id_counts_{split_name}.csv", 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(["class", "n_unique_ids", "n_images"])
        for ci, cname in enumerate(class_names):
            w.writerow([cname, len(ids_by_class[ci]), imgs_by_class[ci]])
        w.writerow(["__TOTAL__", len(all_ids), len(items)])

# ====== Model ======

def build_model(num_classes: int, pretrained: bool = True) -> nn.Module:
    m = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1 if pretrained else None)
    in_features = m.fc.in_features
    m.fc = nn.Linear(in_features, num_classes)
    return m

# ====== Eval (writes artifacts) ======

@torch.no_grad()
def evaluate(model: nn.Module, loader: DataLoader, device: torch.device, class_names: List[str], split_name: str, out_dir: Path):
    model.eval()
    all_probs, all_preds, all_labels = [], [], []
    all_paths, all_pids = [], []

    for imgs, ys, paths, pids in loader:
        imgs = imgs.to(device, non_blocking=True)
        ys = ys.to(device, non_blocking=True)
        logits = model(imgs)
        probs = F.softmax(logits, dim=1)
        preds = probs.argmax(dim=1)

        all_probs.append(probs.cpu().numpy())
        all_preds.append(preds.cpu().numpy())
        all_labels.append(ys.cpu().numpy())
        all_paths.extend(list(paths))
        all_pids.extend(list(pids))

    probs = np.concatenate(all_probs, axis=0) if all_probs else np.zeros((0, len(class_names)))
    preds = np.concatenate(all_preds, axis=0) if all_preds else np.zeros((0,), dtype=np.int64)
    labels = np.concatenate(all_labels, axis=0) if all_labels else np.zeros((0,), dtype=np.int64)

    acc = float(accuracy_score(labels, preds)) if len(labels) else 0.0
    prec_macro, rec_macro, f1_macro, _ = precision_recall_fscore_support(labels, preds, average='macro', zero_division=0)
    prec_weighted, rec_weighted, f1_weighted, _ = precision_recall_fscore_support(labels, preds, average='weighted', zero_division=0)
    per_class = classification_report(labels, preds, target_names=class_names, output_dict=True, zero_division=0)
    cm = confusion_matrix(labels, preds, labels=list(range(len(class_names))))

    metrics = {
        "overall": {
            "accuracy": acc,
            "precision_macro": prec_macro,
            "recall_macro": rec_macro,
            "f1_macro": f1_macro,
            "precision_weighted": prec_weighted,
            "recall_weighted": rec_weighted,
            "f1_weighted": f1_weighted,
            "num_samples": len(labels),
        },
        "per_class": per_class,
    }

    out_dir.mkdir(parents=True, exist_ok=True)
    with open(out_dir / f"metrics_{split_name}.json", 'w', encoding='utf-8') as f:
        json.dump(metrics, f, indent=2)

    with open(out_dir / f"metrics_overall_{split_name}.csv", 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(["split","accuracy","precision_macro","recall_macro","f1_macro","precision_weighted","recall_weighted","f1_weighted","num_samples"])
        o = metrics['overall']
        w.writerow([split_name, o['accuracy'], o['precision_macro'], o['recall_macro'], o['f1_macro'],
                    o['precision_weighted'], o['recall_weighted'], o['f1_weighted'], o['num_samples']])

    with open(out_dir / f"per_class_{split_name}.csv", 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(["class", "precision", "recall", "f1", "support"])
        for cname in class_names:
            row = per_class.get(cname, {"precision":0, "recall":0, "f1-score":0, "support":0})
            w.writerow([cname, row.get("precision",0), row.get("recall",0),
                        row.get("f1-score",0), row.get("support",0)])

    with open(out_dir / f"confusion_{split_name}.csv", 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(["true\\pred"] + class_names)
        for i, cname in enumerate(class_names):
            w.writerow([cname] + list(cm[i]))

    with open(out_dir / f"predictions_{split_name}.csv", 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(["path", "patient_id", "true", "pred"] + [f"p_{c}" for c in class_names])
        for i in range(len(labels)):
            w.writerow([all_paths[i], all_pids[i], class_names[labels[i]],
                        class_names[preds[i]]] + list(map(float, probs[i].tolist())))

    return metrics, cm

# Helper: fast inference to collect arrays (for OOF & ensemble)
@torch.no_grad()
def infer_loader(model: nn.Module, loader: DataLoader, device: torch.device):
    model.eval()
    paths, pids = [], []
    all_probs, all_preds, all_labels = [], [], []
    for imgs, ys, batch_paths, batch_pids in loader:
        imgs = imgs.to(device, non_blocking=True)
        ys = ys.to(device, non_blocking=True)
        logits = model(imgs)
        probs = F.softmax(logits, dim=1)
        preds = probs.argmax(dim=1)

        all_probs.append(probs.cpu().numpy())
        all_preds.append(preds.cpu().numpy())
        all_labels.append(ys.cpu().numpy())
        paths.extend(list(batch_paths))
        pids.extend(list(batch_pids))

    probs = np.concatenate(all_probs, axis=0) if all_probs else np.zeros((0, 0))
    preds = np.concatenate(all_preds, axis=0) if all_preds else np.zeros((0,), dtype=np.int64)
    labels = np.concatenate(all_labels, axis=0) if all_labels else np.zeros((0,), dtype=np.int64)
    return {"probs": probs, "preds": preds, "labels": labels, "paths": paths, "pids": pids}

# ====== Optional GAN loader (TRAIN only) ======

@torch.no_grad()
def _load_stylegan2_ada_G(gan_pkl: str, device: torch.device):
    try:
        import dnnlib
        from training import network as _dummy
        from legacy import load_network_pkl
    except Exception:
        try:
            from torch_utils import misc as _m  # noqa
            from torch_utils import training_stats as _ts  # noqa
            from legacy import load_network_pkl
        except Exception:
            return None
    try:
        with open(gan_pkl, 'rb') as f:
            data = load_network_pkl(f)
        G_ema = data.get('G_ema', None) or data.get('G', None)
        if G_ema is None:
            return None
        G_ema = G_ema.to(device)
        G_ema.eval()
        return G_ema
    except Exception:
        return None

@torch.no_grad()
def maybe_generate_gan_images(gan_pkl: str,
                              class_names: List[str],
                              n_per_class: int,
                              truncation: float,
                              out_dir: Path,
                              device: torch.device,
                              seed: int = 42) -> Optional[Path]:
    if (not gan_pkl) or n_per_class <= 0:
        return None
    G = _load_stylegan2_ada_G(gan_pkl, device)
    if G is None:
        print("[GAN] Could not load StyleGAN2-ADA G_ema. Skipping GAN augmentation.")
        return None
    if getattr(G, 'c_dim', 0) <= 0:
        print("[GAN] Generator is not conditional (c_dim=0). Skipping.")
        return None

    rng = np.random.RandomState(seed)
    synth_root = out_dir / "gan_synth"
    for cname in class_names:
        (synth_root / cname).mkdir(parents=True, exist_ok=True)

    batch = 16
    for ci, cname in enumerate(class_names):
        saved = 0
        while saved < n_per_class:
            bs = min(batch, n_per_class - saved)
            z = torch.from_numpy(rng.randn(bs, G.z_dim)).to(device)
            c = torch.zeros(bs, G.c_dim, device=device); c[:, ci] = 1.0
            try:
                img = G(z, c, truncation_psi=float(truncation), noise_mode='const')
            except TypeError:
                img = G(z, c, truncation_psi=float(truncation))
            img = (img * 127.5 + 128).clamp(0,255).to(torch.uint8).cpu()
            for k in range(bs):
                x = img[k]
                if x.shape[0] == 1:
                    arr = x[0].numpy()
                    pil = Image.fromarray(arr).convert('L')
                else:
                    arr = x.permute(1,2,0).numpy()
                    pil = Image.fromarray(arr).convert('L')
                fname = synth_root / cname / f"gan_{cname}_{saved+k:05d}.png"
                pil.save(fname)
            saved += bs
        print(f"[GAN] Saved {saved} images for class '{cname}'")
    return synth_root

def collate_fn(batch):
    imgs, ys, paths, pids = zip(*batch)
    return torch.stack(imgs, dim=0), torch.tensor(ys, dtype=torch.long), list(paths), list(pids)

# ====== Helper: construir subset balanceado real + GAN ======

def build_balanced_train_subset(real_train_subset: List[Tuple[Path,int,str]],
                                synth_items: List[Tuple[Path,int,str]],
                                num_classes: int,
                                seed: int) -> List[Tuple[Path,int,str]]:
    """
    Toma:
      - real_train_subset: sólo imágenes reales de ese fold
      - synth_items: TODAS las sintéticas (escaneadas desde gan_synth/)
    Devuelve:
      - lista con reales + suficientes GAN por clase para igualar la clase mayor (en REALES).
    """
    from collections import defaultdict

    # Reales por clase
    real_by_class: Dict[int, List[Tuple[Path,int,str]]] = defaultdict(list)
    for item in real_train_subset:
        _, ci, _ = item
        real_by_class[ci].append(item)

    # GAN por clase
    synth_by_class: Dict[int, List[Tuple[Path,int,str]]] = defaultdict(list)
    for item in synth_items:
        _, ci, _ = item
        synth_by_class[ci].append(item)

    # Target = tamaño de la clase mayor (sólo contando REALES)
    counts_real = [len(real_by_class[c]) for c in range(num_classes)]
    target = max(counts_real) if counts_real else 0
    print(f"[BAL] Real class counts (fold): {counts_real} -> target={target}")

    rng = random.Random(seed)
    balanced: List[Tuple[Path,int,str]] = list(real_train_subset)

    for ci in range(num_classes):
        need = target - len(real_by_class[ci])
        if need <= 0:
            continue
        pool = synth_by_class[ci]
        if not pool:
            print(f"[BAL] WARNING: no GAN images for class {ci}, cannot fully balance.")
            continue
        # muestramos con reemplazo si need > len(pool)
        for _ in range(need):
            balanced.append(rng.choice(pool))

    return balanced

# ====== K-Fold Train (CV + OOF + TEST + INTTEST ensemble) ======

def train_kfold(cfg: Config):
    if cfg.fine_tune_real_epochs < 0 or cfg.fine_tune_real_epochs > cfg.epochs:
        raise ValueError("fine_tune_real_epochs must be between 0 and cfg.epochs")

    base_run = Path(cfg.out_dir) / f"kfold_{cfg.kfolds}f_run_{timestamp()}"
    base_run.mkdir(parents=True, exist_ok=True)
    with open(base_run / 'config.json', 'w', encoding='utf-8') as f:
        json.dump(asdict(cfg), f, indent=2)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"[Device] Using {device}")

    # Scan pool
    all_items, class_names = scan_folder(Path(cfg.data_root))
    print(f"[Scan] Pool images: {len(all_items)} | Classes: {class_names}")

    # Optional internal-test holdout (by patient)
    inttest_subset: List[Tuple[Path,int,str]] = []
    if cfg.internal_test_ratio and cfg.internal_test_ratio > 0:
        from sklearn.model_selection import GroupShuffleSplit
        X_all = np.arange(len(all_items))
        y_all = np.array([lbl for _, lbl, _ in all_items])
        g_all = np.array([pid for _, _, pid in all_items])
        gss = GroupShuffleSplit(n_splits=1, test_size=cfg.internal_test_ratio, random_state=cfg.seed)
        trval_idx, inttest_idx = next(gss.split(X_all, y_all, g_all))
        base_pool = [all_items[i] for i in trval_idx]
        inttest_subset = [all_items[i] for i in inttest_idx]
        print(f"[Split] Internal test: {len(inttest_subset)} images ({cfg.internal_test_ratio*100:.1f}%)")
    else:
        base_pool = all_items

    # External test (leak-sanitized vs pool & internal test)
    ext_subset: List[Tuple[Path,int,str]] = []
    if cfg.test_root and len(cfg.test_root.strip()) > 0 and Path(cfg.test_root).exists():
        ext_items, _ = scan_folder(Path(cfg.test_root))
        ids_pool = set(pid for _,_,pid in base_pool)
        if len(inttest_subset) > 0:
            ids_pool |= set(pid for _,_,pid in inttest_subset)
        for it in ext_items:
            if it[2] not in ids_pool:
                ext_subset.append(it)
        removed = len(ext_items) - len(ext_subset)
        if removed > 0:
            print(f"[LEAK-FIX] Removed {removed} overlapping ext-test images by patient ID.")
        print(f"[Scan] External test images (sanitized): {len(ext_subset)}")

    # Build folds grouped by patient (SOLO con REALES)
    X = np.arange(len(base_pool))
    y = np.array([lbl for _, lbl, _ in base_pool])
    groups = np.array([pid for _, _, pid in base_pool])

    if cfg.kfolds < 2:
        raise ValueError("kfolds must be >= 2")

    if HAS_SGTK:
        splitter = StratifiedGroupKFold(n_splits=cfg.kfolds, shuffle=True, random_state=cfg.seed)
        folds_iter = list(splitter.split(X, y, groups))
        print("[Split] Using StratifiedGroupKFold")
    else:
        splitter = GroupKFold(n_splits=cfg.kfolds)
        folds_iter = list(splitter.split(X, groups=groups))
        print("[Split] Using GroupKFold (no stratification)")

    # ===== Pre-generate GAN synth una sola vez =====
    synth_items: List[Tuple[Path,int,str]] = []
    synth_root: Optional[Path] = None
    if cfg.gan_pkl and cfg.gan_pkl.strip():
        # si queremos balancear, ignoramos images_per_class y usamos el máximo global de REALES
        if cfg.balance_gan_to_max:
            from collections import defaultdict
            real_counts_global = defaultdict(int)
            for _, ci, _ in base_pool:
                real_counts_global[ci] += 1
            # Vector ordenado por índice de clase
            counts_vec = [real_counts_global[ci] for ci in range(len(class_names))]
            n_per_class = max(counts_vec) if counts_vec else 0
            print(f"[GAN] balance_gan_to_max=True -> global real counts per class = {counts_vec}, using n_per_class={n_per_class}")
        else:
            n_per_class = cfg.images_per_class

        if n_per_class > 0:
            synth_root = maybe_generate_gan_images(
                gan_pkl=cfg.gan_pkl,
                class_names=class_names,
                n_per_class=n_per_class,
                truncation=cfg.truncation,
                out_dir=base_run,
                device=device,
                seed=cfg.seed,
            )
            if synth_root is not None:
                synth_items, _ = scan_folder(synth_root)
                print(f"[GAN] Synth images generated (global): {len(synth_items)}")
        else:
            print("[GAN] n_per_class <= 0 -> no synthetic images will be generated.")
    else:
        print("[GAN] gan_pkl empty -> NO GAN will be used.")

    # Prepare fixed loaders for ensembles (TEST e INTTEST)
    test_loader_fixed: Optional[DataLoader] = None
    test_ref_paths = test_ref_pids = test_ref_labels = None
    if len(ext_subset) > 0:
        test_ds_fixed = MRIDataset(ext_subset, image_size=cfg.image_size, aug=False)
        test_loader_fixed = DataLoader(test_ds_fixed, batch_size=cfg.batch_size, shuffle=False,
                                       num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)
        # Se rellenan luego con infer_loader

    inttest_loader_fixed: Optional[DataLoader] = None
    int_ref_paths = int_ref_pids = int_ref_labels = None
    if len(inttest_subset) > 0:
        int_ds_fixed = MRIDataset(inttest_subset, image_size=cfg.image_size, aug=False)
        inttest_loader_fixed = DataLoader(int_ds_fixed, batch_size=cfg.batch_size, shuffle=False,
                                          num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)

    # ===== OOF buffers =====
    n_pool = len(base_pool)
    n_cls = len(class_names)
    oof_probs = np.full((n_pool, n_cls), np.nan, dtype=np.float32)
    oof_pred  = np.full((n_pool,), -1, dtype=np.int64)
    oof_true  = np.full((n_pool,), -1, dtype=np.int64)
    oof_path  = [""] * n_pool
    oof_pid   = [""] * n_pool

    # For ensembles
    test_probs_folds: List[np.ndarray] = []
    inttest_probs_folds: List[np.ndarray] = []

    # Per-fold training
    cv_rows = []
    for fold_id, (tr_idx, va_idx) in enumerate(folds_iter, start=1):
        print(f"\n===== FOLD {fold_id}/{cfg.kfolds} =====")
        set_seed(cfg.seed + fold_id)

        fold_dir = base_run / f"fold_{fold_id:02d}"
        fold_dir.mkdir(parents=True, exist_ok=True)

        # REAL train subset and VAL subset
        real_train_subset = [base_pool[i] for i in tr_idx]
        val_subset        = [base_pool[i] for i in va_idx]

        # Leak checks (sólo reales)
        ids_train = [pid for _,_,pid in real_train_subset]
        ids_val   = [pid for _,_,pid in val_subset]
        inter_n, inter_ids = ensure_no_overlap(ids_train, ids_val)
        if inter_n > 0:
            raise RuntimeError(f"[LEAK ERROR] Train↔Val share {inter_n} patient IDs (fold {fold_id}): {inter_ids[:10]}")

        if len(inttest_subset) > 0:
            ids_int = [pid for _,_,pid in inttest_subset]
            n1,_ = ensure_no_overlap(ids_train, ids_int)
            n2,_ = ensure_no_overlap(ids_val, ids_int)
            if n1>0 or n2>0:
                raise RuntimeError("[LEAK ERROR] Fold and INTTEST share patients")

        # ===== construir train_subset =====
        if cfg.balance_gan_to_max and len(synth_items) > 0:
            train_subset = build_balanced_train_subset(
                real_train_subset=real_train_subset,
                synth_items=synth_items,
                num_classes=len(class_names),
                seed=cfg.seed + fold_id,
            )
            print(f"[BAL] Fold {fold_id}: real={len(real_train_subset)} | train_balanced(real+GAN)={len(train_subset)}")
        else:
            # sin balanceo: opcionalmente append de todo synth_items
            train_subset = list(real_train_subset)
            if len(synth_items) > 0:
                train_subset = train_subset + synth_items
                print(f"[TRAIN] Fold {fold_id}: real={len(real_train_subset)} + synth={len(synth_items)} -> {len(train_subset)}")
            else:
                print(f"[TRAIN] Fold {fold_id}: only real images, n={len(real_train_subset)}")

        # Stats CSV
        summarize_id_counts(train_subset, class_names, f'fold{fold_id}_train', fold_dir)
        summarize_id_counts(val_subset,   class_names, f'fold{fold_id}_val',   fold_dir)
        if len(inttest_subset) > 0:
            summarize_id_counts(inttest_subset, class_names, f'fold{fold_id}_inttest', fold_dir)
        if len(ext_subset) > 0:
            summarize_id_counts(ext_subset,     class_names, f'fold{fold_id}_test',    fold_dir)

        # Loaders
        train_ds       = MRIDataset(train_subset,       image_size=cfg.image_size, aug=True)
        real_train_ds  = MRIDataset(real_train_subset,  image_size=cfg.image_size, aug=True)
        val_ds         = MRIDataset(val_subset,         image_size=cfg.image_size, aug=False)

        train_loader      = DataLoader(train_ds,      batch_size=cfg.batch_size, shuffle=True,
                                       num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)
        real_train_loader = DataLoader(real_train_ds, batch_size=cfg.batch_size, shuffle=True,
                                       num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)
        val_loader        = DataLoader(val_ds,        batch_size=cfg.batch_size, shuffle=False,
                                       num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)

        inttest_loader = None
        if len(inttest_subset) > 0:
            int_ds = MRIDataset(inttest_subset, image_size=cfg.image_size, aug=False)
            inttest_loader = DataLoader(int_ds, batch_size=cfg.batch_size, shuffle=False,
                                        num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)

        # Para TEST por fold usamos loader fijo
        test_loader = test_loader_fixed

        # Model/optim
        model = build_model(num_classes=len(class_names), pretrained=cfg.pretrained).to(device)
        optimizer = torch.optim.AdamW(model.parameters(), lr=cfg.lr, weight_decay=cfg.weight_decay)
        scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=cfg.epochs)
        criterion = nn.CrossEntropyLoss()

        amp_enabled = (cfg.amp and device.type == 'cuda')
        scaler = torch.amp.GradScaler('cuda', enabled=amp_enabled)

        best_f1 = -1.0

        # Epoch split: main (train_subset) + fine-tuning (real only)
        epochs_main = cfg.epochs - cfg.fine_tune_real_epochs
        epochs_ft   = cfg.fine_tune_real_epochs

        # ----- Phase 1: train con train_subset (real + GAN o sólo real) -----
        for epoch in range(1, epochs_main + 1):
            model.train()
            running_loss = 0.0; n_seen = 0
            for imgs, ys, _, _ in train_loader:
                imgs = imgs.to(device, non_blocking=True)
                ys = ys.to(device, non_blocking=True)

                optimizer.zero_grad(set_to_none=True)
                with torch.amp.autocast('cuda', enabled=amp_enabled):
                    logits = model(imgs)
                    loss = criterion(logits, ys)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()

                running_loss += float(loss.item()) * imgs.size(0)
                n_seen += imgs.size(0)

            scheduler.step()
            train_loss = running_loss / max(1, n_seen)

            m_val, _ = evaluate(model, val_loader, device, class_names,
                                split_name='val', out_dir=fold_dir)
            f1_macro = m_val['overall']['f1_macro']
            acc_val  = m_val['overall']['accuracy']
            print(f"[Fold {fold_id} | Epoch {epoch:03d}] "
                  f"train_loss={train_loss:.4f} | val_acc={acc_val:.4f} | val_f1_macro={f1_macro:.4f}")

            torch.save({'model': model.state_dict(), 'epoch': epoch, 'class_names': class_names},
                       fold_dir / 'last.ckpt')

            if f1_macro > best_f1:
                best_f1 = f1_macro
                torch.save({'model': model.state_dict(), 'epoch': epoch, 'class_names': class_names},
                           fold_dir / 'best.ckpt')
                print(f"[Fold {fold_id}] New best (F1_macro={best_f1:.4f}) saved.")

        # ----- Phase 2: fine-tuning sólo con REALES -----
        for ft_idx in range(1, epochs_ft + 1):
            epoch = epochs_main + ft_idx
            model.train()
            running_loss = 0.0; n_seen = 0
            for imgs, ys, _, _ in real_train_loader:
                imgs = imgs.to(device, non_blocking=True)
                ys = ys.to(device, non_blocking=True)

                optimizer.zero_grad(set_to_none=True)
                with torch.amp.autocast('cuda', enabled=amp_enabled):
                    logits = model(imgs)
                    loss = criterion(logits, ys)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()

                running_loss += float(loss.item()) * imgs.size(0)
                n_seen += imgs.size(0)

            scheduler.step()
            train_loss = running_loss / max(1, n_seen)

            m_val, _ = evaluate(model, val_loader, device, class_names,
                                split_name='val', out_dir=fold_dir)
            f1_macro = m_val['overall']['f1_macro']
            acc_val  = m_val['overall']['accuracy']
            print(f"[Fold {fold_id} | FT Epoch {epoch:03d}] "
                  f"train_loss={train_loss:.4f} | val_acc={acc_val:.4f} | val_f1_macro={f1_macro:.4f}")

            torch.save({'model': model.state_dict(), 'epoch': epoch, 'class_names': class_names},
                       fold_dir / 'last.ckpt')

            if f1_macro > best_f1:
                best_f1 = f1_macro
                torch.save({'model': model.state_dict(), 'epoch': epoch, 'class_names': class_names},
                           fold_dir / 'best.ckpt')
                print(f"[Fold {fold_id}] New best (F1_macro={best_f1:.4f}) saved (fine-tune).")

        # Re-eval BEST and collect OOF for this fold
        best = torch.load(fold_dir / 'best.ckpt', map_location=device)
        model.load_state_dict(best['model'])

        print(f"[Fold {fold_id}] Re-evaluating BEST on VAL (and collecting OOF)...")
        m_val_best, _ = evaluate(model, val_loader, device, class_names,
                                 split_name='val_best', out_dir=fold_dir)

        # Collect arrays for OOF
        val_inf = infer_loader(model, val_loader, device)  # same order as val_subset (shuffle=False)
        for j, base_i in enumerate(va_idx):
            oof_probs[base_i] = val_inf["probs"][j]
            oof_pred[base_i]  = int(val_inf["preds"][j])
            oof_true[base_i]  = int(val_inf["labels"][j])
            oof_path[base_i]  = val_inf["paths"][j]
            oof_pid[base_i]   = val_inf["pids"][j]

        # Row base with full VAL metrics
        row = {
            "fold": fold_id,
            "val_acc": m_val_best["overall"]["accuracy"],
            "val_precision_macro": m_val_best["overall"]["precision_macro"],
            "val_recall_macro": m_val_best["overall"]["recall_macro"],
            "val_f1_macro": m_val_best["overall"]["f1_macro"],
            "val_precision_weighted": m_val_best["overall"]["precision_weighted"],
            "val_recall_weighted": m_val_best["overall"]["recall_weighted"],
            "val_f1_weighted": m_val_best["overall"]["f1_weighted"],
            "val_n": m_val_best["overall"]["num_samples"],
        }

        # INTERNAL TEST per fold + probs for ensemble
        if inttest_loader is not None:
            print(f"[Fold {fold_id}] Evaluating on INTERNAL TEST...")
            m_int, _ = evaluate(model, inttest_loader, device, class_names,
                                split_name='inttest', out_dir=fold_dir)
            row.update({
                "int_acc": m_int["overall"]["accuracy"],
                "int_precision_macro": m_int["overall"]["precision_macro"],
                "int_recall_macro": m_int["overall"]["recall_macro"],
                "int_f1_macro": m_int["overall"]["f1_macro"],
                "int_precision_weighted": m_int["overall"]["precision_weighted"],
                "int_recall_weighted": m_int["overall"]["recall_weighted"],
                "int_f1_weighted": m_int["overall"]["f1_weighted"],
                "int_n": m_int["overall"]["num_samples"],
            })
            t_int = infer_loader(model, inttest_loader, device)
            inttest_probs_folds.append(t_int["probs"])
            if int_ref_paths is None:
                int_ref_paths  = t_int["paths"]
                int_ref_pids   = t_int["pids"]
                int_ref_labels = t_int["labels"]

        # EXTERNAL TEST per fold + probs for ensemble
        if test_loader is not None:
            print(f"[Fold {fold_id}] Evaluating on EXTERNAL TEST...")
            m_ext, _ = evaluate(model, test_loader, device, class_names,
                                split_name='test', out_dir=fold_dir)
            row.update({
                "test_acc": m_ext["overall"]["accuracy"],
                "test_precision_macro": m_ext["overall"]["precision_macro"],
                "test_recall_macro": m_ext["overall"]["recall_macro"],
                "test_f1_macro": m_ext["overall"]["f1_macro"],
                "test_precision_weighted": m_ext["overall"]["precision_weighted"],
                "test_recall_weighted": m_ext["overall"]["recall_weighted"],
                "test_f1_weighted": m_ext["overall"]["f1_weighted"],
                "test_n": m_ext["overall"]["num_samples"],
            })
            t_inf = infer_loader(model, test_loader, device)
            test_probs_folds.append(t_inf["probs"])
            if test_ref_paths is None:
                test_ref_paths  = t_inf["paths"]
                test_ref_pids   = t_inf["pids"]
                test_ref_labels = t_inf["labels"]

        cv_rows.append(row)

    # === CV summary (mean ± std) ===
    summary_csv = base_run / "cv_summary.csv"
    keys = sorted(set().union(*[r.keys() for r in cv_rows]))
    with open(summary_csv, 'w', newline='', encoding='utf-8') as f:
        w = csv.DictWriter(f, fieldnames=keys)
        w.writeheader()
        for r in cv_rows:
            w.writerow(r)

    def _agg_for(prefix: str) -> Dict[str, Tuple[Optional[float], Optional[float]]]:
        metrics = ["acc","precision_macro","recall_macro","f1_macro",
                   "precision_weighted","recall_weighted","f1_weighted"]
        out: Dict[str, Tuple[Optional[float], Optional[float]]] = {}
        for m in metrics:
            col = f"{prefix}_{m}"
            vals = [float(r[col]) for r in cv_rows if col in r]
            if len(vals) == 0:
                out[m] = (None, None)
            else:
                out[m] = (float(np.mean(vals)), float(np.std(vals)))
        return out

    cv_stats_full = {
        "kfolds": cfg.kfolds,
        "val":     _agg_for("val"),
        "inttest": _agg_for("int") if any(("int_acc" in r) for r in cv_rows) else {},
        "test":    _agg_for("test") if any(("test_acc" in r) for r in cv_rows) else {},
    }
    with open(base_run / "cv_summary_full.json", 'w', encoding='utf-8') as f:
        json.dump(cv_stats_full, f, indent=2)

    with open(base_run / "cv_summary_full.csv", 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(["split","metric","mean","std","k"])
        for split in ["val","inttest","test"]:
            if split not in cv_stats_full or not cv_stats_full[split]:
                continue
            for metric, (mu, sd) in cv_stats_full[split].items():
                if mu is None:
                    continue
                w.writerow([split, metric, mu, sd, cfg.kfolds])

    print("[CV] Summary (mean±std) written to cv_summary_full.{json,csv}")

    # === OOF metrics & artifacts ===
    valid_mask = (~np.isnan(oof_probs).any(axis=1)) & (oof_true != -1)
    oof_probs_v = oof_probs[valid_mask]
    oof_pred_v  = oof_pred[valid_mask]
    oof_true_v  = oof_true[valid_mask]
    oof_path_v  = [oof_path[i] for i, ok in enumerate(valid_mask) if ok]
    oof_pid_v   = [oof_pid[i]  for i, ok in enumerate(valid_mask) if ok]

    if len(oof_true_v) > 0:
        acc = float(accuracy_score(oof_true_v, oof_pred_v))
        prec_m, rec_m, f1_m, _ = precision_recall_fscore_support(
            oof_true_v, oof_pred_v, average='macro', zero_division=0)
        prec_w, rec_w, f1_w, _ = precision_recall_fscore_support(
            oof_true_v, oof_pred_v, average='weighted', zero_division=0)
        per_class = classification_report(
            oof_true_v, oof_pred_v, target_names=class_names,
            output_dict=True, zero_division=0)
        cm = confusion_matrix(oof_true_v, oof_pred_v, labels=list(range(len(class_names))))

        oof_metrics = {
            "overall": {
                "accuracy": acc,
                "precision_macro": prec_m,
                "recall_macro": rec_m,
                "f1_macro": f1_m,
                "precision_weighted": prec_w,
                "recall_weighted": rec_w,
                "f1_weighted": f1_w,
                "num_samples": int(len(oof_true_v)),
            },
            "per_class": per_class,
        }
        with open(base_run / "oof_metrics.json", 'w', encoding='utf-8') as f:
            json.dump(oof_metrics, f, indent=2)

        with open(base_run / "metrics_overall_oof.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["split","accuracy","precision_macro","recall_macro","f1_macro",
                        "precision_weighted","recall_weighted","f1_weighted","num_samples"])
            o = oof_metrics['overall']
            w.writerow(["oof", o['accuracy'], o['precision_macro'], o['recall_macro'],
                        o['f1_macro'], o['precision_weighted'], o['recall_weighted'],
                        o['f1_weighted'], o['num_samples']])

        with open(base_run / "per_class_oof.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["class","precision","recall","f1","support"])
            for cname in class_names:
                row = per_class.get(cname, {"precision":0, "recall":0,
                                            "f1-score":0, "support":0})
                w.writerow([cname, row.get("precision",0), row.get("recall",0),
                            row.get("f1-score",0), row.get("support",0)])

        with open(base_run / "confusion_oof.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["true\\pred"] + class_names)
            for i, cname in enumerate(class_names):
                w.writerow([cname] + list(cm[i]))

        with open(base_run / "oof_predictions.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["path","patient_id","true","pred"] +
                       [f"p_{c}" for c in class_names])
            for i in range(len(oof_true_v)):
                w.writerow([oof_path_v[i], oof_pid_v[i],
                            class_names[oof_true_v[i]],
                            class_names[oof_pred_v[i]]] +
                           list(map(float, oof_probs_v[i].tolist())))

    # === TEST ensemble (avg across folds) ===
    if test_loader_fixed is not None and len(test_probs_folds) > 0:
        probs_stack = np.stack(test_probs_folds, axis=0)     # [K, N_test, C]
        probs_mean  = probs_stack.mean(axis=0)               # [N_test, C]
        preds_ens   = probs_mean.argmax(axis=1)

        acc = float(accuracy_score(test_ref_labels, preds_ens))
        prec_m, rec_m, f1_m, _ = precision_recall_fscore_support(
            test_ref_labels, preds_ens, average='macro', zero_division=0)
        prec_w, rec_w, f1_w, _ = precision_recall_fscore_support(
            test_ref_labels, preds_ens, average='weighted', zero_division=0)
        per_class = classification_report(
            test_ref_labels, preds_ens, target_names=class_names,
            output_dict=True, zero_division=0)
        cm = confusion_matrix(test_ref_labels, preds_ens, labels=list(range(len(class_names))))

        test_ens_metrics = {
            "overall": {
                "accuracy": acc,
                "precision_macro": prec_m,
                "recall_macro": rec_m,
                "f1_macro": f1_m,
                "precision_weighted": prec_w,
                "recall_weighted": rec_w,
                "f1_weighted": f1_w,
                "num_samples": int(len(test_ref_labels)),
            },
            "per_class": per_class,
        }
        with open(base_run / "metrics_test_ensemble.json", 'w', encoding='utf-8') as f:
            json.dump(test_ens_metrics, f, indent=2)

        with open(base_run / "metrics_overall_test_ensemble.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["split","accuracy","precision_macro","recall_macro","f1_macro",
                        "precision_weighted","recall_weighted","f1_weighted","num_samples"])
            o = test_ens_metrics['overall']
            w.writerow(["test_ensemble", o['accuracy'], o['precision_macro'],
                        o['recall_macro'], o['f1_macro'], o['precision_weighted'],
                        o['recall_weighted'], o['f1_weighted'], o['num_samples']])

        with open(base_run / "per_class_test_ensemble.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["class","precision","recall","f1","support"])
            for cname in class_names:
                row = per_class.get(cname, {"precision":0, "recall":0,
                                            "f1-score":0, "support":0})
                w.writerow([cname, row.get("precision",0), row.get("recall",0),
                            row.get("f1-score",0), row.get("support",0)])

        with open(base_run / "confusion_test_ensemble.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["true\\pred"] + class_names)
            for i, cname in enumerate(class_names):
                w.writerow([cname] + list(cm[i]))

        with open(base_run / "test_ensemble_predictions.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["path","patient_id","true","pred"] +
                       [f"p_{c}" for c in class_names])
            for i in range(len(test_ref_labels)):
                w.writerow([test_ref_paths[i], test_ref_pids[i],
                            class_names[int(test_ref_labels[i])],
                            class_names[int(preds_ens[i])]] +
                           list(map(float, probs_mean[i].tolist())))

    # === INTTEST ensemble (avg across folds) ===
    if inttest_loader_fixed is not None and len(inttest_probs_folds) > 0:
        int_probs_stack = np.stack(inttest_probs_folds, axis=0)  # [K, N_int, C]
        int_probs_mean  = int_probs_stack.mean(axis=0)           # [N_int, C]
        int_preds_ens   = int_probs_mean.argmax(axis=1)

        acc = float(accuracy_score(int_ref_labels, int_preds_ens))
        prec_m, rec_m, f1_m, _ = precision_recall_fscore_support(
            int_ref_labels, int_preds_ens, average='macro', zero_division=0)
        prec_w, rec_w, f1_w, _ = precision_recall_fscore_support(
            int_ref_labels, int_preds_ens, average='weighted', zero_division=0)
        per_class = classification_report(
            int_ref_labels, int_preds_ens, target_names=class_names,
            output_dict=True, zero_division=0)
        cm = confusion_matrix(int_ref_labels, int_preds_ens, labels=list(range(len(class_names))))

        int_ens_metrics = {
            "overall": {
                "accuracy": acc,
                "precision_macro": prec_m,
                "recall_macro": rec_m,
                "f1_macro": f1_m,
                "precision_weighted": prec_w,
                "recall_weighted": rec_w,
                "f1_weighted": f1_w,
                "num_samples": int(len(int_ref_labels)),
            },
            "per_class": per_class,
        }
        with open(base_run / "metrics_inttest_ensemble.json", 'w', encoding='utf-8') as f:
            json.dump(int_ens_metrics, f, indent=2)

        with open(base_run / "metrics_overall_inttest_ensemble.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["split","accuracy","precision_macro","recall_macro","f1_macro",
                        "precision_weighted","recall_weighted","f1_weighted","num_samples"])
            o = int_ens_metrics['overall']
            w.writerow(["inttest_ensemble", o['accuracy'], o['precision_macro'],
                        o['recall_macro'], o['f1_macro'], o['precision_weighted'],
                        o['recall_weighted'], o['f1_weighted'], o['num_samples']])

        with open(base_run / "per_class_inttest_ensemble.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["class","precision","recall","f1","support"])
            for cname in class_names:
                row = per_class.get(cname, {"precision":0, "recall":0,
                                            "f1-score":0, "support":0})
                w.writerow([cname, row.get("precision",0), row.get("recall",0),
                            row.get("f1-score",0), row.get("support",0)])

        with open(base_run / "confusion_inttest_ensemble.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["true\\pred"] + class_names)
            for i, cname in enumerate(class_names):
                w.writerow([cname] + list(cm[i]))

        with open(base_run / "inttest_ensemble_predictions.csv", 'w', newline='', encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow(["path","patient_id","true","pred"] +
                       [f"p_{c}" for c in class_names])
            for i in range(len(int_ref_labels)):
                w.writerow([int_ref_paths[i], int_ref_pids[i],
                            class_names[int(int_ref_labels[i])],
                            class_names[int(int_preds_ens[i])]] +
                           list(map(float, int_probs_mean[i].tolist())))

    # (Optional) train on FULL pool and evaluate test once
    if cfg.train_full_after_cv:
        print("\n[Full Train] Training with full pool (no internal test)...")
        run_full = base_run / "full_train"
        run_full.mkdir(parents=True, exist_ok=True)

        real_train_subset = list(base_pool)

        if cfg.balance_gan_to_max and len(synth_items) > 0:
            full_train_subset = build_balanced_train_subset(
                real_train_subset=real_train_subset,
                synth_items=synth_items,
                num_classes=len(class_names),
                seed=cfg.seed + 999,
            )
            print(f"[Full Train] real={len(real_train_subset)} | train_balanced(real+GAN)={len(full_train_subset)}")
        else:
            full_train_subset = real_train_subset + (synth_items if len(synth_items) > 0 else [])
            print(f"[Full Train] real={len(real_train_subset)} + synth={len(synth_items)} -> {len(full_train_subset)}")

        train_ds      = MRIDataset(full_train_subset, image_size=cfg.image_size, aug=False)
        real_train_ds = MRIDataset(real_train_subset, image_size=cfg.image_size, aug=False)

        train_loader      = DataLoader(train_ds,      batch_size=cfg.batch_size, shuffle=True,
                                       num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)
        real_train_loader = DataLoader(real_train_ds, batch_size=cfg.batch_size, shuffle=True,
                                       num_workers=cfg.num_workers, pin_memory=True, collate_fn=collate_fn)

        model = build_model(num_classes=len(class_names), pretrained=cfg.pretrained).to(device)
        optimizer = torch.optim.AdamW(model.parameters(), lr=cfg.lr, weight_decay=cfg.weight_decay)
        scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=cfg.epochs)
        criterion = nn.CrossEntropyLoss()
        amp_enabled = (cfg.amp and device.type == 'cuda')
        scaler = torch.amp.GradScaler('cuda', enabled=amp_enabled)

        epochs_main = cfg.epochs - cfg.fine_tune_real_epochs
        epochs_ft   = cfg.fine_tune_real_epochs

        # Full-train Phase 1: full_train_subset (real + synth balanceado o no)
        for epoch in range(1, epochs_main + 1):
            model.train()
            for imgs, ys, _, _ in train_loader:
                imgs = imgs.to(device, non_blocking=True)
                ys = ys.to(device, non_blocking=True)
                optimizer.zero_grad(set_to_none=True)
                with torch.amp.autocast('cuda', enabled=amp_enabled):
                    loss = criterion(model(imgs), ys)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()
            scheduler.step()

        # Full-train Phase 2: sólo reales
        for ft_idx in range(1, epochs_ft + 1):
            epoch = epochs_main + ft_idx
            model.train()
            for imgs, ys, _, _ in real_train_loader:
                imgs = imgs.to(device, non_blocking=True)
                ys = ys.to(device, non_blocking=True)
                optimizer.zero_grad(set_to_none=True)
                with torch.amp.autocast('cuda', enabled=amp_enabled):
                    loss = criterion(model(imgs), ys)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()
            scheduler.step()

        torch.save({'model': model.state_dict(), 'class_names': class_names},
                   run_full / 'final_full.ckpt')

        if test_loader_fixed is not None:
            print("[Full Train] Evaluating on TEST ...")
            evaluate(model, test_loader_fixed, device, class_names,
                     split_name='test_full', out_dir=run_full)

    print(f"\n[Artifacts] Saved at: {base_run}")
    return base_run

```

```python
# xai_gradcam_resnet.py
# -*- coding: utf-8 -*-
"""
Grad-CAM for ResNet50 models trained with train_kfold(...)
- Loads a single fold checkpoint (best.ckpt)
- Runs Grad-CAM on a folder with structure <root>/<class>/<images>
- Saves overlay images and raw CAMs to disk

Notes:
- This script assumes that the training code lives in a module called
  'train_kfold_api.py' with the following symbols:
    * MRIDataset
    * collate_fn
    * scan_folder
    * build_model

If your file has a different name, just change the import line below.
"""

import os
from pathlib import Path
from typing import List, Optional
from collections import defaultdict

import numpy as np
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
import matplotlib.cm as cm

# Adjust this import if your training file has a different name


# ============================================================
# Device helper
# ============================================================

def get_device() -> torch.device:
    if torch.cuda.is_available():
        return torch.device("cuda")
    else:
        return torch.device("cpu")


# ============================================================
# Grad-CAM implementation with threshold
# ============================================================

class GradCAM:
    """
    Minimal Grad-CAM implementation for a single target layer.
    Allows optional thresholding to shrink the highlighted area.
    """
    def __init__(self, model: nn.Module, target_layer: nn.Module, threshold: float = 0.0):
        """
        threshold: float in [0,1). Values below this (after normalization) are suppressed.
                   e.g., 0.7 -> only top activations remain.
        """
        self.model = model
        self.target_layer = target_layer
        self.threshold = float(threshold)

        self.activations = None
        self.gradients = None
        self.handles = []
        self._register_hooks()

    def _register_hooks(self):
        def forward_hook(module, inputs, output):
            self.activations = output.detach()

        def backward_hook(module, grad_in, grad_out):
            # grad_out[0] has shape [B, C, H, W]
            self.gradients = grad_out[0].detach()

        self.handles.append(self.target_layer.register_forward_hook(forward_hook))
        self.handles.append(self.target_layer.register_backward_hook(backward_hook))

    def remove_hooks(self):
        for h in self.handles:
            h.remove()

    def __call__(self, x: torch.Tensor, target_class: Optional[torch.Tensor] = None) -> torch.Tensor:
        """
        x: [B, 3, H, W] normalized tensor
        target_class: [B] tensor with class indices or None (then argmax is used)

        returns: [B, H_cam, W_cam] CAM normalized to [0,1], optionally thresholded.
        """
        self.model.zero_grad()
        logits = self.model(x)  # [B, C]

        if target_class is None:
            target_class = logits.argmax(dim=1)

        # gather the logit of the target class
        selected = logits.gather(1, target_class.view(-1, 1)).sum()
        selected.backward()

        grads = self.gradients      # [B, C, H, W]
        acts = self.activations     # [B, C, H, W]

        # global average pooling on gradients -> weights [B, C, 1, 1]
        weights = grads.mean(dim=(2, 3), keepdim=True)

        # weighted sum across channels
        cam = (weights * acts).sum(dim=1)  # [B, H, W]
        cam = F.relu(cam)

        # normalize + threshold per image
        cams_norm = []
        eps = 1e-8
        for c in cam:
            c_min = float(c.min())
            c_max = float(c.max())
            if c_max > c_min:
                c_norm = (c - c_min) / (c_max - c_min + eps)
            else:
                c_norm = torch.zeros_like(c)

            if self.threshold > 0.0:
                # suppress values below the threshold
                mask = c_norm >= self.threshold
                if mask.any():
                    c_norm = torch.where(mask, c_norm, torch.zeros_like(c_norm))
                    # re-scale active region back to [0,1]
                    c_norm = c_norm - self.threshold
                    c_norm = c_norm / (float(c_norm.max()) + eps)
                else:
                    c_norm = torch.zeros_like(c_norm)

            cams_norm.append(c_norm)

        cams_norm = torch.stack(cams_norm, dim=0)  # [B, H, W]
        return cams_norm


# ============================================================
# Visualization helpers
# ============================================================

def overlay_cam_on_image(
    image_path: Path,
    cam: np.ndarray,
    out_path: Path,
    alpha: float = 0.5
) -> None:
    """
    image_path: path to original image (will be opened in grayscale)
    cam: [H_cam, W_cam] array in [0,1]
    out_path: where to save overlay
    """
    # load original image in grayscale
    img = Image.open(image_path).convert("L")
    img = img.resize((cam.shape[1], cam.shape[0]), resample=Image.BILINEAR)
    img_np = np.array(img).astype(np.float32)

    # build heatmap using matplotlib colormap
    heatmap = cm.jet(cam)[:, :, :3]  # [H, W, 3] float in [0,1]

    # stack grayscale -> RGB
    img_rgb = np.stack([img_np, img_np, img_np], axis=-1)
    img_rgb = img_rgb / 255.0

    overlay = (1.0 - alpha) * img_rgb + alpha * heatmap
    overlay = np.clip(overlay * 255.0, 0, 255).astype(np.uint8)

    out_path.parent.mkdir(parents=True, exist_ok=True)
    Image.fromarray(overlay).save(out_path)


# ============================================================
# Model + Grad-CAM builder
# ============================================================

def build_gradcam_model(
    ckpt_path: str,
    num_classes: int,
    device: torch.device,
    cam_threshold: float = 0.0
):
    """
    Loads the model from a fold checkpoint and builds Grad-CAM handler.
    Returns:
        model (nn.Module),
        gradcam (GradCAM),
        class_names (List[str])
    """
    ckpt = torch.load(ckpt_path, map_location=device)
    class_names = ckpt.get("class_names", [str(i) for i in range(num_classes)])

    model = build_model(num_classes=num_classes, pretrained=False)
    model.load_state_dict(ckpt["model"])
    model.to(device)
    model.eval()

    # target layer: last conv block of layer4 (ResNet50)
    if isinstance(model.layer4, nn.Sequential):
        target_layer = model.layer4[-1]
    else:
        target_layer = model.layer4

    gradcam = GradCAM(model, target_layer, threshold=cam_threshold)
    return model, gradcam, class_names


# ============================================================
# Subset selection helper
# ============================================================

def select_subset_per_class(
    items: List,
    max_images_per_class: int,
    num_classes: int,
    seed: int = 42
):
    """
    From 'items' (list of (path, class_idx, pid)), select up to 'max_images_per_class' per class.
    """
    rng = np.random.RandomState(seed)
    by_class = defaultdict(list)
    for it in items:
        _, ci, _ = it
        by_class[ci].append(it)

    subset = []
    for ci in range(num_classes):
        lst = by_class[ci]
        if len(lst) == 0:
            continue
        idxs = np.arange(len(lst))
        rng.shuffle(idxs)
        take = idxs[:max_images_per_class]
        subset.extend([lst[i] for i in take])
    return subset


# ============================================================
# Main generation function
# ============================================================

def generate_gradcam_for_folder(
    ckpt_path: str,
    data_root: str,
    out_dir: str,
    image_size: int = 256,
    batch_size: int = 8,
    max_images_per_class: int = 25,
    seed: int = 42,
    cam_threshold: float = 0.0,
    alpha_overlay: float = 0.5
) -> None:
    """
    Main entry point:
    - ckpt_path: path to best.ckpt from some fold
    - data_root: folder with structure <root>/<class>/<images>
    - out_dir: where to save CAM overlays

    cam_threshold:
        - 0.0 -> classic Grad-CAM (full heatmap)
        - 0.5, 0.7, 0.8 -> more focal maps (shrinks highlighted area)

    alpha_overlay:
        - blending factor between original image and heatmap
    """
    device = get_device()
    print(f"[XAI] Using device: {device}")

    data_root = Path(data_root)
    out_dir = Path(out_dir)
    out_dir.mkdir(parents=True, exist_ok=True)

    # Scan folder and build subset
    items, class_names = scan_folder(data_root)
    num_classes = len(class_names)
    print(f"[XAI] Found {len(items)} images in {data_root} | classes={class_names}")

    subset_items = select_subset_per_class(
        items,
        max_images_per_class=max_images_per_class,
        num_classes=num_classes,
        seed=seed
    )
    print(f"[XAI] Using {len(subset_items)} images (balanced subset).")

    ds = MRIDataset(subset_items, image_size=image_size, aug=False)
    loader = DataLoader(
        ds,
        batch_size=batch_size,
        shuffle=False,
        num_workers=2,
        pin_memory=True,
        collate_fn=collate_fn
    )

    # Build model + Grad-CAM
    model, gradcam, class_names_ckpt = build_gradcam_model(
        ckpt_path=ckpt_path,
        num_classes=num_classes,
        device=device,
        cam_threshold=cam_threshold
    )

    print(f"[XAI] Loaded checkpoint from {ckpt_path}")
    print(f"[XAI] Class names in checkpoint: {class_names_ckpt}")
    print(f"[XAI] CAM threshold = {cam_threshold}")

    # Iterate over subset and save overlays
    for imgs, ys, paths, pids in loader:
        imgs = imgs.to(device, non_blocking=True)
        ys = ys.to(device, non_blocking=True)

        # predicted class per image
        with torch.no_grad():
            logits = model(imgs)
            preds = logits.argmax(dim=1)

        # Grad-CAM for predicted class
        cams = gradcam(imgs, target_class=preds)  # [B, H_cam, W_cam]
        cams_up = F.interpolate(
            cams.unsqueeze(1),
            size=(image_size, image_size),
            mode="bilinear",
            align_corners=False
        ).squeeze(1)  # [B, H, W]

        for i in range(imgs.size(0)):
            cam_np = cams_up[i].cpu().numpy()
            true_label = class_names_ckpt[int(ys[i].item())]
            pred_label = class_names_ckpt[int(preds[i].item())]

            in_path = Path(paths[i])

            # Organize outputs by true/pred label
            rel_name = f"{true_label}_true__{pred_label}_pred__{in_path.name}"
            out_path = out_dir / true_label / rel_name

            overlay_cam_on_image(
                image_path=in_path,
                cam=cam_np,
                out_path=out_path,
                alpha=alpha_overlay
            )

    gradcam.remove_hooks()
    print(f"[XAI] Grad-CAM overlays saved to: {out_dir}")

```

```python
# xai_occlusion_resnet.py
# -*- coding: utf-8 -*-
"""
Occlusion sensitivity for ResNet50 models trained with train_kfold(...)

- Loads a single fold checkpoint (best.ckpt)
- Runs occlusion-based importance maps on a folder with structure <root>/<class>/<images>
- Saves overlay images (MRI + occlusion heatmap) to disk

Notes:
- This script assumes that the training code lives in a module called
  'train_kfold_api.py' with the following symbols:
    * MRIDataset
    * collate_fn
    * scan_folder
    * build_model

If your file has a different name, just change the import line below.
"""

import os
from pathlib import Path
from typing import List, Tuple
from collections import defaultdict

import numpy as np
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
import matplotlib.cm as cm

# Adjust this import if your training file has a different name


# ============================================================
# Device helper
# ============================================================

def get_device() -> torch.device:
    if torch.cuda.is_available():
        return torch.device("cuda")
    else:
        return torch.device("cpu")


# ============================================================
# Visualization helper (reuse style from Grad-CAM)
# ============================================================

def overlay_cam_on_image(
    image_path: Path,
    cam: np.ndarray,
    out_path: Path,
    alpha: float = 0.5
) -> None:
    """
    image_path: path to original image (will be opened in grayscale)
    cam: [H_cam, W_cam] array in [0,1]
    out_path: where to save overlay
    alpha: blending factor between original image and heatmap
    """
    # Load original image in grayscale
    img = Image.open(image_path).convert("L")
    img = img.resize((cam.shape[1], cam.shape[0]), resample=Image.BILINEAR)
    img_np = np.array(img).astype(np.float32)

    # Build heatmap using matplotlib colormap
    heatmap = cm.jet(cam)[:, :, :3]  # [H, W, 3] float in [0,1]

    # Stack grayscale -> RGB
    img_rgb = np.stack([img_np, img_np, img_np], axis=-1)
    img_rgb = img_rgb / 255.0

    overlay = (1.0 - alpha) * img_rgb + alpha * heatmap
    overlay = np.clip(overlay * 255.0, 0, 255).astype(np.uint8)

    out_path.parent.mkdir(parents=True, exist_ok=True)
    Image.fromarray(overlay).save(out_path)


# ============================================================
# Subset selection helper (balanced per class)
# ============================================================

def select_subset_per_class(
    items: List[Tuple[Path, int, str]],
    max_images_per_class: int,
    num_classes: int,
    seed: int = 42
) -> List[Tuple[Path, int, str]]:
    """
    From 'items' (list of (path, class_idx, pid)), select up to
    'max_images_per_class' per class.
    """
    rng = np.random.RandomState(seed)
    by_class = defaultdict(list)
    for it in items:
        _, ci, _ = it
        by_class[ci].append(it)

    subset = []
    for ci in range(num_classes):
        lst = by_class[ci]
        if len(lst) == 0:
            continue
        idxs = np.arange(len(lst))
        rng.shuffle(idxs)
        take = idxs[:max_images_per_class]
        subset.extend([lst[i] for i in take])
    return subset


# ============================================================
# Occlusion core
# ============================================================

@torch.no_grad()
def compute_occlusion_map(
    model: nn.Module,
    img: torch.Tensor,
    target_class: int,
    patch_size: int = 32,
    patch_stride: int = 16,
    occlusion_value: float = 0.0,
) -> Tuple[np.ndarray, float]:
    """
    Compute occlusion sensitivity map for a single image.

    Args:
        model: trained classifier (in eval mode)
        img: [1, 3, H, W] normalized image tensor
        target_class: class index to track probability for
        patch_size: side of the square patch to occlude (in pixels)
        patch_stride: stride between occluded patches
        occlusion_value: value used inside the patch in INPUT space
                         (here, input is normalized; 0.0 corresponds to
                         mean intensity after de-normalization).

    Returns:
        occl_up: [H, W] numpy array in [0,1] (higher = more important)
        base_prob: original probability for target_class before occlusion
    """
    model.eval()
    # Base probability
    logits = model(img)  # [1, C]
    probs = F.softmax(logits, dim=1)[0]  # [C]
    base_prob = float(probs[target_class].item())

    _, _, H, W = img.shape

    # Grid size for occlusion positions
    n_rows = 1 + max(0, (H - patch_size) // patch_stride)
    n_cols = 1 + max(0, (W - patch_size) // patch_stride)

    occl_map = np.zeros((n_rows, n_cols), dtype=np.float32)

    for r in range(n_rows):
        for c in range(n_cols):
            top = r * patch_stride
            left = c * patch_stride
            bottom = min(top + patch_size, H)
            right = min(left + patch_size, W)

            img_occ = img.clone()
            img_occ[:, :, top:bottom, left:right] = occlusion_value

            logits_occ = model(img_occ)
            probs_occ = F.softmax(logits_occ, dim=1)[0]
            prob_occ = float(probs_occ[target_class].item())

            diff = base_prob - prob_occ
            if diff < 0:
                diff = 0.0  # clamp negative differences (occlusion helped the class)
            occl_map[r, c] = diff

    # Normalize grid to [0,1]
    max_val = float(occl_map.max())
    if max_val > 1e-8:
        occl_map = occl_map / max_val
    else:
        occl_map[:] = 0.0

    # Upsample grid to full image size [H, W]
    occl_tensor = torch.from_numpy(occl_map).unsqueeze(0).unsqueeze(0).float()  # [1,1,Hc,Wc]
    occl_up = F.interpolate(
        occl_tensor,
        size=(H, W),
        mode="bilinear",
        align_corners=False
    ).squeeze(0).squeeze(0).cpu().numpy()  # [H, W]

    # Final normalization (safety)
    occl_min = float(occl_up.min())
    occl_max = float(occl_up.max())
    if occl_max > occl_min:
        occl_up = (occl_up - occl_min) / (occl_max - occl_min)
    else:
        occl_up = np.zeros_like(occl_up)

    return occl_up, base_prob


# ============================================================
# Model loader
# ============================================================

def build_occlusion_model(
    ckpt_path: str,
    num_classes: int,
    device: torch.device
):
    """
    Loads the model from a fold checkpoint.

    Returns:
        model (nn.Module),
        class_names (List[str])
    """
    ckpt = torch.load(ckpt_path, map_location=device)
    class_names = ckpt.get("class_names", [str(i) for i in range(num_classes)])

    model = build_model(num_classes=num_classes, pretrained=False)
    model.load_state_dict(ckpt["model"])
    model.to(device)
    model.eval()

    return model, class_names


# ============================================================
# Main generation function
# ============================================================

def generate_occlusion_for_folder(
    ckpt_path: str,
    data_root: str,
    out_dir: str,
    image_size: int = 256,
    max_images_per_class: int = 10,
    patch_size: int = 32,
    patch_stride: int = 16,
    target_mode: str = "pred",   # "pred" or "true"
    occlusion_value: float = 0.0,
    alpha_overlay: float = 0.5,
    seed: int = 42,
) -> None:
    """
    Main entry point:
    - ckpt_path: path to best.ckpt from some fold
    - data_root: folder with structure <root>/<class>/<images>
    - out_dir: where to save occlusion overlays

    target_mode:
        - "pred": occlusion importance for the predicted class
        - "true": occlusion importance for the true label

    occlusion_value:
        - value in input (normalized) space inside the patch.
        - 0.0 is a good choice here (maps to gray after de-normalization).
    """
    device = get_device()
    print(f"[OCCL] Using device: {device}")

    data_root = Path(data_root)
    out_dir = Path(out_dir)
    out_dir.mkdir(parents=True, exist_ok=True)

    # Scan folder and build subset
    items, class_names = scan_folder(data_root)
    num_classes = len(class_names)
    print(f"[OCCL] Found {len(items)} images in {data_root} | classes={class_names}")

    subset_items = select_subset_per_class(
        items=items,
        max_images_per_class=max_images_per_class,
        num_classes=num_classes,
        seed=seed
    )
    print(f"[OCCL] Using {len(subset_items)} images (balanced subset).")

    ds = MRIDataset(subset_items, image_size=image_size, aug=False)
    loader = DataLoader(
        ds,
        batch_size=1,  # occlusion is expensive, do 1-by-1
        shuffle=False,
        num_workers=2,
        pin_memory=True,
        collate_fn=collate_fn
    )

    # Build model
    model, class_names_ckpt = build_occlusion_model(
        ckpt_path=ckpt_path,
        num_classes=num_classes,
        device=device
    )

    print(f"[OCCL] Loaded checkpoint from {ckpt_path}")
    print(f"[OCCL] Class names in checkpoint: {class_names_ckpt}")
    print(f"[OCCL] target_mode = {target_mode}, patch_size={patch_size}, stride={patch_stride}")

    for imgs, ys, paths, pids in loader:
        imgs = imgs.to(device, non_blocking=True)     # [1,3,H,W]
        ys = ys.to(device, non_blocking=True)         # [1]

        with torch.no_grad():
            logits = model(imgs)
            probs = F.softmax(logits, dim=1)
            preds = probs.argmax(dim=1)               # [1]

        if target_mode == "true":
            target_class = int(ys[0].item())
        elif target_mode == "pred":
            target_class = int(preds[0].item())
        else:
            raise ValueError(f"Invalid target_mode: {target_mode}")

        occl_map, base_prob = compute_occlusion_map(
            model=model,
            img=imgs,
            target_class=target_class,
            patch_size=patch_size,
            patch_stride=patch_stride,
            occlusion_value=occlusion_value,
        )

        # Build output file name
        true_label = class_names_ckpt[int(ys[0].item())]
        pred_label = class_names_ckpt[int(preds[0].item())]
        in_path = Path(paths[0])

        rel_name = (
            f"{true_label}_true__{pred_label}_pred__"
            f"cls{target_class}_base{base_prob:.3f}__{in_path.name}"
        )
        out_path = out_dir / true_label / rel_name

        overlay_cam_on_image(
            image_path=in_path,
            cam=occl_map,
            out_path=out_path,
            alpha=alpha_overlay
        )

    print(f"[OCCL] Occlusion overlays saved to: {out_dir}")
```

```python
# =========================
# Jupyter single-cell script:
#  - Read training.zip + testing.zip (images)
#  - Read masks.zip (all masks)
#  - Recreate split and write BOTH images and masks to out_dataset/
# =========================

import zipfile
import os
from pathlib import Path
from collections import defaultdict

# ====== CONFIG (EDIT THESE FILENAMES IF NEEDED) ======
ZIP_TRAINING = "/content/drive/MyDrive/TESIS/splits/by_patient_30test1/train.zip"
ZIP_TESTING  = "/content/drive/MyDrive/TESIS/splits/by_patient_30test1/test.zip"
ZIP_MASKS    = "/content/drive/MyDrive/TESIS/datasetcleanmask.zip"

OUT_DIR = Path("/content/localdata/new1")
IMG_ROOT  = OUT_DIR / "images"
MASK_ROOT = OUT_DIR / "masks"

CLASSES = {"glioma", "meningioma", "pituitary"}
MASK_SUFFIX = "_mask.png"
BORDER_SUFFIX = "_border.png"
IMG_EXTS = (".png", ".jpg", ".jpeg", ".bmp", ".tif", ".tiff", ".npy")

# ====== HELPERS ======
def norm_path(p: str) -> str:
    return p.replace("\\", "/")

def is_image_file(name: str) -> bool:
    return name.lower().endswith(IMG_EXTS)

def is_slice_filename(name: str) -> bool:
    n = name.lower()
    return is_image_file(n) and (not n.endswith(MASK_SUFFIX)) and (not n.endswith(BORDER_SUFFIX))

def is_mask_filename(name: str) -> bool:
    return name.lower().endswith(MASK_SUFFIX)

def get_class_from_path_or_name(relpath: str) -> str | None:
    """
    Try to infer class from:
      1) any folder segment equals one of CLASSES
      2) filename pattern: <PID>_<tumor>_<idx>.(ext)  -> tumor is second token
    """
    rel = norm_path(relpath)
    parts = [p.lower() for p in rel.split("/") if p.strip()]
    for seg in parts:
        if seg in CLASSES:
            return seg

    base = os.path.basename(rel).lower()
    stem = Path(base).stem  # e.g. "123_glioma_0007" or "123_glioma_0007_mask"
    stem = stem.replace("_mask", "")
    toks = stem.split("_")
    if len(toks) >= 3 and toks[1] in CLASSES:
        return toks[1]
    return None

def make_key_from_slice_filename(filename: str) -> str:
    # <PID>_<tumor>_<idx>.png -> key = <PID>_<tumor>_<idx>
    return Path(filename).stem

def make_key_from_mask_filename(filename: str) -> str:
    # <PID>_<tumor>_<idx>_mask.png -> key = <PID>_<tumor>_<idx>
    return Path(filename).stem.replace("_mask", "")

def copy_zip_member(z: zipfile.ZipFile, info: zipfile.ZipInfo, dest_path: Path):
    dest_path.parent.mkdir(parents=True, exist_ok=True)
    with z.open(info, "r") as src, dest_path.open("wb") as dst:
        dst.write(src.read())

def list_sample_paths(zip_path: str, n: int = 15):
    with zipfile.ZipFile(zip_path, "r") as z:
        files = [i.filename for i in z.infolist() if not i.is_dir()][:n]
    return files

# ====== PREP OUTPUT DIRS ======
OUT_DIR.mkdir(parents=True, exist_ok=True)
IMG_ROOT.mkdir(parents=True, exist_ok=True)
MASK_ROOT.mkdir(parents=True, exist_ok=True)

print("=== ZIP samples (first 15 entries) ===")
print("[training.zip]")
for s in list_sample_paths(ZIP_TRAINING, 15): print("  ", s)
print("[testing.zip]")
for s in list_sample_paths(ZIP_TESTING, 15): print("  ", s)
print("[masks.zip]")
for s in list_sample_paths(ZIP_MASKS, 15): print("  ", s)

# ====== 1) BUILD MAP key -> (split, class) FROM TRAIN/TEST IMAGES ======
key_to_bucket = {}
collisions = defaultdict(list)
unclassified_examples = []

def scan_images_build_map(zip_path: str, split_name: str):
    global key_to_bucket, collisions, unclassified_examples
    seen = 0
    mapped = 0

    with zipfile.ZipFile(zip_path, "r") as z:
        for info in z.infolist():
            if info.is_dir():
                continue
            rel = norm_path(info.filename)
            base = os.path.basename(rel)

            if not is_slice_filename(base):
                continue

            seen += 1
            cls = get_class_from_path_or_name(rel)
            if cls is None:
                if len(unclassified_examples) < 10:
                    unclassified_examples.append((split_name, rel))
                continue

            key = make_key_from_slice_filename(base)
            bucket = (split_name, cls)

            if key in key_to_bucket and key_to_bucket[key] != bucket:
                collisions[key].append((key_to_bucket[key], bucket, rel))
            else:
                key_to_bucket[key] = bucket
                mapped += 1

    return seen, mapped

seen_tr, mapped_tr = scan_images_build_map(ZIP_TRAINING, "training")
seen_te, mapped_te = scan_images_build_map(ZIP_TESTING, "testing")

print("\n=== Mapping from images ===")
print(f"training.zip: slices found={seen_tr}, mapped={mapped_tr}")
print(f"testing.zip : slices found={seen_te}, mapped={mapped_te}")
print(f"Unique keys mapped: {len(key_to_bucket)}")
if collisions:
    ex = next(iter(collisions.items()))
    print(f"[WARN] Collisions: {len(collisions)} (example key: {ex[0]} -> {ex[1][0]})")
    print("       Collided keys will be skipped for safety.")
if unclassified_examples:
    print("[WARN] Could not infer class for some images. Examples:")
    for sp, p in unclassified_examples:
        print(f"  - ({sp}) {p}")

if len(key_to_bucket) == 0:
    raise RuntimeError(
        "No images were mapped. This usually means class cannot be inferred from path or filename.\n"
        "Expected filename like <PID>_<tumor>_<idx>.png (tumor in {glioma, meningioma, pituitary}).\n"
        "Check the printed ZIP samples above."
    )

# ====== 2) COPY IMAGES TO OUT_DIR/images/<split>/<class>/ ======
def copy_images(zip_path: str, split_name: str):
    copied = 0
    skipped_collision = 0
    skipped_unmapped = 0

    with zipfile.ZipFile(zip_path, "r") as z:
        for info in z.infolist():
            if info.is_dir():
                continue
            rel = norm_path(info.filename)
            base = os.path.basename(rel)
            if not is_slice_filename(base):
                continue

            key = make_key_from_slice_filename(base)
            if key in collisions:
                skipped_collision += 1
                continue

            bucket = key_to_bucket.get(key)
            if bucket is None:
                skipped_unmapped += 1
                continue

            split, cls = bucket
            if split != split_name:
                # safety: if something weird is mapped to the other split, still follow the mapping
                split_name = split

            dst = IMG_ROOT / split / cls / base
            copy_zip_member(z, info, dst)
            copied += 1

    return copied, skipped_unmapped, skipped_collision

img_tr_c, img_tr_skip_u, img_tr_skip_col = copy_images(ZIP_TRAINING, "training")
img_te_c, img_te_skip_u, img_te_skip_col = copy_images(ZIP_TESTING, "testing")

print("\n=== Copy images ===")
print(f"Training images copied: {img_tr_c} | skipped_unmapped={img_tr_skip_u} | skipped_collision={img_tr_skip_col}")
print(f"Testing  images copied: {img_te_c} | skipped_unmapped={img_te_skip_u} | skipped_collision={img_te_skip_col}")

# ====== 3) COPY MASKS FROM masks.zip TO OUT_DIR/masks/<split>/<class>/ USING key_to_bucket ======
mask_copied = 0
mask_skipped_no_match = 0
mask_skipped_collision = 0
mask_skipped_not_mask = 0

with zipfile.ZipFile(ZIP_MASKS, "r") as z:
    for info in z.infolist():
        if info.is_dir():
            continue
        rel = norm_path(info.filename)
        base = os.path.basename(rel)

        if not is_mask_filename(base):
            mask_skipped_not_mask += 1
            continue

        key = make_key_from_mask_filename(base)
        if key in collisions:
            mask_skipped_collision += 1
            continue

        bucket = key_to_bucket.get(key)
        if bucket is None:
            mask_skipped_no_match += 1
            continue

        split, cls = bucket
        dst = MASK_ROOT / split / cls / base
        copy_zip_member(z, info, dst)
        mask_copied += 1

print("\n=== Copy masks ===")
print(f"Masks copied               : {mask_copied}")
print(f"Skipped (not mask file)    : {mask_skipped_not_mask}")
print(f"Skipped (no image match)   : {mask_skipped_no_match}")
print(f"Skipped (collision keys)   : {mask_skipped_collision}")

# ====== 4) SANITY COUNTS ======
print("\n=== Final counts per bucket ===")
for split in ["training", "testing"]:
    for cls in sorted(CLASSES):
        img_dir = IMG_ROOT / split / cls
        msk_dir = MASK_ROOT / split / cls
        ni = len(list(img_dir.glob("*"))) if img_dir.exists() else 0
        nm = len(list(msk_dir.glob("*"))) if msk_dir.exists() else 0
        print(f"{split}/{cls}: images={ni}, masks={nm}")

print(f"\n✅ Done. Output written to: {OUT_DIR.resolve()}")
```

```python
# =========================
# Jupyter: Grad-CAM (pred + true) + IoU/Dice/Pointing Game vs masks
# Uses your xai_gradcam_resnet.py logic (GradCAM class + build_model + MRIDataset pipeline)
# =========================

import os
from pathlib import Path
from typing import Optional, List, Dict
from collections import defaultdict

import numpy as np
import pandas as pd
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader

# ------------------------------------------------------------
# IMPORTANT: you must have these imports available in your environment
# from train_kfold_api import MRIDataset, collate_fn, scan_folder, build_model
# If your module name differs, change it accordingly.
# ------------------------------------------------------------

# =========================
# Grad-CAM (same as your script, but safe backward hook)
# =========================
class GradCAM:
    """
    Minimal Grad-CAM for a single target layer.
    Optionally thresholded (after normalization) to shrink highlighted area.
    """
    def __init__(self, model: nn.Module, target_layer: nn.Module, threshold: float = 0.0):
        self.model = model
        self.target_layer = target_layer
        self.threshold = float(threshold)

        self.activations = None
        self.gradients = None
        self.handles = []
        self._register_hooks()

    def _register_hooks(self):
        def forward_hook(module, inputs, output):
            self.activations = output.detach()

        # Prefer full backward hook in new PyTorch; fallback if not available.
        def backward_hook(module, grad_in, grad_out):
            self.gradients = grad_out[0].detach()

        self.handles.append(self.target_layer.register_forward_hook(forward_hook))
        if hasattr(self.target_layer, "register_full_backward_hook"):
            self.handles.append(self.target_layer.register_full_backward_hook(backward_hook))
        else:
            self.handles.append(self.target_layer.register_backward_hook(backward_hook))

    def remove_hooks(self):
        for h in self.handles:
            h.remove()

    def __call__(self, x: torch.Tensor, target_class: Optional[torch.Tensor] = None) -> torch.Tensor:
        """
        x: [B, 3, H, W] normalized tensor
        target_class: [B] tensor with class indices or None -> argmax
        returns: [B, H_cam, W_cam] CAM normalized to [0,1], optionally thresholded.
        """
        self.model.zero_grad(set_to_none=True)
        logits = self.model(x)  # [B, C]

        if target_class is None:
            target_class = logits.argmax(dim=1)

        selected = logits.gather(1, target_class.view(-1, 1)).sum()
        selected.backward()

        grads = self.gradients      # [B, C, H, W]
        acts  = self.activations    # [B, C, H, W]
        if grads is None or acts is None:
            raise RuntimeError("GradCAM hooks did not capture gradients/activations. Check target_layer.")

        weights = grads.mean(dim=(2, 3), keepdim=True)  # [B, C, 1, 1]
        cam = (weights * acts).sum(dim=1)               # [B, H, W]
        cam = F.relu(cam)

        # normalize + threshold per image
        eps = 1e-8
        cams_norm = []
        for c in cam:
            c_min = float(c.min())
            c_max = float(c.max())
            if c_max > c_min:
                c_norm = (c - c_min) / (c_max - c_min + eps)
            else:
                c_norm = torch.zeros_like(c)

            if self.threshold > 0.0:
                mask = c_norm >= self.threshold
                if mask.any():
                    c_norm = torch.where(mask, c_norm, torch.zeros_like(c_norm))
                    c_norm = c_norm - self.threshold
                    c_norm = c_norm / (float(c_norm.max()) + eps)
                else:
                    c_norm = torch.zeros_like(c_norm)

            cams_norm.append(c_norm)

        return torch.stack(cams_norm, dim=0)  # [B, H, W]


# =========================
# Helpers
# =========================
def get_device() -> torch.device:
    return torch.device("cuda" if torch.cuda.is_available() else "cpu")

def load_mask_binary(mask_path: Path, image_size: int) -> np.ndarray:
    m = Image.open(mask_path).convert("L")
    m = m.resize((image_size, image_size), resample=Image.NEAREST)
    m = (np.array(m).astype(np.float32) / 255.0) >= 0.5
    return m.astype(np.uint8)

def cam_to_topk_binary(cam: np.ndarray, topk_percent: float) -> np.ndarray:
    flat = cam.flatten()
    k = max(1, int(len(flat) * topk_percent))
    thr = np.partition(flat, -k)[-k]
    return (cam >= thr).astype(np.uint8)

def iou_dice(pred_bin: np.ndarray, gt_bin: np.ndarray):
    pred = pred_bin.astype(bool)
    gt = gt_bin.astype(bool)
    inter = np.logical_and(pred, gt).sum()
    union = np.logical_or(pred, gt).sum()
    iou = inter / (union + 1e-8)
    dice = (2.0 * inter) / (pred.sum() + gt.sum() + 1e-8)
    return float(iou), float(dice)

def pointing_game(cam: np.ndarray, gt_bin: np.ndarray) -> int:
    y, x = np.unravel_index(np.argmax(cam), cam.shape)
    return int(gt_bin[y, x] == 1)

def find_mask_for_image(image_path: Path, masks_root: Path, true_label_folder: str) -> Path:
    """
    Expects masks organized as:
      masks_root/<class>/<PID>_<tumor>_<idx>_mask.png
    and image_path is:
      data_root/<class>/<PID>_<tumor>_<idx>.png
    """
    mask_name = f"{image_path.stem}_mask.png"
    return masks_root / true_label_folder / mask_name


# =========================
# Main: compute metrics for BOTH CAM_pred and CAM_true
# =========================
def gradcam_metrics_for_folder(
    ckpt_path: str,
    data_root: str,          # <root>/<class>/<images>
    masks_root: str,         # <root>/<class>/<masks>  (same class folders)
    out_csv: str,
    image_size: int = 256,
    batch_size: int = 8,
    max_images_per_class: int = 10_000,  # set high to evaluate all
    seed: int = 42,
    cam_threshold: float = 0.0,          # keep 0.0 for "classic" Grad-CAM; you can still use TOPK for binarization
    topk_percent: float = 0.15
) -> pd.DataFrame:
    device = get_device()
    print(f"[XAI] Using device: {device}")

    data_root = Path(data_root)
    masks_root = Path(masks_root)
    out_csv = Path(out_csv)
    out_csv.parent.mkdir(parents=True, exist_ok=True)

    # Scan folder (your existing helper)
    items, class_names = scan_folder(data_root)
    num_classes = len(class_names)
    print(f"[XAI] Found {len(items)} images in {data_root} | classes={class_names}")

    # Option: use ALL items (no subset) or balanced subset
    if max_images_per_class is not None and max_images_per_class < 10_000:
        # reuse your helper if you want balanced subset
        def select_subset_per_class(items, max_images_per_class, num_classes, seed=42):
            rng = np.random.RandomState(seed)
            by_class = defaultdict(list)
            for it in items:
                _, ci, _ = it
                by_class[ci].append(it)
            subset = []
            for ci in range(num_classes):
                lst = by_class[ci]
                if not lst:
                    continue
                idxs = np.arange(len(lst))
                rng.shuffle(idxs)
                take = idxs[:max_images_per_class]
                subset.extend([lst[i] for i in take])
            return subset
        use_items = select_subset_per_class(items, max_images_per_class, num_classes, seed=seed)
        print(f"[XAI] Using {len(use_items)} images (balanced subset).")
    else:
        use_items = items
        print(f"[XAI] Using ALL images: {len(use_items)}")

    ds = MRIDataset(use_items, image_size=image_size, aug=False)
    loader = DataLoader(
        ds,
        batch_size=batch_size,
        shuffle=False,
        num_workers=2,
        pin_memory=True,
        collate_fn=collate_fn
    )

    # Load model from ckpt (your format)
    ckpt = torch.load(ckpt_path, map_location=device)
    class_names_ckpt = ckpt.get("class_names", [str(i) for i in range(num_classes)])

    model = build_model(num_classes=num_classes, pretrained=False)
    model.load_state_dict(ckpt["model"])
    model.to(device).eval()

    # target layer: last conv block of layer4 (ResNet50) as in your script
    target_layer = model.layer4[-1] if isinstance(model.layer4, nn.Sequential) else model.layer4
    gradcam = GradCAM(model, target_layer, threshold=cam_threshold)

    rows: List[Dict] = []
    missing_masks = 0

    for imgs, ys, paths, pids in loader:
        imgs = imgs.to(device, non_blocking=True)
        ys = ys.to(device, non_blocking=True)

        with torch.no_grad():
            logits = model(imgs)
            preds = logits.argmax(dim=1)

        # Compute BOTH cams
        cams_pred = gradcam(imgs, target_class=preds)  # [B, h, w]
        cams_true = gradcam(imgs, target_class=ys)     # [B, h, w]

        # Upsample to image resolution
        cams_pred_up = F.interpolate(cams_pred.unsqueeze(1), size=(image_size, image_size),
                                     mode="bilinear", align_corners=False).squeeze(1)
        cams_true_up = F.interpolate(cams_true.unsqueeze(1), size=(image_size, image_size),
                                     mode="bilinear", align_corners=False).squeeze(1)

        for i in range(imgs.size(0)):
            in_path = Path(paths[i])

            true_idx = int(ys[i].item())
            pred_idx = int(preds[i].item())
            true_label = class_names_ckpt[true_idx]
            pred_label = class_names_ckpt[pred_idx]

            cam_p = cams_pred_up[i].detach().cpu().numpy().astype(np.float32)
            cam_t = cams_true_up[i].detach().cpu().numpy().astype(np.float32)

            mask_path = find_mask_for_image(in_path, masks_root, true_label_folder=true_label)
            if not mask_path.exists():
                missing_masks += 1
                continue

            gt = load_mask_binary(mask_path, image_size=image_size)

            # Binarize CAM by top-k
            bin_p = cam_to_topk_binary(cam_p, topk_percent=topk_percent)
            bin_t = cam_to_topk_binary(cam_t, topk_percent=topk_percent)

            iou_p, dice_p = iou_dice(bin_p, gt)
            iou_t, dice_t = iou_dice(bin_t, gt)

            pg_p = pointing_game(cam_p, gt)
            pg_t = pointing_game(cam_t, gt)

            rows.append({
                "image_path": str(in_path),
                "mask_path": str(mask_path),
                "pid": str(pids[i]),
                "true_idx": true_idx,
                "pred_idx": pred_idx,
                "true_label": true_label,
                "pred_label": pred_label,
                "topk_percent": float(topk_percent),
                "cam_threshold": float(cam_threshold),

                "iou_pred": iou_p,
                "dice_pred": dice_p,
                "pointing_pred": pg_p,

                "iou_true": iou_t,
                "dice_true": dice_t,
                "pointing_true": pg_t,
            })

    gradcam.remove_hooks()

    df = pd.DataFrame(rows)
    df.to_csv(out_csv, index=False)

    print(f"[XAI] Missing masks: {missing_masks}")
    print(f"[XAI] Saved per-image metrics to: {out_csv}")
    print(f"[XAI] Evaluated pairs: {len(df)}")

    # Summary table
    if len(df) > 0:
        summary = (df.groupby(["true_label"])
                     .agg(
                         n=("image_path", "count"),
                         iou_pred_mean=("iou_pred", "mean"),
                         dice_pred_mean=("dice_pred", "mean"),
                         pointing_pred_acc=("pointing_pred", "mean"),
                         iou_true_mean=("iou_true", "mean"),
                         dice_true_mean=("dice_true", "mean"),
                         pointing_true_acc=("pointing_true", "mean"),
                     )
                     .reset_index())
        display(summary)

    return df


# =========================
# Example call (EDIT PATHS)
# =========================
# If your dataset is:
# out_dataset/images/testing/<class>/*.png
# out_dataset/masks/testing/<class>/*_mask.png
# then set:
#   data_root  = "out_dataset/images/testing"
#   masks_root = "out_dataset/masks/testing"

# df = gradcam_metrics_for_folder(
#     ckpt_path="path/to/best.ckpt",
#     data_root="out_dataset/images/testing",
#     masks_root="out_dataset/masks/testing",
#     out_csv="out_dataset/xai_gradcam_iou_dice_pointing_testing.csv",
#     image_size=256,
#     batch_size=8,
#     max_images_per_class=100000,  # all
#     cam_threshold=0.0,
#     topk_percent=0.15
# )
```

```python
# =========================
# FULL SCRIPT (Notebook cell)
# + Global metrics (micro + macro) for TESTING and TRAINING
# =========================

import numpy as np
import pandas as pd
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F

# -------------------------
# Config
# -------------------------
OUT_DIR = Path("/content/localdata/new1")
CKPT_FIXED = "/content/drive/MyDrive/TESIS/StyleGan2/CNNResnet50kfoldinge1/inge1.3fold/full_train/final_full.ckpt"

TESTING_CSV  = OUT_DIR / "xai_gradcam_metrics_testing.csv"
TRAINING_CSV = OUT_DIR / "xai_gradcam_metrics_training.csv"

RANK_BY = "dice_true"      # "iou_true" / "dice_true" / "pointing_true"
IMAGE_SIZE = 256

cam_threshold = 0.5        # try 0.5, 0.7, 0.8
alpha_overlay = 0.5        # Grad-CAM overlay alpha
mask_alpha = 0.35          # Mask overlay alpha

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("DEVICE:", DEVICE)

# -------------------------
# Helpers
# -------------------------
def load_grayscale_np(img_path: Path, size: int = 256) -> np.ndarray:
    img = Image.open(img_path).convert("L").resize((size, size), Image.BILINEAR)
    return (np.array(img).astype(np.float32) / 255.0)

def load_mask_np(mask_path: Path, size: int = 256) -> np.ndarray:
    m = Image.open(mask_path).convert("L").resize((size, size), Image.NEAREST)
    return ((np.array(m).astype(np.float32) / 255.0) >= 0.5).astype(np.uint8)

def img_to_tensor_3ch(img_gray: np.ndarray) -> torch.Tensor:
    x = torch.from_numpy(img_gray)[None, None, :, :].float()
    return x.repeat(1, 3, 1, 1)

def overlay_cam(img_gray: np.ndarray, cam: np.ndarray, alpha: float = 0.45) -> np.ndarray:
    img_rgb = np.stack([img_gray, img_gray, img_gray], axis=-1)
    heat = plt.cm.jet(cam)[..., :3]
    out = (1 - alpha) * img_rgb + alpha * heat
    return np.clip(out, 0, 1)

def overlay_mask_on_image(img_gray: np.ndarray, mask: np.ndarray, alpha: float = 0.35) -> np.ndarray:
    img_rgb = np.stack([img_gray, img_gray, img_gray], axis=-1)
    out = img_rgb.copy()
    m = mask.astype(bool)

    # red overlay
    out[m, 0] = np.clip((1 - alpha) * out[m, 0] + alpha * 1.0, 0, 1)
    out[m, 1] = np.clip((1 - alpha) * out[m, 1] + alpha * 0.0, 0, 1)
    out[m, 2] = np.clip((1 - alpha) * out[m, 2] + alpha * 0.0, 0, 1)
    return out

# -------------------------
# Summaries (by class + global)
# -------------------------
METRIC_COLS = [
    "iou_true","dice_true","pointing_true",
    "iou_pred","dice_pred","pointing_pred"
]

def summarize_scores_by_class(csv_path: Path, split_name: str):
    df = pd.read_csv(csv_path)
    needed = ["true_label"] + METRIC_COLS
    for c in needed:
        if c not in df.columns:
            raise ValueError(f"{csv_path} missing column '{c}'.")

    summary = (df.groupby("true_label")
                 .agg(
                     n=("true_label","count"),
                     iou_true_mean=("iou_true","mean"),
                     iou_true_std=("iou_true","std"),
                     dice_true_mean=("dice_true","mean"),
                     dice_true_std=("dice_true","std"),
                     pg_true_mean=("pointing_true","mean"),
                     pg_true_std=("pointing_true","std"),

                     iou_pred_mean=("iou_pred","mean"),
                     iou_pred_std=("iou_pred","std"),
                     dice_pred_mean=("dice_pred","mean"),
                     dice_pred_std=("dice_pred","std"),
                     pg_pred_mean=("pointing_pred","mean"),
                     pg_pred_std=("pointing_pred","std"),
                 )
                 .reset_index()
                 .sort_values("true_label"))

    print(f"\n=== {split_name.upper()} | Scores by class (mean ± std) ===")
    display(summary)
    return df, summary

def compute_global_metrics(df: pd.DataFrame, split_name: str) -> pd.DataFrame:
    """
    Returns a table with:
      - MICRO: mean±std over all samples (weights by n implicitly)
      - MACRO: average of per-class means (equal weight per class)
    """
    # MICRO (all samples)
    micro = {"split": split_name, "avg_type": "micro", "n": int(len(df))}
    for m in METRIC_COLS:
        micro[f"{m}_mean"] = float(df[m].mean())
        micro[f"{m}_std"]  = float(df[m].std())

    # MACRO (average of class means)
    per_class_means = df.groupby("true_label")[METRIC_COLS].mean()
    macro = {"split": split_name, "avg_type": "macro", "n": int(per_class_means.shape[0])}
    for m in METRIC_COLS:
        macro[f"{m}_mean"] = float(per_class_means[m].mean())
        macro[f"{m}_std"]  = float(per_class_means[m].std())  # variability across classes

    out = pd.DataFrame([micro, macro])

    # Pretty print (mean ± std)
    pretty_rows = []
    for _, r in out.iterrows():
        row = {
            "split": r["split"],
            "avg_type": r["avg_type"],
            "n": int(r["n"]),
            "IoU_true":  f'{r["iou_true_mean"]:.3f} ± {r["iou_true_std"]:.3f}',
            "Dice_true": f'{r["dice_true_mean"]:.3f} ± {r["dice_true_std"]:.3f}',
            "PG_true":   f'{r["pointing_true_mean"]:.3f} ± {r["pointing_true_std"]:.3f}',
            "IoU_pred":  f'{r["iou_pred_mean"]:.3f} ± {r["iou_pred_std"]:.3f}',
            "Dice_pred": f'{r["dice_pred_mean"]:.3f} ± {r["dice_pred_std"]:.3f}',
            "PG_pred":   f'{r["pointing_pred_mean"]:.3f} ± {r["pointing_pred_std"]:.3f}',
        }
        pretty_rows.append(row)

    pretty = pd.DataFrame(pretty_rows)
    print(f"\n=== {split_name.upper()} | Global metrics (micro + macro) ===")
    display(pretty)
    return out, pretty

# -------------------------
# Best-per-class selection + plotting
# -------------------------
def best_rows_per_class(df: pd.DataFrame, rank_by: str):
    best_rows = []
    for cls in sorted(df["true_label"].unique()):
        sub = df[df["true_label"] == cls].sort_values(rank_by, ascending=False)
        if len(sub) > 0:
            best_rows.append(sub.iloc[0])
    return pd.DataFrame(best_rows).reset_index(drop=True)

def load_model_from_ckpt(ckpt_path: str, df_for_fallback: pd.DataFrame):
    ckpt = torch.load(ckpt_path, map_location=DEVICE)
    class_names = ckpt.get("class_names", None)
    if class_names is None:
        class_names = sorted(df_for_fallback["true_label"].unique())
    num_classes = len(class_names)

    # build_model must exist in your notebook
    model = build_model(num_classes=num_classes, pretrained=False)

    state = ckpt["model"] if isinstance(ckpt, dict) and "model" in ckpt else ckpt
    model.load_state_dict(state)
    model.to(DEVICE).eval()

    target_layer = model.layer4[-1] if isinstance(model.layer4, nn.Sequential) else model.layer4
    return model, target_layer, class_names

def plot_best_per_class(df_best: pd.DataFrame, model: nn.Module, target_layer: nn.Module,
                        split_name: str, image_size: int, cam_threshold: float,
                        alpha_overlay: float, mask_alpha: float, rank_by: str):

    gradcam = GradCAM(model, target_layer, threshold=float(cam_threshold))

    print(f"\n--- {split_name.upper()} | Best per class by '{rank_by}' ---")
    display(df_best[["true_label","pred_label","iou_true","dice_true","pointing_true",
                     "iou_pred","dice_pred","pointing_pred","image_path","mask_path"]])

    for _, row in df_best.iterrows():
        img_path = Path(row["image_path"])
        mask_path = Path(row["mask_path"])

        if not img_path.exists():
            print(f"[WARN] Image not found: {img_path}")
            continue
        if not mask_path.exists():
            print(f"[WARN] Mask not found: {mask_path}")
            continue

        img_gray = load_grayscale_np(img_path, size=image_size)
        mask = load_mask_np(mask_path, size=image_size)

        x = img_to_tensor_3ch(img_gray).to(DEVICE)
        true_idx = int(row["true_idx"])
        pred_idx = int(row["pred_idx"])

        cam_true = gradcam(x, target_class=torch.tensor([true_idx], device=DEVICE))
        cam_pred = gradcam(x, target_class=torch.tensor([pred_idx], device=DEVICE))

        cam_true = F.interpolate(cam_true.unsqueeze(1), size=(image_size, image_size),
                                 mode="bilinear", align_corners=False)[0,0].detach().cpu().numpy()
        cam_pred = F.interpolate(cam_pred.unsqueeze(1), size=(image_size, image_size),
                                 mode="bilinear", align_corners=False)[0,0].detach().cpu().numpy()

        mask_overlay = overlay_mask_on_image(img_gray, mask, alpha=float(mask_alpha))
        ov_true = overlay_cam(img_gray, cam_true, alpha=float(alpha_overlay))
        ov_pred = overlay_cam(img_gray, cam_pred, alpha=float(alpha_overlay))

        fig, axes = plt.subplots(1, 4, figsize=(16, 4))
        fig.suptitle(
            f"{split_name.upper()} | class={row['true_label']} pred={row['pred_label']} | "
            f"IoU_true={row['iou_true']:.3f} Dice_true={row['dice_true']:.3f} PG_true={row['pointing_true']:.3f} | "
            f"IoU_pred={row['iou_pred']:.3f} Dice_pred={row['dice_pred']:.3f} PG_pred={row['pointing_pred']:.3f}",
            fontsize=10
        )

        axes[0].imshow(img_gray, cmap="gray"); axes[0].set_title("Original"); axes[0].axis("off")
        axes[1].imshow(mask_overlay); axes[1].set_title("Image + Mask overlay"); axes[1].axis("off")
        axes[2].imshow(ov_true); axes[2].set_title("Grad-CAM (true)"); axes[2].axis("off")
        axes[3].imshow(ov_pred); axes[3].set_title("Grad-CAM (pred)"); axes[3].axis("off")

        plt.tight_layout()
        plt.show()

    gradcam.remove_hooks()

# -------------------------
# Run
# -------------------------
df_test,  summary_test  = summarize_scores_by_class(TESTING_CSV,  "testing")
df_train, summary_train = summarize_scores_by_class(TRAINING_CSV, "training")

# Global (micro + macro)
global_raw_test,  global_pretty_test  = compute_global_metrics(df_test,  "testing")
global_raw_train, global_pretty_train = compute_global_metrics(df_train, "training")

# Load model once + plot best per class
model, target_layer, class_names = load_model_from_ckpt(CKPT_FIXED, df_for_fallback=df_test)

df_best_test  = best_rows_per_class(df_test,  RANK_BY)
df_best_train = best_rows_per_class(df_train, RANK_BY)

plot_best_per_class(df_best_test,  model, target_layer, "testing",  IMAGE_SIZE, cam_threshold, alpha_overlay, mask_alpha, RANK_BY)
plot_best_per_class(df_best_train, model, target_layer, "training", IMAGE_SIZE, cam_threshold, alpha_overlay, mask_alpha, RANK_BY)
```

```python
# =========================
# Recompute Grad-CAM recall sweeps + add GLOBAL micro + 95% bootstrap CI for global_micro.
# Outputs:
#   - summary_with_ci.csv (per class + global micro; CI columns only for global_micro rows)
#   - prints pivots for TESTING/TRAINING means (including global_micro)
#   - prints an extra pivot for TESTING with global_micro shown as "mean [ci_lo, ci_hi]"
# =========================

import numpy as np
import pandas as pd
from pathlib import Path
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F

# -------------------------
# Config
# -------------------------
OUT_DIR = Path("/content/localdata/new1")
TESTING_CSV  = OUT_DIR / "xai_gradcam_metrics_testing.csv"
TRAINING_CSV = OUT_DIR / "xai_gradcam_metrics_training.csv"

CKPT_FIXED = "/content/drive/MyDrive/TESIS/StyleGan2/CNNResnet50kfoldinge1/inge1.3fold/full_train/final_full.ckpt"

IMAGE_SIZE = 256
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("DEVICE:", DEVICE)

# Sweeps
TOPK_LIST = [0.01, 0.03, 0.05, 0.10, 0.15]
CAM_THR_LIST = [0.5, 0.7, 0.8]

# Choose CAM target
EVAL_CAM = "true"  # "true" or "pred"

# Bootstrap CI config (global_micro only)
BOOT_N = 2000
CI_ALPHA = 0.05
BOOT_SEED = 123

# Save outputs
OUT_SUMMARY_CSV = OUT_DIR / "recall_sweeps_summary_with_global_micro_ci.csv"

# -------------------------
# I/O helpers
# -------------------------
def load_img_tensor_3ch(img_path: Path, size: int = 256) -> torch.Tensor:
    img = Image.open(img_path).convert("L").resize((size, size), Image.BILINEAR)
    arr = (np.array(img).astype(np.float32) / 255.0)  # [H,W]
    x = torch.from_numpy(arr)[None, None, :, :].float()  # [1,1,H,W]
    x = x.repeat(1, 3, 1, 1)  # [1,3,H,W]
    return x

def load_mask_bool(mask_path: Path, size: int = 256) -> np.ndarray:
    m = Image.open(mask_path).convert("L").resize((size, size), Image.NEAREST)
    return ((np.array(m).astype(np.float32) / 255.0) >= 0.5)

# -------------------------
# Metrics
# -------------------------
def recall_at_topk(cam: np.ndarray, mask: np.ndarray, topk: float) -> float:
    if mask.sum() == 0:
        return float("nan")
    H, W = cam.shape
    k = max(1, int(round(topk * H * W)))
    flat = cam.reshape(-1)
    idx = np.argpartition(flat, -k)[-k:]
    topk_bin = np.zeros_like(flat, dtype=bool)
    topk_bin[idx] = True
    topk_bin = topk_bin.reshape(H, W)
    return float((topk_bin & mask).sum() / mask.sum())

def recall_at_thr(cam: np.ndarray, mask: np.ndarray, thr: float) -> float:
    if mask.sum() == 0:
        return float("nan")
    sel = cam >= thr
    return float((sel & mask).sum() / mask.sum())

# -------------------------
# Bootstrap CI for mean
# -------------------------
def bootstrap_ci_mean(values: np.ndarray, n_boot: int = 2000, alpha: float = 0.05, seed: int = 123):
    v = values[np.isfinite(values)]
    if v.size == 0:
        return float("nan"), float("nan")
    rng = np.random.default_rng(seed)
    n = v.size
    boot_means = np.empty(n_boot, dtype=np.float32)
    for i in range(n_boot):
        samp = rng.choice(v, size=n, replace=True)
        boot_means[i] = samp.mean()
    lo = float(np.quantile(boot_means, alpha / 2.0))
    hi = float(np.quantile(boot_means, 1.0 - alpha / 2.0))
    return lo, hi

def fmt3(x: float) -> str:
    return "nan" if (x is None or not np.isfinite(x)) else f"{x:.3f}"

# -------------------------
# Load model
# -------------------------
ckpt = torch.load(CKPT_FIXED, map_location=DEVICE)

class_names = ckpt.get("class_names", None)
if class_names is None:
    class_names = ["glioma", "meningioma", "pituitary"]
num_classes = len(class_names)

# NOTE: build_model and GradCAM must exist in your environment
model = build_model(num_classes=num_classes, pretrained=False)
state = ckpt["model"] if isinstance(ckpt, dict) and "model" in ckpt else ckpt
model.load_state_dict(state)
model.to(DEVICE).eval()

target_layer = model.layer4[-1] if isinstance(model.layer4, nn.Sequential) else model.layer4
gradcam = GradCAM(model, target_layer, threshold=0.0)

def compute_cam(x: torch.Tensor, class_idx: int) -> np.ndarray:
    cam = gradcam(x, target_class=torch.tensor([class_idx], device=DEVICE))  # [1,h,w] or [h,w]
    if cam.ndim == 3:
        cam = cam[0]
    cam = cam.unsqueeze(0).unsqueeze(0)  # [1,1,h,w]
    cam = F.interpolate(cam, size=(IMAGE_SIZE, IMAGE_SIZE), mode="bilinear", align_corners=False)[0, 0]
    cam = cam.detach().cpu().numpy().astype(np.float32)
    cam = cam - cam.min()
    cam = cam / (cam.max() + 1e-8)
    return cam

# -------------------------
# Evaluate split
# -------------------------
def evaluate_split(csv_path: Path, split_name: str) -> pd.DataFrame:
    df = pd.read_csv(csv_path)
    required = ["true_label", "true_idx", "pred_idx", "image_path", "mask_path"]
    for c in required:
        if c not in df.columns:
            raise ValueError(f"[{split_name}] CSV missing column '{c}'. Columns: {list(df.columns)}")

    rows = []
    for _, r in df.iterrows():
        img_path = Path(r["image_path"])
        mask_path = Path(r["mask_path"])
        if (not img_path.exists()) or (not mask_path.exists()):
            continue

        x = load_img_tensor_3ch(img_path, IMAGE_SIZE).to(DEVICE)
        mask = load_mask_bool(mask_path, IMAGE_SIZE)

        true_idx = int(r["true_idx"])
        pred_idx = int(r["pred_idx"])

        cam_target_idx = true_idx if EVAL_CAM == "true" else pred_idx
        cam = compute_cam(x, cam_target_idx)

        cls = r["true_label"]

        for topk in TOPK_LIST:
            rows.append({
                "split": split_name,
                "true_label": cls,
                "metric": f"recall_topk_{topk:.2f}",
                "value": recall_at_topk(cam, mask, topk)
            })

        for thr in CAM_THR_LIST:
            rows.append({
                "split": split_name,
                "true_label": cls,
                "metric": f"recall_thr_{thr:.2f}",
                "value": recall_at_thr(cam, mask, thr)
            })

    return pd.DataFrame(rows)

# -------------------------
# Run + summarize
# -------------------------
out_test = evaluate_split(TESTING_CSV, "testing")
out_train = evaluate_split(TRAINING_CSV, "training")
out_all = pd.concat([out_test, out_train], ignore_index=True)

# Per-class
summary_cls = (out_all.groupby(["split", "true_label", "metric"])
                    .agg(mean=("value", "mean"),
                         std=("value", "std"),
                         n=("value", "count"))
                    .reset_index())

# Global micro (across all classes)
summary_micro = (out_all.groupby(["split", "metric"])
                      .agg(mean=("value", "mean"),
                           std=("value", "std"),
                           n=("value", "count"))
                      .reset_index())
summary_micro["true_label"] = "global_micro"

# Add bootstrap CI to global_micro rows
ci_rows = []
for (split, metric), g in out_all.groupby(["split", "metric"]):
    vals = g["value"].to_numpy(dtype=np.float32)
    lo, hi = bootstrap_ci_mean(vals, n_boot=BOOT_N, alpha=CI_ALPHA, seed=BOOT_SEED)
    ci_rows.append({
        "split": split,
        "true_label": "global_micro",
        "metric": metric,
        "ci_lo": lo,
        "ci_hi": hi
    })
df_ci = pd.DataFrame(ci_rows)

# Merge CI into global_micro summary; for classes, CI stays NaN
summary_micro = summary_micro.merge(df_ci, on=["split", "true_label", "metric"], how="left")
summary_cls["ci_lo"] = np.nan
summary_cls["ci_hi"] = np.nan

summary = (pd.concat([summary_cls, summary_micro], ignore_index=True)
             .sort_values(["split", "true_label", "metric"])
             .reset_index(drop=True))

# -------------------------
# Print pivots (mean) - same as before
# -------------------------
print("\n=== Recall sweeps summary (mean ± std) + global_micro CI columns ===")
display(summary)

print("\n=== Pivot (mean) TESTING ===")
pivot_mean_testing = summary[summary["split"] == "testing"].pivot_table(
    index="true_label", columns="metric", values="mean"
)
display(pivot_mean_testing)

print("\n=== Pivot (mean) TRAINING ===")
pivot_mean_training = summary[summary["split"] == "training"].pivot_table(
    index="true_label", columns="metric", values="mean"
)
display(pivot_mean_training)

# -------------------------
# NEW: Pivot with CI for TESTING (only global_micro shows mean [lo, hi])
# -------------------------
print("\n=== Pivot (TESTING) with global_micro shown as mean [ci_lo, ci_hi] ===")

# Start from the same mean pivot, but as strings
pivot_ci_testing = pivot_mean_testing.copy()
pivot_ci_testing = pivot_ci_testing.applymap(lambda x: fmt3(float(x)) if np.isfinite(x) else "nan")

# Fill global_micro row with mean [lo, hi] using summary rows
g_rows = summary[(summary["split"] == "testing") & (summary["true_label"] == "global_micro")].copy()
g_map = g_rows.set_index("metric")[["mean", "ci_lo", "ci_hi"]].to_dict(orient="index")

if "global_micro" in pivot_ci_testing.index:
    for metric in pivot_ci_testing.columns:
        if metric in g_map:
            m = g_map[metric]["mean"]
            lo = g_map[metric]["ci_lo"]
            hi = g_map[metric]["ci_hi"]
            pivot_ci_testing.loc["global_micro", metric] = f"{fmt3(m)} [{fmt3(lo)}, {fmt3(hi)}]"

display(pivot_ci_testing)

# Save
summary.to_csv(OUT_SUMMARY_CSV, index=False)
print(f"\nSaved: {OUT_SUMMARY_CSV}")

gradcam.remove_hooks()
print("\nDone.")
```

```python
# =========================
# Grad-CAM evaluation with TOP-K ONLY
# Metrics:
#   - recall_topk_XX
#   - iou_topk_XX
#   - dice_topk_XX
# Includes:
#   - per-class + global_micro
#   - 95% bootstrap CI for global_micro
# Outputs:
#   - gradcam_topk_recall_iou_dice_summary_with_ci.csv
#   - gradcam_topk_metrics_classes_global_ci_testing.csv (compact paper-ready)
# =========================

import numpy as np
import pandas as pd
from pathlib import Path
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F

# -------------------------
# Config
# -------------------------
OUT_DIR = Path("/content/localdata/new1")
TESTING_CSV  = OUT_DIR / "xai_gradcam_metrics_testing.csv"
TRAINING_CSV = OUT_DIR / "xai_gradcam_metrics_training.csv"

CKPT_FIXED = "/content/drive/MyDrive/TESIS/StyleGan2/CNNResnet50kfoldinge1/inge1.3fold/full_train/final_full.ckpt"

IMAGE_SIZE = 256
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("DEVICE:", DEVICE)

# TOP-K sweep only
TOPK_LIST = [0.01, 0.03, 0.05, 0.10, 0.15]

# CAM target
EVAL_CAM = "true"  # "true" or "pred"

# Bootstrap CI (global_micro)
BOOT_N = 2000
CI_ALPHA = 0.05
BOOT_SEED = 123

OUT_SUMMARY_CSV = OUT_DIR / "gradcam_topk_recall_iou_dice_summary_with_ci.csv"

# -------------------------
# I/O helpers
# -------------------------
def load_img_tensor_3ch(img_path: Path, size: int = 256) -> torch.Tensor:
    img = Image.open(img_path).convert("L").resize((size, size), Image.BILINEAR)
    arr = np.array(img).astype(np.float32) / 255.0
    x = torch.from_numpy(arr)[None, None, :, :]
    x = x.repeat(1, 3, 1, 1)
    return x

def load_mask_bool(mask_path: Path, size: int = 256) -> np.ndarray:
    m = Image.open(mask_path).convert("L").resize((size, size), Image.NEAREST)
    return (np.array(m).astype(np.float32) / 255.0) >= 0.5

# -------------------------
# Metrics (TOP-K only)
# -------------------------
def topk_binary(cam: np.ndarray, topk: float) -> np.ndarray:
    """Return boolean mask selecting top-k% pixels of cam."""
    H, W = cam.shape
    k = max(1, int(round(topk * H * W)))
    flat = cam.reshape(-1)
    idx = np.argpartition(flat, -k)[-k:]
    sel = np.zeros_like(flat, dtype=bool)
    sel[idx] = True
    return sel.reshape(H, W)

def recall_topk(cam: np.ndarray, mask: np.ndarray, topk: float) -> float:
    """Recall = |A ∩ M| / |M|"""
    if mask.sum() == 0:
        return float("nan")
    sel = topk_binary(cam, topk)
    return float((sel & mask).sum() / mask.sum())

def iou_topk(cam: np.ndarray, mask: np.ndarray, topk: float) -> float:
    """IoU = |A ∩ M| / |A ∪ M|"""
    if mask.sum() == 0:
        return float("nan")
    sel = topk_binary(cam, topk)
    inter = (sel & mask).sum()
    union = (sel | mask).sum()
    return float(inter / union) if union > 0 else float("nan")

def dice_topk(cam: np.ndarray, mask: np.ndarray, topk: float) -> float:
    """Dice = 2|A ∩ M| / (|A| + |M|)"""
    if mask.sum() == 0:
        return float("nan")
    sel = topk_binary(cam, topk)
    inter = (sel & mask).sum()
    denom = sel.sum() + mask.sum()
    return float((2.0 * inter) / denom) if denom > 0 else float("nan")

# -------------------------
# Bootstrap CI
# -------------------------
def bootstrap_ci_mean(values: np.ndarray, n_boot=2000, alpha=0.05, seed=123):
    v = values[np.isfinite(values)]
    if len(v) == 0:
        return float("nan"), float("nan")
    rng = np.random.default_rng(seed)
    boot = rng.choice(v, size=(n_boot, len(v)), replace=True).mean(axis=1)
    lo = np.quantile(boot, alpha / 2)
    hi = np.quantile(boot, 1 - alpha / 2)
    return float(lo), float(hi)

# -------------------------
# Load model
# -------------------------
ckpt = torch.load(CKPT_FIXED, map_location=DEVICE)

class_names = ckpt.get("class_names", ["glioma", "meningioma", "pituitary"])
num_classes = len(class_names)

# NOTE: build_model and GradCAM must exist in your environment
model = build_model(num_classes=num_classes, pretrained=False)
state = ckpt["model"] if isinstance(ckpt, dict) and "model" in ckpt else ckpt
model.load_state_dict(state)
model.to(DEVICE).eval()

target_layer = model.layer4[-1]
gradcam = GradCAM(model, target_layer)

def compute_cam(x: torch.Tensor, class_idx: int) -> np.ndarray:
    cam = gradcam(x, target_class=torch.tensor([class_idx], device=DEVICE))
    if cam.ndim == 3:
        cam = cam[0]
    cam = cam.unsqueeze(0).unsqueeze(0)
    cam = F.interpolate(
        cam, size=(IMAGE_SIZE, IMAGE_SIZE),
        mode="bilinear", align_corners=False
    )[0, 0]
    cam = cam.detach().cpu().numpy().astype(np.float32)
    cam = cam - cam.min()
    cam = cam / (cam.max() + 1e-8)
    return cam

# -------------------------
# Evaluate split
# -------------------------
def evaluate_split(csv_path: Path, split_name: str) -> pd.DataFrame:
    df = pd.read_csv(csv_path)
    required = ["true_label", "true_idx", "pred_idx", "image_path", "mask_path"]
    for c in required:
        if c not in df.columns:
            raise ValueError(f"[{split_name}] CSV missing column '{c}'. Columns: {list(df.columns)}")

    rows = []
    for _, r in df.iterrows():
        img_path = Path(r["image_path"])
        mask_path = Path(r["mask_path"])
        if not img_path.exists() or not mask_path.exists():
            continue

        x = load_img_tensor_3ch(img_path, IMAGE_SIZE).to(DEVICE)
        mask = load_mask_bool(mask_path, IMAGE_SIZE)

        true_idx = int(r["true_idx"])
        pred_idx = int(r["pred_idx"])
        cam_idx = true_idx if EVAL_CAM == "true" else pred_idx

        cam = compute_cam(x, cam_idx)
        cls = r["true_label"]

        for k in TOPK_LIST:
            rows.append({
                "split": split_name,
                "true_label": cls,
                "metric": f"recall_topk_{k:.2f}",
                "value": recall_topk(cam, mask, k)
            })
            rows.append({
                "split": split_name,
                "true_label": cls,
                "metric": f"iou_topk_{k:.2f}",
                "value": iou_topk(cam, mask, k)
            })
            rows.append({
                "split": split_name,
                "true_label": cls,
                "metric": f"dice_topk_{k:.2f}",
                "value": dice_topk(cam, mask, k)
            })

    return pd.DataFrame(rows)

# -------------------------
# Run evaluation
# -------------------------
df_test = evaluate_split(TESTING_CSV, "testing")
df_train = evaluate_split(TRAINING_CSV, "training")
df_all = pd.concat([df_test, df_train], ignore_index=True)

# Per-class summary
summary_cls = (
    df_all.groupby(["split", "true_label", "metric"])
          .agg(mean=("value", "mean"),
               std=("value", "std"),
               n=("value", "count"))
          .reset_index()
)

# Global micro
summary_micro = (
    df_all.groupby(["split", "metric"])
          .agg(mean=("value", "mean"),
               std=("value", "std"),
               n=("value", "count"))
          .reset_index()
)
summary_micro["true_label"] = "global_micro"

# Bootstrap CI for global_micro (all metrics present)
ci_rows = []
for (split, metric), g in df_all.groupby(["split", "metric"]):
    lo, hi = bootstrap_ci_mean(g["value"].to_numpy(dtype=np.float32), BOOT_N, CI_ALPHA, BOOT_SEED)
    ci_rows.append({
        "split": split,
        "true_label": "global_micro",
        "metric": metric,
        "ci_lo": lo,
        "ci_hi": hi
    })
df_ci = pd.DataFrame(ci_rows)

summary_micro = summary_micro.merge(df_ci, on=["split", "true_label", "metric"], how="left")
summary_cls["ci_lo"] = np.nan
summary_cls["ci_hi"] = np.nan

summary = pd.concat([summary_cls, summary_micro], ignore_index=True)
summary = summary.sort_values(["split", "true_label", "metric"]).reset_index(drop=True)

# -------------------------
# Save + display
# -------------------------
summary.to_csv(OUT_SUMMARY_CSV, index=False)
print(f"Saved: {OUT_SUMMARY_CSV}")

display(summary)

gradcam.remove_hooks()
print("Done.")

# =========================
# Compact DF: per-class + global with CI (paper-ready)
# =========================
def build_class_global_df(summary_df: pd.DataFrame, split_name: str) -> pd.DataFrame:
    """
    Returns a compact DataFrame with:
      true_label | metric | mean | ci_lo | ci_hi
    For classes: ci_lo/ci_hi = NaN
    For global_micro: ci_lo/ci_hi filled
    """
    df = summary_df[summary_df["split"] == split_name].copy()

    # Keep only top-k metrics (cleaner for paper)
    df = df[df["metric"].str.contains("topk")]

    cols = ["true_label", "metric", "mean", "ci_lo", "ci_hi"]
    df = df[cols].sort_values(["metric", "true_label"])
    return df.reset_index(drop=True)

# Build DF for TESTING (recommended for paper)
df_xai_testing = build_class_global_df(summary, "testing")

print("\n=== Compact XAI DF (TESTING): classes + global with CI ===")
display(df_xai_testing)

# Optional: save
OUT_XAI_TABLE = OUT_DIR / "gradcam_topk_metrics_classes_global_ci_testing.csv"
df_xai_testing.to_csv(OUT_XAI_TABLE, index=False)
print(f"Saved: {OUT_XAI_TABLE}")
```
